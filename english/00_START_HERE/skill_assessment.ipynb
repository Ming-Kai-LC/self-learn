## Part 1: Understanding CEFR Levels

The **Common European Framework of Reference for Languages (CEFR)** is the international standard for describing language ability. Let's understand each level:

### CEFR Level Breakdown

| Level | Name | Can Do... |
|-------|------|----------|
| **A1** | Beginner | Understand basic phrases, introduce yourself, ask simple questions |
| **A2** | Elementary | Handle simple exchanges, describe background, immediate needs |
| **B1** | Intermediate | Deal with most travel situations, describe experiences, explain opinions |
| **B2** | Upper-Intermediate | Interact fluently with natives, understand complex texts, present clear arguments |
| **C1** | Advanced | Express ideas fluently, use language flexibly, understand implicit meanings |
| **C2** | Proficient | Understand virtually everything, express yourself spontaneously and precisely |

### Your Learning Journey

This program focuses on the **B1 â†’ B2 â†’ C1** progression, which typically takes:

- **B1 â†’ B2**: 5-7 months with 12 hours/week practice
- **B2 â†’ C1**: Additional 5-11 months (total 12-18 months from B1)

### Active Vocabulary by Level

- **B1**: 2,000-2,500 words
- **B2**: 3,000-4,000 words (including professional terms)
- **C1**: 4,000-8,000 words (including idioms)

## Part 2: Self-Assessment Quiz

Before taking an official test, let's do a quick self-assessment to estimate your current level.

```python
# Self-assessment questions aligned with CEFR descriptors
# Rate yourself 1-5 (1=Strongly Disagree, 5=Strongly Agree)


def self_assessment():
    """
    Quick self-assessment to estimate CEFR level.
    This helps you understand where to start before taking official tests.
    """
    print("=" * 70)
    print("SELF-ASSESSMENT: Rate each statement 1-5")
    print("1 = Strongly Disagree | 5 = Strongly Agree")
    print("=" * 70)
    print()

    # Questions organized by skill area
    questions = {
        "Speaking": [
            "I can handle familiar work and travel situations in English",
            "I can interact with native speakers without strain for both parties",
            "I can express ideas fluently without obviously searching for words",
        ],
        "Listening": [
            "I understand clear speech on familiar topics",
            "I understand 80-90% of native conversations at normal speed",
            "I understand demanding lectures and complex discussions",
        ],
        "Reading": [
            "I can read straightforward texts on familiar subjects",
            "I can read complex texts including technical material in my field",
            "I can understand subtle distinctions and implicit meanings",
        ],
        "Writing": [
            "I can write simple connected texts on familiar topics",
            "I can write clear, detailed texts on diverse subjects",
            "I can write clear, well-structured texts with appropriate style",
        ],
        "Vocabulary": [
            "I have about 2,000-2,500 active words",
            "I have 3,000-4,000 words including professional terms",
            "I have 4,000-8,000 words including idioms and cultural references",
        ],
    }

    # Note: In interactive version, this would collect user input
    # For now, we'll demonstrate the structure

    print("\nðŸ“ EXERCISE: Complete this self-assessment honestly.")
    print("This is for your own benefit - there are no wrong answers!\n")

    total_score = 0
    skill_scores = {}

    for skill, skill_questions in questions.items():
        print(f"\n--- {skill} ---")
        skill_score = 0

        for i, question in enumerate(skill_questions, 1):
            print(f"{i}. {question}")
            # In interactive version: score = int(input("   Your rating (1-5): "))
            # For demonstration:
            score = 0  # Placeholder - user would input their score
            skill_score += score

        skill_scores[skill] = skill_score
        total_score += skill_score

    return total_score, skill_scores


# Run self-assessment (demonstration)
print("\nðŸŽ¯ Complete the self-assessment above by rating each statement 1-5.")
print("Then run the scoring function below to estimate your level.\n")
```

```python
def interpret_self_assessment(total_score):
    """
    Interpret self-assessment score to estimate CEFR level.

    Args:
        total_score: Sum of all ratings (15-75 range)

    Returns:
        Estimated CEFR level and recommendations
    """
    # Score ranges for different levels
    # Each question is rated 1-5, with 15 questions total (max 75 points)

    if total_score >= 60:
        level = "C1-C2"
        description = "Advanced/Proficient"
        recommendation = "Focus on C1-C2 materials, native content, and subtle nuances."
    elif total_score >= 45:
        level = "B2-C1"
        description = "Upper-Intermediate to Advanced"
        recommendation = "Focus on B2-C1 transition, complex topics, and professional fluency."
    elif total_score >= 30:
        level = "B1-B2"
        description = "Intermediate to Upper-Intermediate"
        recommendation = "This program is perfect for you! Focus on B1â†’B2 advancement."
    elif total_score >= 20:
        level = "A2-B1"
        description = "Elementary to Intermediate"
        recommendation = "Start with B1 consolidation, then progress to B2 content."
    else:
        level = "A1-A2"
        description = "Beginner to Elementary"
        recommendation = "Consider starting with A2 materials before this program."

    print("\n" + "=" * 70)
    print("SELF-ASSESSMENT RESULTS")
    print("=" * 70)
    print(f"Total Score: {total_score}/75")
    print(f"Estimated Level: {level} ({description})")
    print(f"\nðŸ’¡ Recommendation: {recommendation}")
    print("=" * 70)

    print("\nâš ï¸  IMPORTANT: This is just an estimate!")
    print("Please take an official assessment (next section) for accurate results.")

    return level


# Example usage (replace with your actual total score)
# example_score = 35  # Replace this with your total from above
# estimated_level = interpret_self_assessment(example_score)
```

```python
# Record your baseline assessment results


def save_baseline_results(
    test_name,
    overall_score,
    reading_score=None,
    listening_score=None,
    speaking_score=None,
    writing_score=None,
    notes="",
):
    """
    Save your baseline assessment results for progress tracking.

    Args:
        test_name: Name of test (e.g., "EF SET", "British Council")
        overall_score: Your overall CEFR level (e.g., "B1", "B2")
        reading_score: Reading section score (optional)
        listening_score: Listening section score (optional)
        speaking_score: Speaking section score (optional)
        writing_score: Writing section score (optional)
        notes: Any additional observations
    """
    baseline_data = {
        "date": datetime.now().isoformat(),
        "test_name": test_name,
        "overall_level": overall_score,
        "breakdown": {
            "reading": reading_score,
            "listening": listening_score,
            "speaking": speaking_score,
            "writing": writing_score,
        },
        "notes": notes,
    }

    # Save to JSON file
    baseline_file = progress_dir / "baseline_assessment.json"

    with open(baseline_file, "w", encoding="utf-8") as f:
        json.dump(baseline_data, f, indent=2)

    print("âœ… Baseline results saved successfully!")
    print(f"\nYour starting point: {overall_score}")
    print(f"Date: {datetime.now().strftime('%Y-%m-%d')}")
    print(f"\nResults saved to: {baseline_file}")

    return baseline_data


# Example usage - REPLACE WITH YOUR ACTUAL SCORES:
# my_baseline = save_baseline_results(
#     test_name="EF SET",
#     overall_score="B1",
#     reading_score="52",
#     listening_score="48",
#     notes="Struggled with fast native speech, reading was stronger"
# )

print("\nðŸ“ Uncomment the code above and enter your actual test results.")
```

## Part 4: Setting Your Learning Goals

Based on your baseline assessment, let's set realistic, measurable goals.

```python
# Goal-setting framework based on current level


def create_learning_plan(current_level, weekly_hours=12):
    """
    Create a personalized learning plan based on current CEFR level.

    Args:
        current_level: Your baseline CEFR level (e.g., "B1", "B2")
        weekly_hours: Hours you can dedicate per week (default: 12)

    Returns:
        Dictionary with timeline and milestones
    """
    # Timeline estimates based on research (with 12 hours/week)
    timelines = {
        "A2": {
            "to_B1": "3-4 months",
            "to_B2": "8-11 months",
            "to_C1": "15-22 months",
        },
        "B1": {
            "to_B2": "5-7 months",
            "to_C1": "12-18 months",
        },
        "B2": {
            "to_C1": "7-11 months",
            "to_C2": "15-24 months",
        },
    }

    # Adjust timeline if different weekly hours
    time_multiplier = 12 / weekly_hours if weekly_hours != 12 else 1.0

    print("=" * 70)
    print(f"YOUR PERSONALIZED LEARNING PLAN (Starting Level: {current_level})")
    print("=" * 70)
    print(f"\nâ° Weekly commitment: {weekly_hours} hours")

    if current_level in timelines:
        print(f"\nðŸ“… Estimated Timeline:")
        for target, duration in timelines[current_level].items():
            print(f"   â€¢ {current_level} â†’ {target.replace('to_', '')}: {duration}")
            if time_multiplier != 1.0:
                print(
                    f"     (Adjusted for {weekly_hours}h/week: ~{duration} Ã— {time_multiplier:.1f})"
                )

    # Recommended focus areas by level
    focus_areas = {
        "A2": [
            "Basic grammar consolidation (tenses, questions, negatives)",
            "Core vocabulary (1,500-2,000 common words)",
            "Simple conversation practice",
            "Building confidence and fluency",
        ],
        "B1": [
            "Grammar refinement (conditionals, passive, reported speech)",
            "Business vocabulary expansion (2,500-3,500 words)",
            "Professional speaking scenarios",
            "Writing longer, structured texts",
        ],
        "B2": [
            "Advanced grammar mastery (mixed conditionals, subjunctive)",
            "Idiomatic expressions and collocations (4,000-6,000 words)",
            "Abstract discussion and debate",
            "Sophisticated writing with style variation",
        ],
    }

    if current_level in focus_areas:
        print(f"\nðŸŽ¯ Recommended Focus Areas:")
        for i, area in enumerate(focus_areas[current_level], 1):
            print(f"   {i}. {area}")

    # Assessment schedule
    print(f"\nðŸ“Š Progress Assessment Schedule:")
    print(f"   â€¢ Week 8: First progress check (mini-assessment)")
    print(f"   â€¢ Month 3-4: Mid-term EF SET re-test")
    print(f"   â€¢ Month 6-7: Check for level progression")
    print(f"   â€¢ Month 12: Major milestone assessment")

    print("\n" + "=" * 70)

    return {
        "current_level": current_level,
        "weekly_hours": weekly_hours,
        "timeline": timelines.get(current_level, {}),
        "focus_areas": focus_areas.get(current_level, []),
    }


# Example usage - REPLACE WITH YOUR LEVEL:
# my_plan = create_learning_plan(current_level="B1", weekly_hours=12)

print("\nðŸ“ Run create_learning_plan() with your baseline level to see your plan.")
```

```python
# Save your complete learning plan


def save_learning_plan(
    baseline_level, target_level, weekly_hours, target_date=None, personal_goals=None
):
    """
    Save your learning plan for ongoing tracking.

    Args:
        baseline_level: Your starting CEFR level
        target_level: Your goal CEFR level
        weekly_hours: Hours committed per week
        target_date: When you want to achieve your goal (optional)
        personal_goals: List of specific personal goals (optional)
    """
    plan_data = {
        "created_date": datetime.now().isoformat(),
        "baseline_level": baseline_level,
        "target_level": target_level,
        "weekly_commitment_hours": weekly_hours,
        "target_completion_date": target_date,
        "personal_goals": personal_goals or [],
        "milestones": [
            {"week": 8, "goal": "Complete B1 consolidation, first progress check"},
            {"month": 3, "goal": "Mid-term assessment, evaluate B2 readiness"},
            {"month": 6, "goal": "Solid B2 skills, professional fluency"},
            {"month": 12, "goal": f"Achieve {target_level} level certification"},
        ],
    }

    plan_file = progress_dir / "learning_plan.json"

    with open(plan_file, "w", encoding="utf-8") as f:
        json.dump(plan_data, f, indent=2)

    print("âœ… Learning plan saved successfully!")
    print(f"\nðŸŽ¯ Goal: {baseline_level} â†’ {target_level}")
    print(f"â° Weekly commitment: {weekly_hours} hours")

    if personal_goals:
        print(f"\nðŸ’ª Your Personal Goals:")
        for goal in personal_goals:
            print(f"   â€¢ {goal}")

    print(f"\nðŸ“ Plan saved to: {plan_file}")

    return plan_data


# Example usage - CUSTOMIZE WITH YOUR GOALS:
# my_learning_plan = save_learning_plan(
#     baseline_level="B1",
#     target_level="C1",
#     weekly_hours=12,
#     target_date="2026-12-31",
#     personal_goals=[
#         "Lead meetings confidently in English",
#         "Watch movies without subtitles",
#         "Write professional reports and proposals",
#         "Interact naturally with native speakers",
#     ]
# )

print("\nðŸ“ Customize and run save_learning_plan() with your specific goals.")
```
