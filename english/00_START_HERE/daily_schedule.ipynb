## Part 2: Daily Foundation (Monday-Friday)

Let's break down each daily session with specific activities.

### Morning Routine (20 minutes)

**Best time**: Right after waking up or with breakfast

**Activities**:
1. **Vocabulary building (10 min)**
   - Use WordUp app or Memrise
   - Learn 20-30 new words weekly
   - Focus on business/professional contexts
   - Review yesterday's words with spaced repetition

2. **Quick reading (10 min)**
   - Read one article from Breaking News English (at your level)
   - Or News in Levels (Level 2-3)
   - Save 5-10 new vocabulary words for later AI practice

**Goal**: Start your day with English, build vocabulary systematically

---

### Lunch Break (15 minutes)

**Best time**: During lunch or commute

**Activities**:
- **Podcast listening with transcripts**
  - British Council LearnEnglish Podcasts (A2-B1)
  - Practising English by Mike Bilbrough (B1-B2)
  - Real Exam English (B2+) - features 6 different English accents!
  - 6 Minute English (perfect for 15-minute slots)

**Strategy**:
- First listen: With transcript
- Second listen: Without transcript (next day)
- Note down interesting phrases for evening AI practice

**Goal**: Train your ear for natural speech, build listening stamina

---

### Evening Core Practice (45-60 minutes)

**Best time**: After dinner, before bed

**Activities** (rotate daily):

**3. Writing Practice (30 min) - Alternate days**

   - Or Grammarly for instant corrections
   - Practice professional vs. casual tone

**4. Review & Consolidation (15 min) - EVERY DAY**
   - Discuss your learning with ChatGPT or Claude
   - Practice using new vocabulary in context
   - Review AI corrections from earlier exercises
   - Ask questions about confusing grammar points

**Goal**: Active skill development with immediate AI feedback

## Part 4: Daily Practice Tracker

Consistency is key! Let's create a system to track your daily practice.

```python
# Daily practice logging system


def log_daily_practice(
    date, morning_done=False, lunch_done=False, evening_done=False, total_minutes=0, notes=""
):
    """
    Log daily practice completion.

    Args:
        date: Practice date (YYYY-MM-DD format or datetime object)
        morning_done: Completed morning routine (bool)
        lunch_done: Completed lunch podcast (bool)
        evening_done: Completed evening practice (bool)
        total_minutes: Actual minutes practiced
        notes: Any observations or challenges
    """
    if isinstance(date, datetime):
        date_str = date.strftime("%Y-%m-%d")
    else:
        date_str = date

    practice_log = {
        "date": date_str,
        "sessions_completed": {
            "morning": morning_done,
            "lunch": lunch_done,
            "evening": evening_done,
        },
        "total_minutes": total_minutes,
        "notes": notes,
        "logged_at": datetime.now().isoformat(),
    }

    # Load existing logs
    log_file = progress_dir / "daily_practice_log.json"

    if log_file.exists():
        with open(log_file, "r", encoding="utf-8") as f:
            all_logs = json.load(f)
    else:
        all_logs = []

    # Add new log (or update existing date)
    existing_index = next((i for i, log in enumerate(all_logs) if log["date"] == date_str), None)

    if existing_index is not None:
        all_logs[existing_index] = practice_log
    else:
        all_logs.append(practice_log)

    # Save updated logs
    with open(log_file, "w", encoding="utf-8") as f:
        json.dump(all_logs, f, indent=2)

    # Display confirmation
    completed_count = sum(practice_log["sessions_completed"].values())
    print(f"âœ… Practice logged for {date_str}")
    print(f"   Sessions completed: {completed_count}/3")
    print(f"   Total time: {total_minutes} minutes")
    if notes:
        print(f"   Notes: {notes}")

    return practice_log


# Example usage:
# today_practice = log_daily_practice(
#     date=datetime.now(),
#     morning_done=True,
#     lunch_done=True,
#     evening_done=False,
#     total_minutes=35,
#     notes="Missed evening session due to late meeting"
# )


print("\nðŸ“ Use log_daily_practice() each day to track your consistency!")
```

```python
# Weekly summary and streak tracking


def get_practice_stats(days=7):
    """
    Get practice statistics for the last N days.

    Args:
        days: Number of days to analyze (default: 7 for weekly)
    """
    log_file = progress_dir / "daily_practice_log.json"

    if not log_file.exists():
        print("âŒ No practice logs found yet. Start logging your daily practice!")
        return None

    with open(log_file, "r", encoding="utf-8") as f:
        all_logs = json.load(f)

    # Get logs from last N days
    cutoff_date = (datetime.now() - timedelta(days=days)).date()
    recent_logs = [
        log for log in all_logs if datetime.fromisoformat(log["date"]).date() >= cutoff_date
    ]

    if not recent_logs:
        print(f"âŒ No logs found in the last {days} days.")
        return None

    # Calculate statistics
    total_sessions = 0
    total_minutes = 0
    days_practiced = len(recent_logs)

    for log in recent_logs:
        total_sessions += sum(log["sessions_completed"].values())
        total_minutes += log["total_minutes"]

    # Display stats
    print("=" * 70)
    print(f"PRACTICE STATISTICS (Last {days} days)")
    print("=" * 70)
    print(f"\nðŸ“… Days with practice: {days_practiced}/{days}")
    print(f"â±ï¸  Total practice time: {total_minutes} minutes ({total_minutes/60:.1f} hours)")
    print(f"ðŸ“Š Average per day: {total_minutes/days if days > 0 else 0:.0f} minutes")
    print(f"âœ… Sessions completed: {total_sessions}")

    # Weekly goal progress (12 hours = 720 minutes)
    if days == 7:
        weekly_goal = 720  # 12 hours in minutes
        progress_pct = (total_minutes / weekly_goal) * 100
        print(f"\nðŸŽ¯ Weekly Goal Progress: {progress_pct:.1f}% ({total_minutes}/{weekly_goal} min)")

        # Visual progress bar
        bar_length = 50
        filled = int((total_minutes / weekly_goal) * bar_length)
        bar = "â–ˆ" * filled + "â–‘" * (bar_length - filled)
        print(f"   {bar}")

        if total_minutes >= weekly_goal:
            print(f"\nðŸŽ‰ Congratulations! You hit your weekly goal!")
        else:
            remaining = weekly_goal - total_minutes
            print(f"\nðŸ’ª {remaining} minutes to go this week - you can do it!")

    print("\n" + "=" * 70)

    return {
        "days_practiced": days_practiced,
        "total_minutes": total_minutes,
        "total_sessions": total_sessions,
    }


# Example usage:
# weekly_stats = get_practice_stats(days=7)
# monthly_stats = get_practice_stats(days=30)


print("\nðŸ“Š Use get_practice_stats() to see your progress!")
```

## Part 5: Customizing Your Schedule

The 12-hour framework is flexible. Here's how to adapt it to your life.

### Common Adjustments

**1. Can't do mornings?**
- Move morning routine to evening commute
- Or split into two 10-minute blocks during the day

**2. Irregular work schedule?**
- Keep total 12 hours/week but distribute differently
- Example: 4 days Ã— 1.5 hours + 2 days Ã— 3 hours

**3. Weekend work commitments?**
- Do intensive sessions on days off
- Or extend weekday evenings to 90 minutes

**4. Less than 12 hours available?**
- Still progress, just slower (double timeline estimates)
- Prioritize: AI conversations (15 min) + one other skill daily
- Maintain consistency over intensity

### Non-Negotiables

Whatever your schedule, keep these:

âœ… **Daily AI conversation** (even if just 10 minutes)  
âœ… **Weekend longer sessions** (at least one 2-hour block)  
âœ… **All four skills weekly** (speaking, listening, reading, writing)  
âœ… **Consistency** (better 45 min/day than 5 hours once/week)

## Part 6: Overcoming Common Obstacles

### "I'm too tired in the evenings"

**Solutions**:
- Do AI conversation first (most engaging when tired)
- Save writing practice for weekends
- Listen to podcasts (passive, less draining)
- 15 minutes is better than 0 minutes!

### "I miss days and feel guilty"

**Solutions**:
- One missed day â‰  failure
- Just resume next day, don't try to "catch up"
- Aim for 5-6 days/week, not perfection
- Track practice to see you're doing better than you think

### "I can't focus for full sessions"

**Solutions**:
- Break into smaller chunks (5-10 min)
- Use Pomodoro: 25 min work, 5 min break
- Vary activities within session
- Make it fun (topics you care about)

### "My schedule is unpredictable"

**Solutions**:
- Set weekly goal (12 hours), not daily requirements
- Keep "emergency" activities ready (podcast queue, vocabulary app)
- Use waiting time productively (commutes, queues)
- Weekend sessions can compensate for busy weekdays

### "I don't see progress"

**Solutions**:
- Progress is invisible week-to-week, visible month-to-month
- Take monthly assessments for objective data
- Keep error log to see which mistakes disappeared
- Revisit old content to notice how much easier it is now
