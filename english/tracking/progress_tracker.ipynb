## Setup

Run this cell first to import required libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import json
from pathlib import Path

# Create data directory if it doesn't exist
Path('data').mkdir(exist_ok=True)

print('âœ… Setup complete!')
```

## Initialize Your Progress Log

Run this cell ONCE when you first start using this tracker:

```python
# Create empty progress log
progress_log = pd.DataFrame(columns=[
    'date', 'level', 'skill', 'notebook_number', 'notebook_name',
    'time_spent_min', 'difficulty_rating', 'completed', 'notes'
])

# Save to CSV
progress_log.to_csv('data/my_progress.csv', index=False)
print('âœ… Progress log initialized!')
print('ğŸ“ File saved to: data/my_progress.csv')
```

## Log a Completed Notebook

Fill in your details and run this cell after completing each notebook:

```python
# ===== EDIT THESE VALUES =====
level = 'B2'  # A1, A2, B1, B2, C1, or C2
skill = 'writing'  # writing, speaking, listening, or reading
notebook_number = '01'  # 01, 02, 03, etc.
notebook_name = 'Passive Voice Business Writing'  # Brief description
time_spent_min = 45  # How long did it take?
difficulty_rating = 3  # 1=too easy, 3=just right, 5=too hard
completed = True  # True if you finished it
notes = 'Need more practice with complex passive structures'  # Optional notes
# ==============================

# Load existing log
df = pd.read_csv('data/my_progress.csv')

# Add new entry
new_entry = pd.DataFrame([{
    'date': datetime.now().strftime('%Y-%m-%d %H:%M'),
    'level': level,
    'skill': skill,
    'notebook_number': notebook_number,
    'notebook_name': notebook_name,
    'time_spent_min': time_spent_min,
    'difficulty_rating': difficulty_rating,
    'completed': completed,
    'notes': notes
}])

df = pd.concat([df, new_entry], ignore_index=True)
df.to_csv('data/my_progress.csv', index=False)

print(f'âœ… Logged: {level}/{skill}/{notebook_number} - {notebook_name}')
print(f'â±ï¸ Time: {time_spent_min} minutes')
print(f'ğŸ“Š Total notebooks completed: {df[df.completed==True].shape[0]}')
```

## View Your Progress Summary

```python
# Load your progress
df = pd.read_csv('data/my_progress.csv')

# Overall stats
total_notebooks = df[df.completed==True].shape[0]
total_time = df['time_spent_min'].sum()
avg_time = df['time_spent_min'].mean()

print('=== YOUR PROGRESS SUMMARY ===')
print(f'\nğŸ“š Notebooks completed: {total_notebooks}/335 ({total_notebooks/335*100:.1f}%)')
print(f'â±ï¸ Total study time: {total_time/60:.1f} hours')
print(f'âŒš Average time per notebook: {avg_time:.0f} minutes')
print(f'ğŸ“… Study days: {df.date.nunique()}')

# Progress by level
print('\n=== PROGRESS BY LEVEL ===')
level_progress = df[df.completed==True].groupby('level').size()
level_totals = {'A1': 40, 'A2': 48, 'B1': 60, 'B2': 80, 'C1': 60, 'C2': 40}
for level in ['A1', 'A2', 'B1', 'B2', 'C1', 'C2']:
    completed = level_progress.get(level, 0)
    total = level_totals[level]
    pct = (completed/total*100) if total > 0 else 0
    bar = 'â–ˆ' * int(pct/5) + 'â–‘' * (20 - int(pct/5))
    print(f'{level}: {bar} {completed}/{total} ({pct:.0f}%)')

# Progress by skill
print('\n=== PROGRESS BY SKILL ===')
skill_progress = df[df.completed==True].groupby('skill').size()
for skill in ['writing', 'speaking', 'listening', 'reading']:
    completed = skill_progress.get(skill, 0)
    total = 82  # Each skill has 82 notebooks total
    pct = (completed/total*100) if total > 0 else 0
    bar = 'â–ˆ' * int(pct/5) + 'â–‘' * (20 - int(pct/5))
    print(f'{skill}: {bar} {completed}/{total} ({pct:.0f}%)')
```

## Visualize Your Progress

```python
# Load progress
df = pd.read_csv('data/my_progress.csv')
df['date'] = pd.to_datetime(df['date'])

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Progress over time
df_completed = df[df.completed==True].copy()
df_completed['cumulative'] = range(1, len(df_completed)+1)
axes[0,0].plot(df_completed['date'], df_completed['cumulative'], marker='o')
axes[0,0].set_title('Cumulative Progress Over Time', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Notebooks Completed')
axes[0,0].grid(True, alpha=0.3)

# 2. Study time by level
time_by_level = df.groupby('level')['time_spent_min'].sum() / 60
time_by_level.plot(kind='bar', ax=axes[0,1], color='steelblue')
axes[0,1].set_title('Study Time by Level (hours)', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Level')
axes[0,1].set_ylabel('Hours')
axes[0,1].tick_params(axis='x', rotation=0)

# 3. Completion by skill
skill_counts = df[df.completed==True].groupby('skill').size()
skill_counts.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', startangle=90)
axes[1,0].set_title('Notebooks Completed by Skill', fontsize=14, fontweight='bold')
axes[1,0].set_ylabel('')

# 4. Difficulty ratings
difficulty_dist = df['difficulty_rating'].value_counts().sort_index()
difficulty_dist.plot(kind='bar', ax=axes[1,1], color='coral')
axes[1,1].set_title('Difficulty Ratings Distribution', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Rating (1=easy, 5=hard)')
axes[1,1].set_ylabel('Count')
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('data/progress_visualization.png', dpi=300, bbox_inches='tight')
print('âœ… Visualizations saved to: data/progress_visualization.png')
plt.show()
```

## Weekly Report

See what you accomplished this week:

```python
# Load progress
df = pd.read_csv('data/my_progress.csv')
df['date'] = pd.to_datetime(df['date'])

# Filter to last 7 days
last_week = df[df['date'] >= datetime.now() - timedelta(days=7)]

print('=== THIS WEEK\'S PROGRESS ===')
print(f'\nğŸ“š Notebooks completed: {last_week[last_week.completed==True].shape[0]}')
print(f'â±ï¸ Study time: {last_week.time_spent_min.sum()/60:.1f} hours')
print(f'ğŸ“… Study days: {last_week.date.dt.date.nunique()}')
print(f'âŒš Avg time/session: {last_week.time_spent_min.mean():.0f} minutes')

if len(last_week) > 0:
    print('\nğŸ“– Completed this week:')
    for _, row in last_week[last_week.completed==True].iterrows():
        print(f"  â€¢ {row['level']}/{row['skill']}/{row['notebook_number']} - {row['notebook_name']}")
else:
    print('\nâŒ No study sessions this week. Time to get back to it!')
```

## Identify Areas for Improvement

```python
# Load progress
df = pd.read_csv('data/my_progress.csv')

print('=== AREAS FOR IMPROVEMENT ===')

# Skills with fewer completions
skill_counts = df[df.completed==True].groupby('skill').size()
if len(skill_counts) > 0:
    weakest_skill = skill_counts.idxmin()
    print(f'\nğŸ“Š Focus more on: {weakest_skill} ({skill_counts[weakest_skill]} notebooks)')

# Difficult notebooks (rated 4-5)
difficult = df[df.difficulty_rating >= 4]
if len(difficult) > 0:
    print(f'\nğŸ”¥ Challenging notebooks ({len(difficult)} rated as difficult):')
    for _, row in difficult.head(5).iterrows():
        print(f"  â€¢ {row['level']}/{row['skill']}/{row['notebook_number']} - {row['notebook_name']}")
    print('   â†’ Consider reviewing these topics again')

# Notebooks with notes
with_notes = df[df.notes.notna() & (df.notes != '')]
if len(with_notes) > 0:
    print(f'\nğŸ“ Your notes ({len(with_notes)} notebooks):')
    for _, row in with_notes.tail(3).iterrows():
        print(f"  â€¢ {row['level']}/{row['skill']}/{row['notebook_number']}: {row['notes']}")
```

## Set Goals for Next Week

```python
# Calculate recommended weekly goal based on past performance
df = pd.read_csv('data/my_progress.csv')
df['date'] = pd.to_datetime(df['date'])

if len(df) > 0:
    # Calculate average notebooks per week
    days_active = (df['date'].max() - df['date'].min()).days + 1
    weeks_active = max(days_active / 7, 1)
    avg_per_week = len(df) / weeks_active
    
    # Suggest goal (10% increase)
    suggested_goal = int(avg_per_week * 1.1) + 1
    
    print('=== GOALS FOR NEXT WEEK ===')
    print(f'\nğŸ“Š Your average: {avg_per_week:.1f} notebooks/week')
    print(f'ğŸ¯ Suggested goal: {suggested_goal} notebooks')
    print(f'â±ï¸ Estimated time: {suggested_goal * 45} minutes ({suggested_goal * 45 / 60:.1f} hours)')
    print('\nğŸ“ Write your goal here:')
    print(f'   "This week I will complete {suggested_goal} notebooks"')
    print('\nğŸ’¡ Tips:')
    print('   - Study at the same time each day')
    print('   - Mix different skills for variety')
    print('   - Review difficult topics')
    print('   - Practice with AI tools daily')
else:
    print('=== START YOUR FIRST WEEK! ===')
    print('\nğŸ¯ Suggested goal: 3-5 notebooks')
    print('â±ï¸ Time needed: 2-4 hours')
    print('\nFocus on building the habit this week!')
```

## Export Your Progress

Create a shareable summary of your progress:

```python
# Create summary report
df = pd.read_csv('data/my_progress.csv')
df['date'] = pd.to_datetime(df['date'])

summary = {
    'generated_date': datetime.now().strftime('%Y-%m-%d'),
    'total_notebooks_completed': int(df[df.completed==True].shape[0]),
    'total_study_hours': float(df['time_spent_min'].sum() / 60),
    'study_days': int(df.date.dt.date.nunique()),
    'average_time_per_notebook': float(df['time_spent_min'].mean()),
    'progress_by_level': df[df.completed==True].groupby('level').size().to_dict(),
    'progress_by_skill': df[df.completed==True].groupby('skill').size().to_dict()
}

# Save as JSON
with open('data/progress_summary.json', 'w') as f:
    json.dump(summary, f, indent=2)

print('âœ… Progress summary exported!')
print('ğŸ“ Saved to: data/progress_summary.json')
print('\nShare this file to show your progress! ğŸ‰')
```
