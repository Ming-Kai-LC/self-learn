## 1. What is the Present Simple?

The Present Simple tense describes:
- **Habits and routines**: "I drink coffee every morning."
- **Facts and general truths**: "The sun rises in the east."
- **Things that are always or usually true**: "She works in a hospital."

### âš ï¸ Important Note:
Present Simple is NOT used for actions happening right now! (That's Present Continuous, which you'll learn later)

---

## 2. How to Form Present Simple (Affirmative)

### Basic Structure:
```
Subject + Base Verb (+ -s for he/she/it)
```

### Complete Table:

| Person | Verb Form | Example |
|--------|-----------|----------|
| I | **work** | I work every day |
| You | **work** | You work hard |
| He | **works** | He works at home |
| She | **works** | She works in an office |
| It | **works** | It works well |
| We | **work** | We work together |
| They | **work** | They work downtown |

### ğŸ¯ Key Rule:
- **I, you, we, they** = base verb (work, play, study)
- **He, she, it** = base verb + **-s** (works, plays, studies)

## 4. Spelling Rules for Third-Person -S

There are **3 different ways** to add the third-person ending:

### Rule 1: Most verbs â†’ Add -S
```
work â†’ works
play â†’ plays
eat â†’ eats
sleep â†’ sleeps
```

### Rule 2: Verbs ending in -s, -sh, -ch, -x, -o â†’ Add -ES
```
kiss â†’ kisses
wash â†’ washes
watch â†’ watches
fix â†’ fixes
go â†’ goes
do â†’ does
```

### Rule 3: Verbs ending in consonant + y â†’ Change Y to -IES
```
study â†’ studies
try â†’ tries
cry â†’ cries
fly â†’ flies
```

**BUT:** If the verb ends in vowel + y, just add -s:
```
play â†’ plays (NOT plaies)
say â†’ says
enjoy â†’ enjoys
```

## 5. Irregular Verbs: HAVE

The verb **HAVE** is irregular in third person:

| Person | Form |
|--------|------|
| I | **have** |
| You | **have** |
| He/She/It | **has** (NOT haves!) |
| We | **have** |
| They | **have** |

### Examples:
- I **have** a car.
- She **has** three children.
- They **have** a big house.

## 7. Adverbs of Frequency

We often use **adverbs of frequency** with Present Simple to show how often we do things.

### Frequency Scale (100% â†’ 0%):
```
always (100%)    â†’ I always brush my teeth.
usually (80%)    â†’ She usually walks to work.
often (60%)      â†’ We often eat pizza.
sometimes (40%)  â†’ He sometimes watches TV.
rarely (20%)     â†’ They rarely go out.
never (0%)       â†’ I never smoke.
```

### Position in Sentence:
**BEFORE the main verb:**
- I **always** eat breakfast.
- She **usually** works late.
- They **sometimes** play tennis.

**AFTER the verb "to be":**
- I am **always** happy.
- He is **usually** late.
- We are **sometimes** tired.

## 8. Common Time Expressions

Present Simple often appears with these time expressions:

### Frequency:
- **every** day/week/month/year
- **once** a week
- **twice** a day
- **three times** a week

### Days/Times:
- **on** Mondays
- **in** the morning/afternoon/evening
- **at** night
- **at** 8 o'clock

### Examples:
- I go to the gym **every day**.
- She visits her parents **once a week**.
- We have English class **on Tuesdays and Thursdays**.
- He wakes up **at 6 AM**.
