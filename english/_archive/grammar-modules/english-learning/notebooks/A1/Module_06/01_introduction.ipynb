## 1. What Are Nouns? ğŸ“

**Nouns** are words that name people, places, things, or ideas.

### Categories of Nouns:

| Category | Examples |
|----------|----------|
| **People** | teacher, student, friend, doctor, child |
| **Places** | city, school, park, country, home |
| **Things** | book, car, phone, table, apple |
| **Ideas** | love, happiness, time, music, freedom |

### Singular vs Plural:

- **Singular** = ONE thing: **a book**, **one cat**, **the teacher**
- **Plural** = TWO OR MORE things: **books**, **three cats**, **the teachers**

### Why Plurals Matter:

âœ… They tell us the quantity (one or many)  
âœ… They affect verb forms: "The cat **is** sleeping" vs "The cats **are** sleeping"  
âœ… They work with different determiners: "**a** cat" vs "**some** cats"

## 5. Vowel + Y â†’ YS ğŸ”‘

When a noun ends in a **vowel + y**, we simply **add -s** (don't change the y).

### Rule:

**Vowel + y + s = Plural**

Vowels = **a, e, i, o, u**

### Examples:

| Singular | Analysis | Plural |
|----------|----------|--------|
| da**y** | a (vowel) + y | da**ys** |
| bo**y** | o (vowel) + y | bo**ys** |
| ke**y** | e (vowel) + y | ke**ys** |
| to**y** | o (vowel) + y | to**ys** |
| mone**y** | e (vowel) + y | (uncountable) |

### Compare Consonant + Y vs Vowel + Y:

| Type | Singular | Plural |
|------|----------|--------|
| Consonant + y | cit**y** (t+y) | cit**ies** |
| Vowel + y | da**y** (a+y) | da**ys** |
| Consonant + y | bab**y** (b+y) | bab**ies** |
| Vowel + y | bo**y** (o+y) | bo**ys** |
| Consonant + y | lad**y** (d+y) | lad**ies** |
| Vowel + y | ke**y** (e+y) | ke**ys** |

### ğŸ”Š Listen and Compare:

````python
# Listen to vowel + y plurals
examples_ys = [
    "one day, seven days",
    "one boy, three boys",
    "one key, two keys",
    "one toy, many toys",
]

print("Vowel + Y â†’ add -s:")
for example in examples_ys:
    print(f"ğŸ”Š {example}")
    audio.play_audio(example, accent="us")

print("\nConsonant + Y â†’ change to -ies:")
comparison = ["one city, two cities", "one baby, three babies"]
for example in comparison:
    print(f"ğŸ”Š {example}")
    audio.play_audio(example, accent="us")
````

## 8. Introduction to Countable and Uncountable Nouns ğŸ’§

In English, nouns are divided into two main categories:

### Countable Nouns (Can Count Them) ğŸ”¢

**Countable nouns** are things you can count: 1, 2, 3...

- They have **singular** and **plural** forms
- You can use **a/an** with singular countable nouns
- You can use numbers with them

**Examples**:

| Singular | Plural | With Numbers |
|----------|--------|---------------|
| **a** book | books | one book, two books, three books |
| **an** apple | apples | one apple, five apples |
| **a** car | cars | one car, ten cars |
| **a** friend | friends | one friend, many friends |

### Uncountable Nouns (Cannot Count Them) ğŸ’§

**Uncountable nouns** are things you **cannot** count individually.

- They have **NO** plural form
- You **cannot** use **a/an** with them
- You **cannot** use numbers directly with them
- They are always **singular**

**Common Uncountable Nouns**:

| Category | Examples |
|----------|----------|
| **Liquids** | water, milk, coffee, tea, juice, oil |
| **Materials** | wood, paper, glass, plastic, metal |
| **Food** | rice, bread, cheese, meat, sugar, salt |
| **Abstract** | information, advice, knowledge, music |
| **Activities** | homework, work, housework, research |
| **Weather** | rain, snow, weather, sunshine |
| **Other** | money, time, news, furniture, luggage |

### How to Express Quantity with Uncountable Nouns:

Use measure words or containers:

| Uncountable | How to Count |
|-------------|---------------|
| water | **a glass of** water, **a bottle of** water |
| coffee | **a cup of** coffee, **two cups of** coffee |
| rice | **a bowl of** rice, **a kilo of** rice |
| bread | **a loaf of** bread, **a slice of** bread |
| advice | **a piece of** advice, **some** advice |
| furniture | **a piece of** furniture, **some** furniture |

### Compare:

| Countable | Uncountable |
|-----------|-------------|
| âœ… **a** book | âŒ ~~**a** water~~ |
| âœ… **an** apple | âŒ ~~**an** advice~~ |
| âœ… two books | âŒ ~~two waters~~ |
| âœ… many books | âœ… much water |
| âœ… a few books | âœ… a little water |

**Note**: We'll learn more about countable/uncountable nouns in future modules!

## 9. Common Mistakes to Avoid âš ï¸

### Mistake 1: Wrong Plural Ending

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~two boxs~~ | two box**es** |
| ~~three dishs~~ | three dish**es** |
| ~~many buss~~ | many bus**es** |

**Remember**: After -s, -x, -z, -ch, -sh â†’ add **-es**!

### Mistake 2: Consonant + Y Errors

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~two citys~~ | two cit**ies** |
| ~~three babys~~ | three bab**ies** |
| ~~many countrys~~ | many countr**ies** |

**Remember**: Consonant + y â†’ change to **-ies**!

### Mistake 3: Irregular Plural Errors

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~two childs~~ | two **children** |
| ~~three mans~~ | three **men** |
| ~~many peoples~~ | many **people** |
| ~~two mouses~~ | two **mice** |
| ~~two foots~~ | two **feet** |

**Remember**: Irregular plurals must be memorized!

### Mistake 4: Adding Plural to Uncountable Nouns

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~two waters~~ | two **glasses of water** |
| ~~many informations~~ | much **information** |
| ~~three advices~~ | three **pieces of advice** |
| ~~two breads~~ | two **loaves of bread** |
| ~~many furnitures~~ | much **furniture** |

**Remember**: Uncountable nouns have NO plural form!

### Mistake 5: Using 'A/An' with Plural or Uncountable

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~a books~~ | **a book** OR **books** |
| ~~an apples~~ | **an apple** OR **apples** |
| ~~a water~~ | **water** OR **a glass of water** |
| ~~an information~~ | **information** OR **some information** |

**Remember**: Use a/an ONLY with singular countable nouns!

### Mistake 6: Same-Form Plurals

| âŒ Wrong | âœ… Correct |
|---------|----------|
| ~~two sheeps~~ | two **sheep** |
| ~~three deers~~ | three **deer** |
| ~~many fishs~~ | many **fish** |

**Remember**: Some nouns have the same singular and plural form!
