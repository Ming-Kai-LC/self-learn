## Section 1: What Are Adjectives? ğŸ¨

### Definition

**Adjectives** are words that **describe** or **modify** nouns. They tell us more information about people, places, things, or ideas.

### Examples:

| Without Adjective | With Adjective | What it describes |
|---|---|---|
| I have a car | I have a **red** car | Color |
| She lives in a house | She lives in a **big** house | Size |
| He is a teacher | He is a **good** teacher | Quality |
| It's a day | It's a **beautiful** day | Quality |
| They are students | They are **smart** students | Quality |

### Key Point:

Adjectives give us **MORE INFORMATION** about nouns. They answer questions like:
- **What kind?** (a **fast** car)
- **What color?** (a **blue** shirt)
- **How many?** (**three** books)
- **Which one?** (**this** phone)

## Section 2: Adjectives BEFORE Nouns ğŸ“

### Pattern: Article + Adjective + Noun

This is the **most common** way to use adjectives in English.

### Structure:

```
[a/an/the] + [ADJECTIVE] + [NOUN]
```

### Examples:

| Article | Adjective | Noun | Full Phrase |
|---|---|---|---|
| a | **big** | house | a big house |
| an | **old** | car | an old car |
| the | **beautiful** | garden | the beautiful garden |
| a | **new** | phone | a new phone |
| the | **happy** | child | the happy child |

### Remember:

- Use **a** before consonant sounds: a **big** house
- Use **an** before vowel sounds: an **old** car
- The adjective goes **BETWEEN** the article and the noun

### âŒ Common Mistakes:

| âŒ Wrong | âœ… Correct |
|---|---|
| a house big | a big house |
| an car old | an old car |
| big a house | a big house |

### More Examples:

- I want a **cold** drink.
- She has a **beautiful** smile.
- They live in a **small** apartment.
- He drives a **fast** car.
- We saw a **funny** movie.

## Section 3: Adjectives AFTER 'Be' ğŸ¯

### Pattern: Subject + be + Adjective

Adjectives can also come **after the verb 'be'** (am, is, are, was, were).

### Structure:

```
[SUBJECT] + [BE] + [ADJECTIVE]
```

### Examples:

| Subject | Be | Adjective | Full Sentence |
|---|---|---|---|
| I | am | **happy** | I am happy. |
| She | is | **beautiful** | She is beautiful. |
| They | are | **tired** | They are tired. |
| The car | is | **new** | The car is new. |
| We | are | **ready** | We are ready. |

### Forms of 'Be':

**Present:**
- I **am** happy
- You **are** smart
- He/She/It **is** tall
- We/You/They **are** busy

**Past:**
- I/He/She/It **was** happy
- You/We/They **were** happy

### More Examples:

- The weather **is cold**.
- My sister **is tall**.
- The students **are quiet**.
- This coffee **is hot**.
- Those books **are interesting**.

### Compare Both Patterns:

| Before Noun | After 'Be' |
|---|---|
| I have a **big** house. | My house **is big**. |
| She is a **beautiful** woman. | She **is beautiful**. |
| He drives a **fast** car. | His car **is fast**. |
| We saw a **good** movie. | The movie **was good**. |

---

````python
# Listen to adjectives after 'be'
examples_after_be = [
    "I am happy.",
    "She is beautiful.",
    "They are tired.",
    "The car is new.",
    "We are ready.",
    "The weather is cold.",
    "My sister is tall.",
]

print("ğŸ§ Adjectives AFTER 'Be':\n")
for example in examples_after_be:
    print(f"ğŸ”Š {example}")
    audio.play_audio(example, accent="us")
    print()
````

## Section 7: Basic Adjective Order ğŸ“‹

When you use **TWO OR MORE** adjectives before a noun, they follow a specific order.

### The Order (from first to last):

1. **Opinion** (beautiful, nice, ugly)
2. **Size** (big, small, large)
3. **Age** (old, new, young)
4. **Color** (red, blue, green)
5. **Origin** (Chinese, Italian, American)
6. **Material** (wooden, metal, plastic)
7. **NOUN**

### Memory Trick: **OSACO M**

**O**pinion â†’ **S**ize â†’ **A**ge â†’ **C**olor â†’ **O**rigin â†’ **M**aterial

### Examples:

| Opinion | Size | Age | Color | Origin | Material | Noun |
|---|---|---|---|---|---|---|
| beautiful | big | old | red | Italian | wooden | table |
| nice | small | new | blue | - | - | car |
| - | large | - | green | - | plastic | bag |
| ugly | - | old | - | - | metal | chair |

### Full Examples:

- a **beautiful big old red Italian wooden** table
- a **nice small new blue** car
- a **large green plastic** bag
- an **ugly old metal** chair

### Common Combinations (2 adjectives):

| Pattern | Example |
|---|---|
| Opinion + Size | a **nice big** house |
| Size + Color | a **big red** car |
| Age + Color | an **old blue** shirt |
| Size + Age | a **small new** phone |
| Opinion + Color | a **beautiful white** dress |

### Real-Life Examples:

- I bought a **beautiful new red** dress.
- She lives in a **nice big old** house.
- He drives a **small blue Japanese** car.
- We have a **lovely old wooden** table.

### Important Note:

At A1 level, you usually use **1-2 adjectives** at most. Don't worry about using many adjectives together!

## Section 8: Adjectives DON'T Change! ğŸ¯

### VERY IMPORTANT RULE:

In English, **adjectives NEVER change form**. They are the same for:
- Singular and plural nouns
- Masculine and feminine
- Different subjects

### Examples:

| Singular | Plural | Rule |
|---|---|---|
| a **big** house | **big** houses | Same adjective! |
| a **red** car | **red** cars | No plural form! |
| a **beautiful** woman | **beautiful** women | No change! |

### âŒ Common Mistakes:

| âŒ Wrong | âœ… Correct | Explanation |
|---|---|---|
| reds cars | red cars | Adjectives don't have plural |
| bigs houses | big houses | No -s on adjectives |
| beautifuls flowers | beautiful flowers | Never add -s |

### Compare with Other Languages:

Some languages change adjectives (like Spanish or French), but **English does NOT**:

- **Spanish:** un carro rojo (masculine) / una casa roja (feminine)
- **English:** a red car / a red house (SAME!)

This makes English adjectives **easier**!
