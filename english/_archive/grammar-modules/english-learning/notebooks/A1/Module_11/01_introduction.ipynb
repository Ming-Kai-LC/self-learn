## 1. What are Question Words?

Question words are special words we use at the beginning of questions to ask for specific information.

### The Six Essential Question Words:

| Question Word | Asks About | Example |
|---------------|------------|----------|
| **WHAT** | Things, actions, information | What is your name? |
| **WHERE** | Places, locations | Where do you live? |
| **WHO** | People | Who is your teacher? |
| **WHEN** | Time | When is your birthday? |
| **WHY** | Reasons | Why are you happy? |
| **HOW** | Method, condition | How do you go to work? |

### üéØ Key Point:
Question words come at the **beginning** of the question!

---

## 2. WHAT - Asking About Things and Information üì¶

**WHAT** asks about things, objects, actions, or information.

### Uses of WHAT:

**1. What + noun** (asking for specific information):
- What **time** is it?
- What **color** is your car?
- What **day** is today?

**2. What + be** (asking for identification):
- What is this?
- What is your name?
- What are those?

**3. What + do** (asking about actions):
- What do you do? (job)
- What does she do on weekends?
- What did you do yesterday?

### Common WHAT Questions:
- What is this?
- What is your favorite food?
- What time is it?
- What color do you like?
- What day is today?
- What do you want?

## 3. WHERE - Asking About Places and Locations üìç

**WHERE** asks about places, locations, and positions.

### Structure:
```
Where + be + subject?
Where + do/does + subject + verb?
```

### Examples:
- Where **is** the bathroom?
- Where **are** you from?
- Where **do** you live?
- Where **does** she work?

### Examples with Audio üîä

````python
# WHERE examples
audio.play_audio("Where are you from?", accent="us")
````

````python
audio.play_audio("Where do you live?", accent="us")
````

````python
audio.play_audio("Where is the post office?", accent="us")
````

### Answering WHERE Questions:
- Where are you? ‚Üí I'm **at home** / **in the office** / **on the bus**
- Where do you live? ‚Üí I live **in London** / **on Main Street**
- Where is the bank? ‚Üí It's **next to the post office** / **on the corner**

## 4. WHO - Asking About People üë§

**WHO** asks about people (identity).

### Structure:
```
Who + be + ...?
Who + verb + ...? (Who as subject)
```

### Examples:
- Who **is** that man?
- Who **is** your teacher?
- Who **are** those people?
- Who **lives** here? (Who as subject)
- Who **wants** ice cream? (Who as subject)

### Examples with Audio üîä

````python
audio.play_audio("Who lives in that house?", accent="us")
````

### Answering WHO Questions:
- Who is that? ‚Üí That's **my brother** / **Maria** / **my teacher**
- Who lives here? ‚Üí **John** lives here / **My family** lives here
- Who wants coffee? ‚Üí **I do** / **We all do**

## 5. WHEN - Asking About Time ‚è∞

**WHEN** asks about time (day, date, time of day, etc.).

### Structure:
```
When + be + event?
When + do/does + subject + verb?
```

### Examples:
- When **is** your birthday?
- When **is** the meeting?
- When **do** you wake up?
- When **does** the class start?

### Answering WHEN Questions:
- When is your birthday? ‚Üí It's **on May 15th** / **in July**
- When do you work? ‚Üí I work **from Monday to Friday** / **in the morning**
- When does the store close? ‚Üí It closes **at 9 PM** / **at night**

## 6. WHY - Asking About Reasons ü§î

**WHY** asks for reasons and explanations.

### Structure:
```
Why + be + subject + adjective?
Why + do/does + subject + verb?
```

### Examples:
- Why **are** you late?
- Why **is** she happy?
- Why **do** you study English?
- Why **does** he work here?

### Answer with BECAUSE:
When someone asks WHY, answer with **BECAUSE**:
- Why are you tired? ‚Üí **Because** I didn't sleep well.
- Why do you study English? ‚Üí **Because** I need it for work.

## 7. HOW - Asking About Method and Condition üõ§Ô∏è

**HOW** asks about the way/method of doing something, or about someone's condition.

### Uses of HOW:

**1. How + be** (asking about condition):
- How **are** you? (condition/health)
- How **is** the weather?
- How **is** your mother?

**2. How + do** (asking about method):
- How **do** you go to work? (by car, by bus, etc.)
- How **do** you cook this?
- How **does** it work?

**3. How + adjective/adverb** (asking about degree):
- How **old** are you?
- How **tall** is he?
- How **much** does it cost?
- How **many** students are there?
- How **often** do you exercise?

- How do you go to work? ‚Üí I go **by car** / **by bus** / **on foot**
- How old are you? ‚Üí I'm **25 years old**
- How much does it cost? ‚Üí It costs **$10**

## 8. Question Word Summary Table üìä

| Question Word | Asks About | Answer Type | Example Question | Example Answer |
|---------------|------------|-------------|------------------|----------------|
| **WHAT** | Things, actions | Specific information | What is your name? | My name is John. |
| **WHERE** | Places, locations | Place | Where do you live? | I live in London. |

## 9. Word Order in Wh-Questions

### With BE:
```
Wh-word + BE + Subject + ...?
```
Examples:
- Where **are** you?
- When **is** the meeting?
- Who **is** that man?

### With DO/DOES:
```
Wh-word + DO/DOES + Subject + Base Verb + ...?
```
Examples:
- What **do** you eat for breakfast?
- Where **does** she work?
- When **do** they arrive?

### Exception - WHO as Subject:
When WHO is the subject, we don't use DO/DOES:
- Who **lives** here? (NOT: Who does live here?)
- Who **wants** coffee?
- Who **knows** the answer?

## 10. Common Question Combinations

### WHAT + Noun:

- What **type** of food do you prefer?

### HOW + Adjective/Adverb:
- How **old** are you?
- How **tall** is the building?
- How **far** is the station?
- How **long** does it take?
- How **often** do you exercise?
- How **much** money do you need?
- How **many** books do you have?

## 11. Common Mistakes to Avoid ‚ö†Ô∏è

### Mistake 1: Wrong word order
‚ùå Where you are? ‚Üí ‚úÖ Where **are you**?
‚ùå What you do? ‚Üí ‚úÖ What **do you do**?

### Mistake 2: Forgetting DO/DOES
‚ùå Where you work? ‚Üí ‚úÖ Where **do** you work?
‚ùå When she starts? ‚Üí ‚úÖ When **does** she start?

### Mistake 3: Using DO/DOES with WHO as subject
‚ùå Who does live here? ‚Üí ‚úÖ Who **lives** here?
‚ùå Who does want coffee? ‚Üí ‚úÖ Who **wants** coffee?

### Mistake 4: Confusing question words
‚ùå What are you from? ‚Üí ‚úÖ **Where** are you from?
‚ùå When is your name? ‚Üí ‚úÖ **What** is your name?

### Mistake 5: Not using BECAUSE with WHY
‚ùå Why are you late? "I missed the bus." ‚Üí ‚úÖ Why are you late? "**Because** I missed the bus."
