## 1. What is the Verb "To Be"?

The verb "**to be**" describes **existence**, **identity**, or **state**. It answers questions like:
- Who are you?
- What is your name?
- Where are you from?
- How are you?

### Present Tense Forms

The verb "to be" has **three forms** in the present tense:

| Pronoun | Verb Form | Contraction |
|---------|-----------|-------------|
| I       | **am**    | I'm         |
| You     | **are**   | You're      |
| He      | **is**    | He's        |
| She     | **is**    | She's       |
| It      | **is**    | It's        |
| We      | **are**   | We're       |
| They    | **are**   | They're     |

### üéØ Key Rule
- Use **AM** with **I**
- Use **IS** with **he, she, it** (singular)
- Use **ARE** with **you, we, they** (plural)

## 3. Questions with "To Be"

To make a question, **reverse the subject and verb**:

### Structure:
```
Verb "to be" + Subject + Complement + ?
```

### Statement ‚Üí Question Transformation

| Statement | Question |
|-----------|----------|
| You **are** happy. | **Are** you happy? |
| She **is** a doctor. | **Is** she a doctor? |
| They **are** ready. | **Are** they ready? |

### Examples with Audio üîä

````python
# Question examples
audio.play_audio("Are you ready?", accent="us")
audio.play_audio("Is she your friend?", accent="us")
audio.play_audio("Are they from Spain?", accent="us")
````

## 4. Negative Sentences

To make a negative sentence, add **NOT** after the verb:

### Structure:
```
Subject + Verb "to be" + NOT + Complement
```

### Full Form vs Contraction

| Full Form | Contraction |
|-----------|-------------|
| I **am not** | I'm not |
| You **are not** | You **aren't** / You're not |
| He **is not** | He **isn't** / He's not |
| She **is not** | She **isn't** / She's not |
| It **is not** | It **isn't** / It's not |
| We **are not** | We **aren't** / We're not |
| They **are not** | They **aren't** / They're not |

### Examples with Audio üîä

````python
# Negative examples
audio.play_audio("I am not tired.", accent="us")
audio.play_audio("She isn't from Germany.", accent="us")
audio.play_audio("They aren't at home.", accent="us")
````

## 5. Short Answers

When someone asks you a yes/no question, you can give a **short answer**:

### Structure:
```
Yes, + subject pronoun + verb
No, + subject pronoun + verb + not
```

### Examples:

| Question | Positive Answer | Negative Answer |
|----------|----------------|------------------|
| Are you happy? | Yes, I **am**. | No, I'm **not**. |
| Is she a teacher? | Yes, she **is**. | No, she **isn't**. |
| Are they ready? | Yes, they **are**. | No, they **aren't**. |

## 6. Common Mistakes to Avoid ‚ö†Ô∏è

### ‚ùå Wrong ‚Üí ‚úÖ Right

1. ‚ùå I **is** happy ‚Üí ‚úÖ I **am** happy
2. ‚ùå She **are** beautiful ‚Üí ‚úÖ She **is** beautiful
3. ‚ùå They **is** students ‚Üí ‚úÖ They **are** students
4. ‚ùå You **is** late ‚Üí ‚úÖ You **are** late
5. ‚ùå He **am** tired ‚Üí ‚úÖ He **is** tired

### Remember:
- **I + AM** (always together!)
- **He/She/It + IS** (singular)
- **You/We/They + ARE** (plural)
