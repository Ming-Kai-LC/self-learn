## 1. What is CAN?

**CAN** is a **modal verb**. Modal verbs are special helping verbs that change the meaning of the main verb.

### Key Features of CAN:

1. **Same form for all subjects** (no -s for he/she/it):
   - I can ‚úÖ
   - You can ‚úÖ
   - He/She/It can ‚úÖ (NOT "cans")
   - We can ‚úÖ
   - They can ‚úÖ

2. **Followed by base verb** (without "to"):
   - I can swim ‚úÖ (NOT "I can to swim")
   - She can dance ‚úÖ (NOT "She can to dance")

3. **No DO/DOES in questions**:
   - Can you swim? ‚úÖ (NOT "Do you can swim?")
   - Can she dance? ‚úÖ (NOT "Does she can dance?")

### üéØ Key Point:
CAN is very simple - it **never changes form**!

## 3. CAN'T for Inability (What You Cannot Do) üö´

We use **CAN'T** (cannot) to talk about things we are not able to do.

### Structure:
```
Subject + CAN'T + base verb
Subject + CANNOT + base verb (more formal)
```

### Examples:

**Physical abilities:**
- I **can't** fly. (Nobody can!)
- She **can't** swim. (She doesn't know how)
- He **can't** run fast. (He's not able to)

**Skills and knowledge:**
- I **can't** speak Chinese.
- He **can't** drive a car.
- We **can't** play chess.

**Temporary inability:**
- I **can't** come to the party. (I'm busy)
- She **can't** see without glasses.
- They **can't** hear you. (Too noisy)

### Contractions:
| Full Form | Contraction | Usage |
|-----------|-------------|-------|
| cannot | can't | Common in speaking and informal writing |
| cannot | cannot | Formal writing |

**Note:** "Cannot" is ONE word, but the contraction "can't" is more common in everyday English.

### Examples with Audio üîä

````python
# CAN'T for inability examples
audio.play_audio("I can't swim.", accent="us")
````

````python
audio.play_audio("He can't speak Spanish.", accent="us")
````

````python
audio.play_audio("They can't come to the meeting.", accent="us")
````

## 4. Questions with CAN ‚ùì

To make questions with CAN, we put CAN before the subject.

### Structure:
```
CAN + subject + base verb + ...?
```

### Yes/No Questions:
- **Can** you swim?
- **Can** she speak French?
- **Can** they come tomorrow?
- **Can** he play the guitar?

### Wh-Questions:
We can combine question words with CAN:
- **What** can you do?
- **Where** can I buy tickets?
- **When** can you start?
- **Who** can help me?
- **How** can I get there?

### Examples with Audio üîä

````python
audio.play_audio("Where can I find a taxi?", accent="us")
````

## 5. Short Answers with CAN üëçüëé

When answering Yes/No questions with CAN, we use short answers.

### Structure:
```
Yes, + subject + can.
No, + subject + can't.
```

### Examples:

| Question | Positive Answer | Negative Answer |
|----------|----------------|------------------|
| Can you swim? | Yes, I can. | No, I can't. |
| Can she drive? | Yes, she can. | No, she can't. |
| Can they come? | Yes, they can. | No, they can't. |
| Can he speak French? | Yes, he can. | No, he can't. |

### Important Rules:
1. **Don't repeat the main verb** in short answers:
   - Can you swim? ‚Üí Yes, I can. ‚úÖ (NOT "Yes, I can swim." in short answers)
   
2. **Don't use contractions in positive short answers:**
   - Yes, I can. ‚úÖ (NOT "Yes, I can't.")
   - No, I can't. ‚úÖ

### Conversation Example:

## 6. CAN for Permission (Asking if Something is Allowed) üö™

We use **CAN** to ask for permission (to ask if something is allowed).

### Structure:
```
Can I/we + base verb + ...?
```

### Asking for Permission:
- **Can I** use your phone?
- **Can I** sit here?
- **Can I** ask a question?
- **Can I** open the window?
- **Can we** leave early?
- **Can we** take a break?

### Giving Permission:
- Yes, you can. ‚úÖ
- Sure! / Of course! ‚úÖ
- No problem! ‚úÖ

### Refusing Permission:
- No, you can't. (Direct)
- I'm sorry, but no. (Polite)
- I'm afraid not. (Polite)

### Examples with Audio üîä

## 7. CAN for Requests (Asking Someone to Do Something) üôè

We use **CAN** to make requests - to ask someone to do something for us.

### Structure:
```
Can you + base verb + ...?
```

### Making Requests:
- **Can you** help me?
- **Can you** open the door?
- **Can you** pass the salt?
- **Can you** speak more slowly?
- **Can you** tell me the time?
- **Can you** wait a moment?

### Responding to Requests:

**Agreeing:**
- Yes, I can. / Sure! / Of course! / No problem!

**Refusing politely:**
- I'm sorry, I can't. (+ reason)
- I'm afraid I can't.

### Polite Tip:
Add **"please"** to make requests more polite:
- Can you help me, **please**?
- Can you **please** open the window?

### Examples with Audio üîä

````python
audio.create_conversation_audio(
    [
        ("Person A", "Can you lend me ten dollars?"),
        ("Person B", "I'm sorry, I can't. I don't have any money with me."),
    ],
    accent="us",
)
````

## 8. CAN for Offers (Offering to Do Something) üéÅ

We use **CAN** to offer help or to volunteer to do something.

### Structure:
```
I/We can + base verb + ...
Can I/we + base verb + ...? (as a question)
```

### Making Offers:
- **I can** help you.
- **I can** carry that for you.
- **We can** give you a ride.
- **Can I** help you?
- **Can I** carry your bags?
- **Can we** give you a hand?

### Accepting Offers:
- Yes, please. / Thank you! / That would be great!

### Declining Offers:
- No, thank you. / It's okay, I can manage. / That's kind, but I'm fine.

### Examples with Audio üîä

````python
# Offer examples
audio.play_audio("I can help you with that.", accent="us")
````

## 9. Summary Table of CAN Uses üìä

| Use | Structure | Example | Translation Idea |
|-----|-----------|---------|------------------|
| **Ability** | S + can + verb | I can swim. | I am able to swim |
| **Inability** | S + can't + verb | I can't drive. | I am not able to drive |
| **Questions (ability)** | Can + S + verb? | Can you dance? | Are you able to dance? |
| **Permission (asking)** | Can I/we + verb? | Can I sit here? | Am I allowed to sit here? |
| **Permission (giving)** | You can + verb | You can sit here. | You are allowed to sit |
| **Requests** | Can you + verb? | Can you help me? | Please help me |
| **Offers** | Can I + verb? / I can + verb | Can I help you? | Do you want me to help? |

## 10. Word Order Rules üìè

### Positive Statements:
```
Subject + CAN + base verb + (object/complement)
```
Examples:
- I can speak English.
- She can play the piano.
- They can swim very fast.

### Negative Statements:
```
Subject + CAN'T + base verb + (object/complement)
```
Examples:
- I can't speak Chinese.
- He can't play tennis.
- We can't come tomorrow.

### Questions:
```
CAN + subject + base verb + (object/complement)?
```
Examples:
- Can you speak English?
- Can she play the piano?
- Can they swim?

### With Question Words:
```
Question word + CAN + subject + base verb?
```
Examples:
- What can you do?
- Where can I buy tickets?
- When can you start?

## 11. Common Mistakes to Avoid ‚ö†Ô∏è

### Mistake 1: Adding -s to CAN
‚ùå He **cans** swim. ‚Üí ‚úÖ He **can** swim.

‚ùå She **cans** speak English. ‚Üí ‚úÖ She **can** speak English.

**Remember:** CAN never changes form!

### Mistake 2: Using "to" after CAN
‚ùå I can **to swim**. ‚Üí ‚úÖ I can swim.

‚ùå She can **to drive**. ‚Üí ‚úÖ She can drive.

**Remember:** Use the base verb (without "to") after CAN!

### Mistake 3: Using DO/DOES in questions
‚ùå **Do** you can swim? ‚Üí ‚úÖ **Can** you swim?

‚ùå **Does** he can drive? ‚Üí ‚úÖ **Can** he drive?

**Remember:** Don't use DO/DOES with CAN!

### Mistake 4: Using DO/DON'T in negatives
‚ùå I **don't can** swim. ‚Üí ‚úÖ I **can't** swim.

‚ùå She **doesn't can** drive. ‚Üí ‚úÖ She **can't** drive.

**Remember:** Use CAN'T, not don't/doesn't!

### Mistake 5: Wrong verb form after CAN
‚ùå I can **swimming**. ‚Üí ‚úÖ I can **swim**.

‚ùå He can **plays** guitar. ‚Üí ‚úÖ He can **play** guitar.

‚ùå She can **drove**. ‚Üí ‚úÖ She can **drive**.

**Remember:** Always use the base form of the verb!

### Mistake 6: Confusing CAN and CAN'T pronunciation
**CAN** = /k√¶n/ (weak) or /k√¶n/ (strong)

**CAN'T** = /k…ëÀênt/ (US) or /k…ëÀênt/ (UK)

Listen carefully to the difference!
