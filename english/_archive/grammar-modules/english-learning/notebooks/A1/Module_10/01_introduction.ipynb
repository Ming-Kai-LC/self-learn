## 1. What are Prepositions?

**Prepositions** are small words that show the relationship between things.

They tell us:
- üîπ **WHERE** something is (place)
- üîπ **WHEN** something happens (time)
- üîπ The **relationship** between objects

### Common Prepositions:

| Type | Prepositions |
|------|-------------|
| **Place** | in, on, at, under, over, behind, in front of, next to, between, near |
| **Time** | in, on, at, before, after, during, for, since |
| **Direction** | to, from, into, out of, toward |

### Structure:

```
preposition + noun/pronoun
```

### Examples:

| Preposition | Example |
|-------------|----------|
| in | The book is **in** the bag. |
| on | The cup is **on** the table. |
| at | I'm **at** home. |
| under | The cat is **under** the chair. |
| before | I eat breakfast **before** work. |

### Important:
- Prepositions are **always** followed by a noun or pronoun
- The same preposition can have different meanings in different contexts
- Some prepositions are fixed in certain expressions

## 9. Common Mistakes to Avoid

### ‚ùå Mistake 1: Confusing in/on/at for place

| Wrong | Correct | Rule |
|-------|---------|------|
| ‚ùå I live **on** London. | ‚úÖ I live **in** London. | Use "in" for cities |
| ‚ùå The book is **in** the table. | ‚úÖ The book is **on** the table. | Use "on" for surfaces |
| ‚ùå I'm **in** home. | ‚úÖ I'm **at** home. | Use "at" for home |
| ‚ùå He works **in** Main Street. | ‚úÖ He works **on** Main Street. | Use "on" for streets |

### ‚ùå Mistake 2: Confusing in/on/at for time

| Wrong | Correct | Rule |
|-------|---------|------|
| ‚ùå The meeting is **in** Monday. | ‚úÖ The meeting is **on** Monday. | Use "on" for days |
| ‚ùå I wake up **on** 7 AM. | ‚úÖ I wake up **at** 7 AM. | Use "at" for times |
| ‚ùå My birthday is **at** June. | ‚úÖ My birthday is **in** June. | Use "in" for months |
| ‚ùå It's cold **on** winter. | ‚úÖ It's cold **in** winter. | Use "in" for seasons |

### ‚ùå Mistake 3: Using wrong preposition with transportation

| Wrong | Correct | Rule |
|-------|---------|------|
| ‚ùå I'm **in** the bus. | ‚úÖ I'm **on** the bus. | Use "on" for public transport |
| ‚ùå He's **on** a taxi. | ‚úÖ He's **in** a taxi. | Use "in" for cars/taxis |
| ‚ùå We go **with** train. | ‚úÖ We go **by** train. | Use "by" for mode of transport |

### ‚ùå Mistake 4: Forgetting prepositions

| Wrong | Correct | Rule |
|-------|---------|------|
| ‚ùå I arrive Monday. | ‚úÖ I arrive **on** Monday. | Need preposition with days |
| ‚ùå The book is the table. | ‚úÖ The book is **on** the table. | Need preposition for position |
| ‚ùå I go home 6 PM. | ‚úÖ I go home **at** 6 PM. | Need preposition with times |

### ‚ùå Mistake 5: Using "in" instead of "at" for night

| Wrong | Correct | Explanation |
|-------|---------|-------------|
| ‚ùå **In** the night | ‚úÖ **At** night | Fixed expression |
| ‚úÖ **In** the morning | ‚úÖ **In** the morning | Correct |
| ‚úÖ **In** the evening | ‚úÖ **In** the evening | Correct |

### ‚ùå Mistake 6: Wrong preposition with fixed expressions

| Wrong | Correct |
|-------|----------|
| ‚ùå in home | ‚úÖ **at** home |
| ‚ùå in work | ‚úÖ **at** work |
| ‚ùå in the weekend | ‚úÖ **on/at** the weekend |
| ‚ùå in time (punctual) | ‚úÖ **on** time |

### Memory Tips:

**Place - Size Rule:**
```
IN  = INSIDE (enclosed)     ‚Üí in the room, in London
ON  = CONTACT (surface)     ‚Üí on the table, on the wall
AT  = POINT (specific spot) ‚Üí at the door, at home
```

**Time - Specificity Rule:**
```
IN  = GENERAL (longer)  ‚Üí in June, in 2024, in the morning
ON  = SPECIFIC DAY      ‚Üí on Monday, on March 15th
AT  = EXACT TIME        ‚Üí at 3 PM, at noon, at midnight
```
