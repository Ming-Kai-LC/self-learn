## Activity 5: Its vs It's in Daily Life üêï

### Task:
Practice the **CRITICAL** difference between **its** (possessive) and **it's** (it is/it has).

### Remember:
- **its** = possessive (The dog wagged **its** tail)
- **it's** = it is OR it has (**It's** raining = It is raining)

### Test Method:
Can you replace it with "it is" or "it has"? ‚Üí Use **it's**
If not ‚Üí Use **its**
