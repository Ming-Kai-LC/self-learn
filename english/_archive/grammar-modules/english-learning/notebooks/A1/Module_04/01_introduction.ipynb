## 2. Present Simple Negatives

To make negative sentences, we use **do not (don't)** or **does not (doesn't)** + base verb.

### Structure:
```
Subject + don't/doesn't + base verb
```

### Complete Table:

| Person | Negative | Example |
|--------|----------|----------|
| I | **don't** work | I don't work on weekends |
| You | **don't** work | You don't work here |
| He/She/It | **doesn't** work | She doesn't work Fridays |
| We | **don't** work | We don't work late |
| They | **don't** work | They don't work together |

### üéØ Key Rules:
1. **I/you/we/they** ‚Üí use **don't**
2. **He/she/it** ‚Üí use **doesn't**
3. ‚ö†Ô∏è **ALWAYS use base verb** (no -s!) after don't/doesn't

---

## 3. Important: Base Verb in Negatives! ‚≠ê

This is a **critical rule** that students often forget:

### Affirmative vs Negative:

‚úÖ **Affirmative:** She work**s** ‚Üí The -s goes on the verb

‚úÖ **Negative:** She **doesn't** work ‚Üí The -s is in "doesn't", so we use base verb!

### Common Mistakes:

‚ùå She doesn't works ‚Üí ‚úÖ She doesn't work

‚ùå He doesn't likes ‚Üí ‚úÖ He doesn't like

‚ùå It doesn't rains ‚Üí ‚úÖ It doesn't rain

## 4. Contractions: Don't vs Do Not

In spoken and informal written English, we usually use contractions:

| Full Form | Contraction | Usage |
|-----------|-------------|-------|
| do not | **don't** | Common in speech and writing |
| does not | **doesn't** | Common in speech and writing |
| I do not | I **don't** | More formal (written) |
| She does not | She **doesn't** | More formal (written) |

### When to use which:
- **Conversation, emails, casual writing:** Use contractions (don't, doesn't)
- **Formal writing, emphasis:** Use full forms (do not, does not)

## 5. Yes/No Questions with Do/Does

To ask yes/no questions, we use **Do** or **Does** at the beginning.

### Structure:
```
Do/Does + subject + base verb + ...?
```

### Complete Table:

| Person | Question | Example |
|--------|----------|----------|
| I | **Do** I...? | Do I work too much? |
| You | **Do** you...? | Do you work here? |
| He/She/It | **Does** he/she/it...? | Does she work here? |
| We | **Do** we...? | Do we work tomorrow? |
| They | **Do** they...? | Do they work together? |

### üéØ Key Rules:
1. **I/you/we/they** ‚Üí use **Do**
2. **He/she/it** ‚Üí use **Does**
3. ‚ö†Ô∏è **ALWAYS use base verb** (no -s!) after Do/Does

## 6. Short Answers

We don't just say "Yes" or "No" to questions. We use short answers with do/does.

### Structure:
```
Yes, + subject + do/does
No, + subject + don't/doesn't
```

### Examples:

| Question | Positive Answer | Negative Answer |
|----------|----------------|------------------|
| Do you like coffee? | Yes, I **do**. | No, I **don't**. |
| Does she work here? | Yes, she **does**. | No, she **doesn't**. |
| Do they speak Spanish? | Yes, they **do**. | No, they **don't**. |

### ‚ö†Ô∏è Common Mistake:
‚ùå "Do you like coffee?" ‚Üí "Yes, I like." (WRONG)

‚úÖ "Do you like coffee?" ‚Üí "Yes, I do." (CORRECT)

## 7. Wh-Questions (Information Questions)

Wh-questions ask for specific information (not just yes/no).

### Structure:
```
Wh-word + do/does + subject + base verb + ...?
```

### Wh-Words:
- **What** = asking about things
- **Where** = asking about places
- **When** = asking about time
- **Who** = asking about people
- **Why** = asking about reasons
- **How** = asking about manner/way

### Examples:

| Wh-word | Question | Example Answer |
|---------|----------|----------------|
| What | What **do** you **do**? | I work as a teacher. |
| Where | Where **does** she **live**? | She lives in London. |
| When | When **do** they **start** work? | They start at 9 AM. |
| Who | Who **do** you **live** with? | I live with my family. |
| Why | Why **does** he **study** English? | Because he wants a better job. |
| How | How **do** you **get** to work? | I take the bus. |

## 8. Special Case: Questions About the Subject

When "who" or "what" is the **subject** of the question, we DON'T use do/does!

### Compare:

**Normal question (about object):**
- Who **do** you like? ‚Üí I like Maria.

**Subject question (who is the subject):**
- Who like**s** you? ‚Üí Maria likes me.

### More Examples:

| About Object (use do/does) | About Subject (no do/does) |
|----------------------------|---------------------------|
| What **do** you eat? | What happen**s** next? |
| Who **do** they know? | Who know**s** the answer? |
| What **does** she want? | What make**s** you happy? |

**Note:** We'll practice this more later. For now, focus on regular questions!

## 9. Common Mistakes to Avoid ‚ö†Ô∏è

### Mistake 1: Adding -s to the verb
‚ùå She doesn't like**s** coffee ‚Üí ‚úÖ She doesn't like coffee

‚ùå Does he work**s** here? ‚Üí ‚úÖ Does he work here?

### Mistake 2: Forgetting "do/does" in questions
‚ùå You speak English? ‚Üí ‚úÖ Do you speak English?

‚ùå Where she lives? ‚Üí ‚úÖ Where does she live?

### Mistake 3: Wrong word order
‚ùå What you do? ‚Üí ‚úÖ What do you do?

‚ùå Where lives she? ‚Üí ‚úÖ Where does she live?

### Mistake 4: Using "to be" instead of "do/does"
‚ùå Are you speak English? ‚Üí ‚úÖ Do you speak English?

‚ùå Is she works here? ‚Üí ‚úÖ Does she work here?

### Mistake 5: Short answers without do/does
‚ùå "Do you like it?" ‚Üí "Yes, I like." ‚Üí ‚úÖ "Yes, I do."
