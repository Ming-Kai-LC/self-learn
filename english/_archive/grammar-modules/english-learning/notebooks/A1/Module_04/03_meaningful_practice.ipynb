## üìå Key Points Review:

‚úÖ **Negatives:** Subject + don't/doesn't + base verb

‚úÖ **Yes/No Questions:** Do/Does + subject + base verb?

‚úÖ **Short Answers:** Yes, I do / No, I don't (etc.)

‚úÖ **Wh-Questions:** Wh-word + do/does + subject + base verb?

‚úÖ **Critical Rule:** NO -s on the verb after do/does/don't/doesn't!
