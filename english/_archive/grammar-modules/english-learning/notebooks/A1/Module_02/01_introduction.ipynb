## 1. What are Personal Pronouns?

Personal pronouns replace **nouns** (names of people, animals, things) to make sentences smoother and avoid repetition.

### Two Types of Personal Pronouns:

1. **Subject Pronouns** - do the action (before the verb)
2. **Object Pronouns** - receive the action (after the verb)

### Complete Table

| Person | Subject Pronoun | Object Pronoun | Example (Subject) | Example (Object) |
|--------|-----------------|----------------|-------------------|------------------|
| 1st singular | **I** | **me** | **I** love pizza | Give it to **me** |
| 2nd singular | **you** | **you** | **You** are nice | I see **you** |
| 3rd singular (male) | **he** | **him** | **He** is tall | Call **him** |
| 3rd singular (female) | **she** | **her** | **She** is smart | Help **her** |
| 3rd singular (thing) | **it** | **it** | **It** is cold | Take **it** |
| 1st plural | **we** | **us** | **We** are happy | Join **us** |
| 2nd plural | **you** | **you** | **You** are ready | I'll tell **you** |
| 3rd plural | **they** | **them** | **They** are here | Meet **them** |

## 3. Object Pronouns (me, you, him, her, it, us, them)

Object pronouns **receive the action**. They come **after the verb** or **after a preposition**.

### Structure:
```
Subject + Verb + Object Pronoun
Subject + Verb + Preposition + Object Pronoun
```

## 4. Subject vs Object: The Key Difference

### üéØ Simple Rule:
- **BEFORE the verb** = Subject pronoun
- **AFTER the verb** = Object pronoun

### Examples:

1. **I** know **him**.
   - **I** = subject (before verb "know")
   - **him** = object (after verb "know")

2. **She** loves **me**.
   - **She** = subject (before verb "loves")
   - **me** = object (after verb "loves")

3. **They** are with **us**.
   - **They** = subject (before verb "are")
   - **us** = object (after preposition "with")

### Visual Aid:

```
[SUBJECT PRONOUN] + VERB + [OBJECT PRONOUN]
       ‚Üì                         ‚Üì
  Does action             Receives action
```

## 6. Special Notes

### üìå Note 1: "You" is the same
"**You**" is both subject and object pronoun:
- Subject: "**You** are nice."
- Object: "I like **you**."

### üìå Note 2: "It" is the same
"**It**" is both subject and object pronoun:
- Subject: "**It** is cold."
- Object: "Take **it**."

### üìå Note 3: "He" vs "Him" (Common Mistake!)
‚ùå Wrong: **Him** is a doctor. ‚Üí ‚úÖ Right: **He** is a doctor.
‚ùå Wrong: I know **he**. ‚Üí ‚úÖ Right: I know **him**.

### üìå Note 4: After Prepositions, Use Object Pronouns
Prepositions: with, for, to, at, from, about...
- Come with **me** ‚úÖ (not: Come with I ‚ùå)
- For **him** ‚úÖ (not: For he ‚ùå)
- To **them** ‚úÖ (not: To they ‚ùå)

---

## 7. Common Mistakes to Avoid ‚ö†Ô∏è

### ‚ùå Wrong ‚Üí ‚úÖ Right

1. ‚ùå **Me** am a student ‚Üí ‚úÖ **I** am a student
   - Use "I" as subject, not "me"

2. ‚ùå **Him** is my friend ‚Üí ‚úÖ **He** is my friend
   - Use "he" as subject, not "him"

3. ‚ùå Give it to **I** ‚Üí ‚úÖ Give it to **me**
   - After "to" (preposition), use object pronoun

4. ‚ùå I know **she** ‚Üí ‚úÖ I know **her**
   - After verb, use object pronoun

5. ‚ùå **Us** are happy ‚Üí ‚úÖ **We** are happy
   - Use "we" as subject, not "us"

6. ‚ùå Come with **I** ‚Üí ‚úÖ Come with **me**
   - After "with" (preposition), use object pronoun
