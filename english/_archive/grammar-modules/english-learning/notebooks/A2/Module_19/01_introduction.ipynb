- Form adverbs from adjectives: adjective + -ly

- Place adverbs correctly in sentences
- Distinguish between adjectives and adverbs
- Form comparative and superlative adverbs

**Estimated Time:** 5-6 hours
**Comparison Focus:** adverbs of manner

- adjective + -ly
- irregular adverbs
- adverb placement

### Common Errors to Avoid:
- ❌ more better → ✅ better
- ❌ more taller → ✅ taller
- ❌ She is taller her brother → ✅ She is taller than her brother
