## Task 1: Interview: What Were You Doing?

**Objective:** Interview 3 classmates about what they were doing at different times yesterday

**Instructions:**
- Work in pairs or small groups
- Use past continuous throughout
- Ask and answer questions naturally
- Present findings to the class

**Time:** 15-20 minutes

**Language Focus:**
- What were you doing...?
- I was...
- Were you...?
- Yes, I was. / No, I wasn't.

```
# Task 1: Interview: What Were You Doing?
print("Interview: What Were You Doing?")
print("=" * 60)
print("\nInterview 3 classmates about what they were doing at different times yesterday")
print("\nUseful questions:")
print("- What were you doing at [time]?")
print("- Where were you [verb-ing]?")
print("- Who was [verb-ing] with you?")
print("- Were you [verb-ing] alone?")
print("- How long were you [verb-ing]?")
```
