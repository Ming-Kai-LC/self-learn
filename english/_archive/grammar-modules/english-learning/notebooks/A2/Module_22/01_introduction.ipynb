## Module Overview

Master essential connectors to show cause-result and contrast relationships

### What You'll Learn

1. Use 'because' and 'so' to connect cause and result
2. Understand the difference between 'because' (reason) and 'so' (result)
3. Use 'but' to express contrast between ideas
4. Use 'although' for concession (despite contrast)
5. Apply correct punctuation with connectors
