### Step 2: The 4-Read Strategy

#### Read 1: Skim (2-3 minutes)

**Goal**: Get the gist

**How**:
1. Read title and subtitle
2. Read first sentence of each paragraph
3. Read conclusion
4. Don't stop for unknown words!

**Ask yourself**:
- What is this article about?
- What is the main point?
- Is it news, opinion, analysis?

#### Read 2: Scan for Details (3-5 minutes)

**Goal**: Find specific information

**How**:
1. Read the comprehension questions first
2. Scan article looking for answers
3. Note key numbers, names, dates

**Tip**: You don't need to understand every word to answer questions!

#### Read 3: Deep Read (10-15 minutes)

**Goal**: Full comprehension

**How**:
1. Read carefully from start to finish
2. Try to guess unknown words from context
3. Only look up 3-5 key words (maximum!)
4. Note useful phrases and expressions

**Ask yourself**:
- Do I understand the main arguments?
- What is the author's opinion?
- What are the supporting details?

#### Read 4: Review with Exercises (10-15 minutes)

**Goal**: Consolidate learning

**How**:
1. Complete comprehension questions
2. Do vocabulary matching exercise
3. Check answers
4. Review any mistakes

**Total time**: 25-40 minutes per article

## Part 3: Active Reading Techniques

### Technique 1: Predicting Content

**Before reading**, look at:

**Ask yourself**:

**Example**:

### Technique 2: SQ3R Method

**S**urvey â†’ **Q**uestion â†’ **R**ead â†’ **R**ecite â†’ **R**eview

**Survey**: Skim the article (2 min)

**Question**: Turn headings into questions
- Heading: "Rising Sales"
- Question: "Why are sales rising?"

**Read**: Read to answer your questions

**Recite**: Summarize in your own words (no looking!)

**Review**: Check your summary against article

## Part 4: Vocabulary Learning from Context

### The B2 Vocabulary Challenge

**Problem**: News articles contain many unknown words

**Wrong approach** âŒ: Look up every unknown word
- Interrupts reading flow
- Takes too long
- You forget most words anyway

**Right approach** âœ…: Guess from context
- Faster
- Builds inference skills
- Better retention

### Context Clues Strategies

#### Strategy 1: Definition Clues

**Example**:
"The company uses **sustainable** practices, meaning they operate in ways that don't harm the environment."

**Clue**: "meaning" signals definition
**Guess**: sustainable = environmentally friendly

#### Strategy 2: Example Clues

**Example**:
"Renewable energy sources such as solar, wind, and hydroelectric power are becoming more popular."

**Clue**: examples given (solar, wind, hydroelectric)
**Guess**: renewable = natural, doesn't run out

#### Strategy 3: Contrast Clues

**Example**:
"Unlike traditional cars that produce emissions, electric vehicles are environmentally friendly."

**Clue**: "unlike" signals contrast
**Guess**: emissions = pollution/harmful gases

#### Strategy 4: General Context

**Example**:
"The government implemented new policies to **mitigate** climate change. These measures include carbon taxes and renewable energy incentives."

**Clue**: Context suggests addressing/reducing climate change
**Guess**: mitigate = reduce/lessen

### Which Words to Actually Look Up

**Look up words that are**:
1. **Repeated** multiple times in article
2. **Essential** to main idea
3. **Useful** for other contexts

**Skip words that are**:
1. Only appear once
2. Very technical/specialized
3. You can guess well enough

**Rule of thumb**: Maximum 5-8 words to look up per article

## Part 5: Improving Reading Speed

### Why Speed Matters at B2

**Current B2 learner**: 80-120 words/min  
**B2 target**: 200-250 words/min  
**Native speaker**: 250-300 words/min

**Why you're slow**:
1. **Subvocalization** (saying words in your head)
2. **Regression** (re-reading sentences)
3. **Word-by-word reading** (not reading in chunks)
4. **Dictionary dependence** (looking up too many words)

### Speed Building Techniques

#### Technique 1: Timed Reading

**How**:
1. Choose Breaking News English article
2. Note word count (e.g., 650 words)
3. Set timer
4. Read without stopping
5. Calculate: words Ã· minutes = WPM

**Example**:
- Article: 650 words
- Time: 8 minutes
- Speed: 650 Ã· 8 = 81 WPM

**Weekly goal**: Increase by 10-20 WPM

#### Technique 2: Chunking

**Don't read**: word â†’ by â†’ word â†’ like â†’ this

**Do read**: in phrases / like this / using natural chunks

**Practice**:
```
The government / announced new policies / to combat climate change /
including carbon taxes / and renewable energy incentives.
```

#### Technique 3: Reduce Regression

**Use pointer** (finger, pen, cursor):
- Move smoothly left to right
- Never move backward
- Forces forward momentum
- Increases speed 20-30%

#### Technique 4: Progressive Speed Challenge

**Week 1**: Read at comfortable speed (baseline)
**Week 2**: Force yourself to read 10% faster (less comfortable)
**Week 3**: Return to comfortable speed (now faster than Week 1!)
**Week 4**: Push 10% faster again

**Goal**: Comfortable speed increases over time

````python
# Reading speed tracker and calculator

def calculate_reading_speed(word_count, minutes):
    """Calculate words per minute"""
    wpm = word_count / minutes
    return round(wpm)

def assess_reading_level(wpm, comprehension):
    """Assess reading level based on speed and comprehension"""
    if wpm >= 200 and comprehension >= 70:
        return "B2 - Excellent!"
    elif wpm >= 150 and comprehension >= 70:
        return "B2 - Good progress"
    elif wpm >= 120 and comprehension >= 65:
        return "B1+ - Keep practicing"
    elif wpm < 120:
        return "B1 - Focus on building speed"
    elif comprehension < 65:
        return "Slow down - focus on comprehension first"
    
# Example usage
words = 650
time = 8
comprehension = 75

speed = calculate_reading_speed(words, time)
level = assess_reading_level(speed, comprehension)

print("="*70)
print("READING SPEED ASSESSMENT")
print("="*70)
print(f"\nArticle length: {words} words")
print(f"Reading time: {time} minutes")
print(f"Reading speed: {speed} WPM")
print(f"Comprehension: {comprehension}%")
print(f"\nAssessment: {level}")
print("\nB2 Target: 200+ WPM with 70%+ comprehension")
print("="*70)

# Weekly progress tracker
print("\nðŸ“Š WEEKLY PROGRESS EXAMPLE")
print("="*70)

weekly_progress = [
    {"week": 1, "wpm": 85, "comprehension": 70},
    {"week": 2, "wpm": 95, "comprehension": 72},
    {"week": 3, "wpm": 110, "comprehension": 73},
    {"week": 4, "wpm": 125, "comprehension": 75}
]

for entry in weekly_progress:
    print(f"Week {entry['week']}: {entry['wpm']} WPM, {entry['comprehension']}% comprehension")

improvement = weekly_progress[-1]['wpm'] - weekly_progress[0]['wpm']
print(f"\nâœ… Total improvement: +{improvement} WPM in 4 weeks!")
print("="*70)
````

## Part 7: Common B2 Reading Challenges

### Challenge 1: Unknown Vocabulary Overload

**Problem**: Too many unknown words â†’ frustration

**Solution**:
- Start one level lower than you think (Level 3, not 4)
- Accept 70% understanding (not 100%)
- Look up maximum 5-8 words per article
- Trust context clues

### Challenge 2: Losing Main Idea in Details

**Problem**: Get lost in details, forget overall message

**Solution**:
- Read introduction and conclusion first
- Write one-sentence summary of each paragraph
- Ask: "What's the big picture here?"
- Use SQ3R method

### Challenge 3: Complex Sentence Structures

**Problem**: Long sentences with multiple clauses

**Example**:
"The scientists, who have been studying climate change for over two decades and have published numerous papers on the subject, warn that sea levels could rise faster than previously predicted unless immediate action is taken."

**Solution**:
1. Find the main subject and verb:
   - Subject: "The scientists"
   - Verb: "warn"
2. Identify core message: "Scientists warn sea levels could rise"
3. Add details back in: "unless action is taken", "faster than predicted"

### Challenge 4: Staying Focused

**Problem**: Mind wanders while reading

**Solution**:
- Use active reading techniques (annotation, summarizing)
- Read in shorter sessions (15 min)
- Choose interesting topics
- Eliminate distractions (phone away)
- Take breaks between paragraphs

### Challenge 5: Reading Too Slowly

**Problem**: Takes 20+ min to read 500-word article

**Solution**:
- Use pointer/finger to maintain pace
- Reduce subvocalization
- Don't re-read unless necessary
- Practice timed reading weekly
- Gradually increase target speed
