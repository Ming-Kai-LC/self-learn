## Part 2: Reported Speech Fundamentals

### Basic Transformation Rules

#### Rule 1: Tense Backshift

When the reporting verb is in the past (said, told, asked), shift tenses back:

| Direct Speech | Reported Speech |
|---------------|----------------|
| Present Simple: "I **work** here" | Past Simple: He said he **worked** there |
| Present Continuous: "I'm **working**" | Past Continuous: She said she **was working** |
| Past Simple: "I **worked**" | Past Perfect: He said he **had worked** |
| Present Perfect: "I **have worked**" | Past Perfect: She said she **had worked** |
| Will: "I **will finish**" | Would: He said he **would finish** |
| Can: "I **can help**" | Could: She said she **could help** |
| Must: "I **must go**" | Had to: He said he **had to go** |

#### Rule 2: Pronoun Changes

**Direct**: "**I** will send **you** the report."
- John said to me ‚Üì

**Reported**: John said **he** would send **me** the report.

#### Rule 3: Time/Place Changes

| Direct | Reported |
|--------|----------|
| today | that day |
| tomorrow | the next day / the following day |
| yesterday | the day before / the previous day |
| next week | the following week |
| last week | the previous week |
| here | there |
| this | that |
| these | those |

**Example**:
- Direct: "I'll finish this **today**."
- Reported: She said she would finish that **that day**.

### When NOT to Backshift

**No backshift if**:

1. **The statement is still true**:
   - Direct: "The office **is** in London."
   - Reported: He said the office **is** in London. (still true)

2. **General truths**:
   - Direct: "Water **boils** at 100¬∞C."
   - Reported: She said water **boils** at 100¬∞C.

3. **Recent past with present reporting**:
   - Direct: "I **finished** the report." (just now)
   - Reported: John says he **has finished** the report.

## Part 3: Reporting Statements

### Common Reporting Verbs for Statements

**Beyond "said"**, use varied reporting verbs:

#### Neutral Reporting
- **said** (that): She **said** she would attend.
- **told** + person + (that): He **told me** the deadline was Friday.
- **mentioned** (that): They **mentioned** they were interested.
- **reported** (that): The team **reported** progress was on track.
- **stated** (that): The CEO **stated** profits had increased.

#### Emphatic/Strong
- **insisted** (that): She **insisted** the data was accurate.
- **emphasized** (that): He **emphasized** quality was priority.
- **claimed** (that): They **claimed** they had met requirements.
- **announced** (that): The company **announced** it would expand.

#### Adding Information
- **added** (that): She **added** the budget was approved.
- **explained** (that): He **explained** the process was complex.
- **confirmed** (that): They **confirmed** the meeting was scheduled.
- **informed** + person + (that): She **informed us** the project was delayed.

### Structure Patterns

**Pattern 1**: Subject + said/mentioned/reported + (that) + clause
- The manager **mentioned** the deadline had been extended.

**Pattern 2**: Subject + told/informed + person + (that) + clause
- She **told me** she would review the proposal.

**Pattern 3**: Subject + announced/stated + (that) + clause
- The company **announced** it was hiring new staff.

**Note**: "that" is optional but recommended for clarity in formal writing

## Part 4: Reporting Questions

### Yes/No Questions

**Structure**: asked + if/whether + statement order

**Direct**: "**Do you** have the report?"
**Reported**: She asked **if I had** the report.

**Direct**: "**Can you** attend the meeting?"
**Reported**: He asked **whether I could** attend the meeting.

**Important**: Change to statement word order (not question order)
- ‚ùå She asked if **did I have** the report
- ‚úÖ She asked if **I had** the report

### Wh- Questions

**Structure**: asked + wh-word + statement order

**Direct**: "**When will** you finish?"
**Reported**: He asked **when I would** finish.

**Direct**: "**What is** the deadline?"
**Reported**: She asked **what the deadline was**.

**Direct**: "**Who is** responsible?"
**Reported**: They asked **who was** responsible.

### Reporting Verbs for Questions

- **asked**: neutral
- **wondered**: thinking/curiosity
- **wanted to know**: direct interest
- **inquired**: formal

**Examples**:
- The client **asked** when we could deliver.
- She **wondered** if the budget was sufficient.
- They **wanted to know** who would lead the project.
- He **inquired** whether we had received the documents.

## Part 5: Reporting Requests and Commands

### Requests (Polite)

**Structure**: asked + person + to + infinitive

**Direct**: "Could you **send** the report?"
**Reported**: She **asked me to send** the report.

**Direct**: "Would you **review** this, please?"
**Reported**: He **asked me to review** it.

**Direct**: "Can you **attend** the meeting?"
**Reported**: They **asked me to attend** the meeting.

### Commands (Direct Orders)

**Structure**: told/ordered + person + to + infinitive

**Direct**: "**Submit** the report by Friday."
**Reported**: The manager **told us to submit** the report by Friday.

**Direct**: "**Complete** the training."
**Reported**: HR **instructed employees to complete** the training.

### Negative Requests/Commands

**Structure**: asked/told + person + not + to + infinitive

**Direct**: "**Don't forget** the deadline."
**Reported**: She **told me not to forget** the deadline.

**Direct**: "**Please don't share** this information."
**Reported**: He **asked us not to share** the information.

### Reporting Verbs for Requests/Commands

**Polite requests**:
- asked
- requested
- invited

**Strong requests/commands**:
- told
- instructed
- ordered
- required

**Suggestions**:
- suggested + gerund: She **suggested reviewing** the data.
- recommended + gerund: He **recommended postponing** the launch.
- advised + person + to: They **advised us to wait**.

## Part 8: Common Errors and How to Fix Them

### Error 1: Wrong Word Order in Reported Questions

‚ùå She asked **when would I finish** the report.  
‚úÖ She asked **when I would finish** the report.

**Rule**: Use statement word order, not question word order

### Error 2: Forgetting to Backshift Tenses

‚ùå He said he **is** busy. (reporting past statement)  
‚úÖ He said he **was** busy.

**Rule**: Backshift when reporting verb is past

### Error 3: Using "Say" Instead of "Tell"

‚ùå She **said me** the deadline was Friday.  
‚úÖ She **told me** the deadline was Friday.

**Rule**: 
- **Say** (that): She said (that) the deadline was Friday.
- **Tell** + person: She told me the deadline was Friday.

### Error 4: Using "That" After Asked

‚ùå He asked **that** I could attend.  
‚úÖ He asked **if/whether** I could attend.

**Rule**: Use if/whether for yes/no questions, not "that"

### Error 5: Incorrect Infinitive Structure for Requests

‚ùå She asked me **that I send** the report.  
‚úÖ She asked me **to send** the report.

**Rule**: Use "to + infinitive" for requests, not "that + clause"

### Error 6: Forgetting Pronoun Changes

‚ùå John said to me: "**I** will help **you**" ‚Üí John said **I** would help **you**.  
‚úÖ John said **he** would help **me**.

**Rule**: Change pronouns based on perspective

### Error 7: Not Changing Time References

‚ùå Yesterday, she said: "I'll finish it tomorrow" ‚Üí She said she would finish it **tomorrow**.  
‚úÖ She said she would finish it **the next day**.

**Rule**: Adjust time references (tomorrow ‚Üí the next day, etc.)
