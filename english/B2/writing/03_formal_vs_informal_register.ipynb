## Register Levels in Business English

### The Register Scale

**Very Informal** (casual chat):  
"Hey! Wanna grab coffee later?"

**Informal** (friendly colleague):  
"Hi! Would you like to get coffee this afternoon?"

**Neutral Professional** (standard business):  
"Good morning. Would you be available for a coffee meeting this afternoon?"

**Formal** (senior management, clients):  
"Good morning. I would like to invite you to a brief meeting this afternoon to discuss the project."

**Very Formal** (legal, official documents):  
"Dear Sir/Madam, I am writing to request a meeting at your earliest convenience."

### When to Use Each Level

| Situation | Register | Example |
|-----------|----------|----------|
| Close colleague (chat) | Informal | "Can you check this?" |
|Team email | Neutral | "Could you please review this document?" |
| Client email | Formal | "I would appreciate your feedback on the attached proposal." |
| Board report | Very Formal | "This document presents the quarterly performance analysis." |

## Formal vs Informal: Key Differences

### 1. Vocabulary Choices

| Informal | Formal | Example |
|----------|--------|----------|
| get | receive, obtain | We will **obtain** the data next week |
| ask for | request | I would like to **request** an extension |
| find out | discover, determine | We need to **determine** the cause |
| tell | inform, notify | Please **inform** the team |
| think about | consider | We should **consider** other options |
| show | demonstrate, illustrate | The data **demonstrates** growth |
| help | assist, facilitate | This will **facilitate** the process |
| big | significant, substantial | A **significant** increase |
| good | beneficial, effective | An **effective** solution |

### 2. Contractions

**Informal**: ❌ Don't, can't, we'll, it's  
**Formal**: ✅ Do not, cannot, we will, it is

### 3. Phrasal Verbs → Single Verbs

| Phrasal Verb (Informal) | Single Verb (Formal) |
|-------------------------|------------------------|
| put off | postpone |
| call off | cancel |
|bring up | mention, raise |
| look into | investigate |
| come up with | propose, develop |
| go over | review |
| set up | establish, arrange |

### 4. Sentence Structure

**Informal**: Short, simple sentences  
"The project is delayed. We need more time. Can we extend the deadline?"

**Formal**: Complex sentences with linking  
"Due to unforeseen delays in the project timeline, we would like to request a deadline extension."

````python
# Practice: Transform informal to formal

transformations = [
    {"informal": "Thanks for your email.", "formal": "Thank you for your email."},
    {"informal": "Sorry for the delay.", "formal": "I apologize for the delay."},
    {"informal": "I'll get back to you soon.", "formal": "I will respond at my earliest convenience."},
    {"informal": "Can you send me the report?", "formal": "Could you please forward the report?"},
    {"informal": "We need to talk about this.", "formal": "We need to discuss this matter."},
]

print("FORMAL TRANSFORMATION PRACTICE")
print("="*60)
for i, t in enumerate(transformations, 1):
    print(f"\n{i}. Informal: {t['informal']}")
    print(f"   Formal:   {t['formal']}")
````

## Common Register Mistakes

### Mistake 1: Too Casual for Context

❌ "Hey! Just wanted to let you know the meeting got pushed back."  
✅ "I am writing to inform you that the meeting has been rescheduled."

### Mistake 2: Mixing Registers

❌ "I would like to request an extension because stuff came up."  
✅ "I would like to request an extension due to unforeseen circumstances."

### Mistake 3: Overly Formal for Internal Team

❌ "I am writing to humbly request that you might consider reviewing..." (to close colleague)  
✅ "Could you please review this when you have a chance?"
