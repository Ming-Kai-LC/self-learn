## Part 4: Understanding Gliglish Corrections

### Types of Corrections You'll See

#### 1. Grammar Corrections

**Example**:
- ‚ùå You said: "I have work here for 3 years"
- ‚úÖ Correction: "I have **worked** here for 3 years"
- üìù Rule: Present perfect needs past participle

#### 2. Word Choice Corrections

**Example**:
- ‚ùå You said: "I want to say my opinion"
- ‚úÖ Correction: "I'd like to **share** my opinion"
- üìù Rule: "Share opinion" is more natural than "say opinion"

#### 3. Pronunciation Feedback

**Example**:
- ‚ùå You said: /pro-JECT/ (noun)
- ‚úÖ Correction: /PRO-ject/ (noun) vs. /pro-JECT/ (verb)
- üìù Rule: Noun stress on first syllable

#### 4. Fluency Suggestions

**Example**:
- ‚ùå "Um... I think... well... maybe we should..."
- ‚úÖ "I think we should..."
- üìù Rule: Reduce fillers for professional speech

### How to Use Corrections Effectively

**During conversation**:
1. Note the correction (Gliglish shows it in real-time)
2. Repeat the corrected version aloud immediately
3. Continue conversation using correct form

**After conversation**:
1. Review conversation history in Gliglish
2. Identify patterns (same error multiple times?)
3. Write down top 3 errors in error log
4. Practice corrected versions with AI tutor

**Pro tip**: If same error appears 3+ times, create a ChatGPT drill:
```
"I keep making this error: [your error]. 
Please create 10 practice sentences for me to say correctly.
After each, confirm if I've used it right."
```

## Part 7: Common B2 Speaking Errors (And How Gliglish Fixes Them)

### Error Category 1: Present Perfect vs. Simple Past

‚ùå **Common mistake**: "I work here for 5 years"  
‚úÖ **Gliglish correction**: "I've worked here for 5 years"  
üìù **Rule**: Use present perfect for actions starting in past, continuing now

‚ùå "Did you finish the report yet?"  
‚úÖ "Have you finished the report yet?"  
üìù **Rule**: "Yet" typically pairs with present perfect

### Error Category 2: Preposition Mistakes

‚ùå "I'm responsible of this project"  
‚úÖ "I'm responsible **for** this project"  

‚ùå "We need to focus in quality"  
‚úÖ "We need to focus **on** quality"  

‚ùå "I'm looking forward to meet you"  
‚úÖ "I'm looking forward to **meeting** you"  
üìù **Rule**: "To" here is a preposition, needs gerund

### Error Category 3: Article Usage (The/A/An/Zero)

‚ùå "I work in IT industry"  
‚úÖ "I work in **the** IT industry"  

‚ùå "The teamwork is important"  
‚úÖ "Teamwork is important" (abstract concept, no article)  

‚ùå "I need a advice"  
‚úÖ "I need **some** advice" (advice is uncountable)  

### Error Category 4: Word Order in Questions

‚ùå "You can tell me where is the meeting?"  
‚úÖ "Can you tell me where **the meeting is**?"  
üìù **Rule**: Embedded questions use statement word order

### Error Category 5: Modal Verb Mistakes

‚ùå "I must to finish this today"  
‚úÖ "I must finish this today" (no "to" after modals)  

‚ùå "We should to consider other options"  
‚úÖ "We should consider other options"  

### How Gliglish Helps

**Real-time correction**: You see the error immediately while speaking  
**Pattern recognition**: After 3-5 instances, you internalize the correction  
**No embarrassment**: Practice mistakes safely before real conversations  
**Conversation history**: Review all corrections after session

## Part 8: Advanced Techniques - Fluency Building

### Technique 1: Reduce Fillers

**Common B2 problem**: Too many "um", "uh", "like", "you know"

**Gliglish shows**:
- Your filler count per session
- Where fillers appear (usually before complex sentences)

**Solution**:
1. Pause silently instead of saying "um"
2. Use professional fillers: "Let me think...", "That's a good question..."
3. Practice with Gliglish: Force yourself to speak without fillers

**Weekly goal**: Reduce filler count by 20%

### Technique 2: Expand Answers (Avoid One-Word Responses)

‚ùå **Weak B2**:  
Q: "Do you enjoy your job?"  
A: "Yes"  

‚úÖ **Strong B2**:  
Q: "Do you enjoy your job?"  
A: "Yes, I do. I particularly enjoy the project management aspect because it allows me to work with different teams. The most challenging part is managing deadlines, but that's also what makes it interesting."

**Formula**: **Answer + Reason + Example/Detail**

### Technique 3: Sophisticated Connectors

**Beginner connectors**: and, but, because  
**B2 connectors**: however, although, whereas, furthermore, consequently

**Practice in Gliglish**:
- "**Although** we faced delays, we still met the deadline"
- "**Consequently**, we need to revise our approach"
- "**Furthermore**, this solution is more cost-effective"

### Technique 4: Self-Correction

**B2 speakers can self-correct naturally**:
- "We need to... sorry, I mean we should consider..."
- "The meeting is on... actually, let me check... it's on Tuesday"

**Don't aim for perfection**: Native speakers self-correct too!

### Technique 5: Speed Control

**Not too fast**: Clarity > Speed  
**Not too slow**: Sounds unnatural

**Target**: 140-160 words per minute (natural professional pace)

**Gliglish shows**: Your speaking speed, helps you adjust
