## What is n8n?

n8n (pronounced "n-eight-n") is a powerful **workflow automation tool** that lets you connect different apps and services together without writing complex code.

Think of it like this:
- **IFTTT** (If This Then That) - but more powerful and self-hosted
- **Zapier** - but you own your data and it's free
- **Microsoft Power Automate** - but open-source and more flexible

### Real-World Examples

Here's what you can automate with n8n:

1. **Email Management**: Auto-save Gmail attachments to Google Drive
2. **Social Media**: Post the same content to Twitter, LinkedIn, and Facebook simultaneously
3. **Data Sync**: Keep your CRM updated when new leads come from your website
4. **Notifications**: Get Telegram messages when your website goes down
5. **Report Generation**: Create weekly reports from your business data automatically

### Hardware Requirements

| Component | Minimum | Recommended | Production |
|-----------|---------|-------------|------------|
| **RAM** | 4 GB | 8 GB | 16 GB |
| **CPU** | 2 cores | 4 cores | 4-8 cores |
| **Storage** | 20 GB free | 50 GB free | 100+ GB |
| **Disk Type** | HDD | SSD | NVMe SSD |

**Your Use Case**:
- **Testing/Learning**: Minimum specs are fine
- **Personal Automation**: Recommended specs
- **Business Critical**: Production specs required

````python
# Let's create a simple system check helper
import platform
import psutil
import os

def check_system_requirements():
    """
    Check if the system meets minimum requirements for n8n.
    This runs on Linux (Jupyter), but shows what to check on Windows.
    """
    print("=== System Information ===")
    print(f"Platform: {platform.system()}")
    print(f"Platform Version: {platform.version()}")
    print(f"Architecture: {platform.machine()}")
    print()
    
    # Check RAM
    ram_gb = psutil.virtual_memory().total / (1024**3)
    print(f"Total RAM: {ram_gb:.2f} GB")
    
    if ram_gb >= 16:
        print("  âœ… Excellent for production use")
    elif ram_gb >= 8:
        print("  âœ… Good for personal automation")
    elif ram_gb >= 4:
        print("  âš ï¸  Minimum met - testing/learning only")
    else:
        print("  âŒ Insufficient RAM - need at least 4GB")
    print()
    
    # Check CPU
    cpu_count = psutil.cpu_count(logical=False)
    print(f"CPU Cores (Physical): {cpu_count}")
    
    if cpu_count >= 4:
        print("  âœ… Good for production")
    elif cpu_count >= 2:
        print("  âœ… Sufficient for learning and personal use")
    else:
        print("  âš ï¸  May be slow with complex workflows")
    print()
    
    # Check disk space
    disk = psutil.disk_usage('/')
    free_gb = disk.free / (1024**3)
    print(f"Free Disk Space: {free_gb:.2f} GB")
    
    if free_gb >= 100:
        print("  âœ… Plenty of space")
    elif free_gb >= 50:
        print("  âœ… Good for most uses")
    elif free_gb >= 20:
        print("  âš ï¸  Minimum space available")
    else:
        print("  âŒ Insufficient disk space")
    
# Run the check
check_system_requirements()
````

## Choosing Your Installation Method

There are three main ways to install n8n on Windows:

### 1. npm (Node.js) Installation â­ RECOMMENDED FOR BEGINNERS

**Best For**: Learning, testing, personal automation

**Pros**:
- âœ… Fastest to set up (under 10 minutes)
- âœ… Works on Windows Home edition
- âœ… Lowest RAM usage (100-320 MB)
- âœ… Best battery life on laptops
- âœ… Easy to update

**Cons**:
- âŒ Not isolated from other apps
- âŒ Harder to manage multiple versions
- âŒ Not production-ready for critical workflows

### 2. Docker Desktop Installation

**Best For**: Production use, team environments

**Pros**:
- âœ… Production-ready isolation
- âœ… Easy version management
- âœ… Clean rollback capabilities
- âœ… Consistent across machines

**Cons**:
- âŒ Requires Windows 10/11 Pro or Enterprise
- âŒ High RAM overhead (2-4 GB)
- âŒ Worse battery life
- âŒ More complex troubleshooting

### 3. WSL2 (Linux on Windows)

**Best For**: Linux-focused developers

**Pros**:
- âœ… Works on Windows Home
- âœ… Linux-native experience
- âœ… Better performance than Docker sometimes

**Cons**:
- âŒ Most complex setup
- âŒ Requires managing Windows + Linux
- âŒ Memory management issues

````python
# Installation method decision helper

def recommend_installation_method():
    """
    Interactive helper to recommend the best installation method.
    """
    print("=== n8n Installation Method Recommender ===")
    print()
    
    # Simulate user inputs (in real usage, these would be interactive)
    windows_edition = "home"  # Options: home, pro, enterprise
    use_case = "learning"  # Options: learning, personal, production
    laptop_or_desktop = "laptop"  # Options: laptop, desktop
    
    print("Based on your configuration:")
    print(f"  Windows Edition: {windows_edition.title()}")
    print(f"  Use Case: {use_case.title()}")
    print(f"  Device Type: {laptop_or_desktop.title()}")
    print()
    
    # Recommendation logic
    if use_case == "learning":
        method = "npm (Node.js)"
        reason = "Fastest setup and lowest resource usage for learning"
    elif use_case == "production" and windows_edition in ["pro", "enterprise"]:
        method = "Docker Desktop"
        reason = "Production-ready with isolation and easy version management"
    elif use_case == "production" and windows_edition == "home":
        method = "WSL2"
        reason = "Docker not available on Home, WSL2 provides production capabilities"
    elif laptop_or_desktop == "laptop":
        method = "npm (Node.js)"
        reason = "Better battery life and lower resource usage on laptops"
    else:
        method = "npm (Node.js)"
        reason = "Most versatile option for personal automation"
    
    print(f"âœ… RECOMMENDED: {method}")
    print(f"   Reason: {reason}")
    print()
    print("Next Steps:")
    if method == "npm (Node.js)":
        print("  â†’ Continue to Module 02: npm Installation")
    elif method == "Docker Desktop":
        print("  â†’ Continue to Module 03: Docker Desktop Installation")
    else:
        print("  â†’ See documentation for WSL2 setup")

# Run the recommender
recommend_installation_method()
````
