### Step 2.1: Trend Indicators

---

```python
# TODO: Implement all indicator functions from scratch
# Use the functions you learned in Modules 04-07

# Function library (complete these based on previous modules)

def calculate_sma(prices, period):
    """Calculate Simple Moving Average"""
    return prices.rolling(window=period).mean()

def calculate_ema(prices, period):
    """Calculate Exponential Moving Average"""
    alpha = 2 / (period + 1)
    return prices.ewm(alpha=alpha, adjust=False).mean()

def calculate_macd(prices, fast=12, slow=26, signal=9):
    """Calculate MACD, Signal Line, and Histogram"""
    ema_fast = calculate_ema(prices, fast)
    ema_slow = calculate_ema(prices, slow)
    macd_line = ema_fast - ema_slow
    signal_line = calculate_ema(macd_line, signal)
    histogram = macd_line - signal_line
    
    return pd.DataFrame({
        'MACD': macd_line,
        'Signal': signal_line,
        'Histogram': histogram
    })

def calculate_rsi(prices, period=14):
    """Calculate RSI with Wilder's smoothing"""
    delta = prices.diff()
    gains = delta.where(delta > 0, 0)
    losses = -delta.where(delta < 0, 0)
    
    avg_gains = gains.ewm(alpha=1/period, adjust=False).mean()
    avg_losses = losses.ewm(alpha=1/period, adjust=False).mean()
    
    rs = avg_gains / avg_losses
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_stochastic(high, low, close, k_period=14, d_period=3):
    """Calculate Stochastic Oscillator"""
    lowest_low = low.rolling(window=k_period).min()
    highest_high = high.rolling(window=k_period).max()
    
    k = ((close - lowest_low) / (highest_high - lowest_low)) * 100
    d = k.rolling(window=d_period).mean()
    
    return pd.DataFrame({'%K': k, '%D': d})

def calculate_bollinger_bands(prices, period=20, num_std=2):
    """Calculate Bollinger Bands"""
    sma = calculate_sma(prices, period)
    std = prices.rolling(window=period).std()
    
    upper = sma + (std * num_std)
    lower = sma - (std * num_std)
    
    return pd.DataFrame({
        'Middle': sma,
        'Upper': upper,
        'Lower': lower,
        '%B': (prices - lower) / (upper - lower),
        'Bandwidth': ((upper - lower) / sma) * 100
    })

def calculate_atr(high, low, close, period=14):
    """Calculate Average True Range"""
    tr1 = high - low
    tr2 = abs(high - close.shift())
    tr3 = abs(low - close.shift())
    
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    atr = tr.ewm(alpha=1/period, adjust=False).mean()
    
    return atr

print("âœ“ Indicator functions loaded")
print("âœ“ Ready to calculate all technical indicators")
```

```python
# Calculate all indicators
print("Calculating all technical indicators...")
print("="*80)

# Trend indicators
stock_data['SMA_20'] = calculate_sma(stock_data['Close'], 20)
stock_data['SMA_50'] = calculate_sma(stock_data['Close'], 50)
stock_data['SMA_200'] = calculate_sma(stock_data['Close'], 200)
stock_data['EMA_12'] = calculate_ema(stock_data['Close'], 12)
stock_data['EMA_26'] = calculate_ema(stock_data['Close'], 26)

# MACD
macd_data = calculate_macd(stock_data['Close'])
stock_data = pd.concat([stock_data, macd_data], axis=1)

# Momentum indicators
stock_data['RSI'] = calculate_rsi(stock_data['Close'])
stoch_data = calculate_stochastic(stock_data['High'], stock_data['Low'], stock_data['Close'])
stock_data = pd.concat([stock_data, stoch_data], axis=1)

# Volatility indicators
bb_data = calculate_bollinger_bands(stock_data['Close'])
stock_data = pd.concat([stock_data, bb_data], axis=1)
stock_data['ATR'] = calculate_atr(stock_data['High'], stock_data['Low'], stock_data['Close'])

# Calculate returns
stock_data['Returns'] = stock_data['Close'].pct_change()

print("âœ“ All indicators calculated!")
print(f"\nTotal columns: {len(stock_data.columns)}")
print("\nIndicator columns:")
indicator_cols = [col for col in stock_data.columns if col not in ['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']]
for col in sorted(indicator_cols):
    print(f"  - {col}")
```

```python
# Calculate risk metrics
print("ðŸ“Š RISK METRICS ANALYSIS")
print("="*80)

# Volatility measures
daily_returns = stock_data['Returns'].dropna()
volatility_daily = daily_returns.std()
volatility_annual = volatility_daily * np.sqrt(252)  # Annualized

print("\nðŸ“ˆ Volatility Measures:")
print("-"*80)
print(f"Daily volatility: {volatility_daily * 100:.2f}%")
print(f"Annualized volatility: {volatility_annual * 100:.2f}%")
print(f"Average daily return: {daily_returns.mean() * 100:.3f}%")
print(f"Best day: {daily_returns.max() * 100:.2f}%")
print(f"Worst day: {daily_returns.min() * 100:.2f}%")

# Risk assessment
if volatility_annual < 0.15:
    risk_level = "LOW (< 15% annual volatility)"
elif volatility_annual < 0.25:
    risk_level = "MODERATE (15-25% annual volatility)"
elif volatility_annual < 0.40:
    risk_level = "HIGH (25-40% annual volatility)"
else:
    risk_level = "VERY HIGH (> 40% annual volatility)"

print(f"\nRisk Level: {risk_level}")

# Sharpe-like ratio (simplified, assuming risk-free rate = 0)
sharpe = (daily_returns.mean() / daily_returns.std()) * np.sqrt(252)
print(f"Sharpe Ratio (simplified): {sharpe:.2f}")

# Maximum drawdown
cumulative = (1 + daily_returns).cumprod()
running_max = cumulative.expanding().max()
drawdown = (cumulative - running_max) / running_max
max_dd = drawdown.min() * 100

print(f"\nMaximum Drawdown: {max_dd:.2f}%")
print(f"Recovery gain needed: {(abs(max_dd) / (100 - abs(max_dd))) * 100:.2f}%")
```

```python
# TRADING PARAMETERS (TODO: Customize these)
ACCOUNT_SIZE = 50000  # TODO: Your account size in RM
RISK_PER_TRADE = 0.02  # TODO: Risk percentage (2% = 0.02)
ENTRY_PRICE = current['Close']  # Current price

# Determine stop loss based on ATR
ATR_MULTIPLIER = 2  # 2x ATR for stop loss
STOP_LOSS = ENTRY_PRICE - (current['ATR'] * ATR_MULTIPLIER)

# Determine target based on risk/reward ratio
RISK_REWARD_RATIO = 2  # Target is 2x the risk
TARGET_PRICE = ENTRY_PRICE + ((ENTRY_PRICE - STOP_LOSS) * RISK_REWARD_RATIO)

print("ðŸ’¼ POSITION SIZING CALCULATION")
print("="*80)
print("\nAccount Information:")
print("-"*80)
print(f"Account size: RM {ACCOUNT_SIZE:,.2f}")
print(f"Risk per trade: {RISK_PER_TRADE * 100:.1f}%")
print(f"Maximum risk amount: RM {ACCOUNT_SIZE * RISK_PER_TRADE:,.2f}")

print("\nTrade Setup:")
print("-"*80)
print(f"Entry price: RM {ENTRY_PRICE:.2f}")
print(f"Stop loss: RM {STOP_LOSS:.2f} ({((ENTRY_PRICE - STOP_LOSS) / ENTRY_PRICE * 100):.2f}% below entry)")
print(f"Target price: RM {TARGET_PRICE:.2f} ({((TARGET_PRICE - ENTRY_PRICE) / ENTRY_PRICE * 100):.2f}% above entry)")
print(f"Risk/Reward ratio: 1:{RISK_REWARD_RATIO}")

# Calculate position size
risk_per_share = ENTRY_PRICE - STOP_LOSS
risk_amount = ACCOUNT_SIZE * RISK_PER_TRADE
num_shares = int(risk_amount / risk_per_share)
position_value = num_shares * ENTRY_PRICE
position_pct = (position_value / ACCOUNT_SIZE) * 100

print("\nPosition Sizing:")
print("-"*80)
print(f"Risk per share: RM {risk_per_share:.2f}")
print(f"Number of shares to buy: {num_shares:,} shares")
print(f"Total position value: RM {position_value:,.2f} ({position_pct:.1f}% of account)")

print("\nScenario Analysis:")
print("-"*80)
# If stopped out
loss_if_stopped = num_shares * (ENTRY_PRICE - STOP_LOSS)
print(f"If stopped out:")
print(f"  Loss: RM {loss_if_stopped:,.2f} ({(loss_if_stopped / ACCOUNT_SIZE * 100):.2f}% of account)")
print(f"  New account value: RM {ACCOUNT_SIZE - loss_if_stopped:,.2f}")

# If target hit
profit_if_target = num_shares * (TARGET_PRICE - ENTRY_PRICE)
print(f"\nIf target hit:")
print(f"  Profit: RM {profit_if_target:,.2f} ({(profit_if_target / ACCOUNT_SIZE * 100):.2f}% of account)")
print(f"  New account value: RM {ACCOUNT_SIZE + profit_if_target:,.2f}")

print("\n" + "="*80)
print(f"ðŸ’¡ Summary: Risk RM {loss_if_stopped:,.0f} to potentially make RM {profit_if_target:,.0f}")
```
