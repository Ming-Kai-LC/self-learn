# Jupyter Keyboard Shortcuts Guide

Master these shortcuts to become 10x more productive!

**Note**: Jupyter has two modes:
- **Command Mode** (blue border): Press `Esc` to enter
- **Edit Mode** (green border): Press `Enter` to enter

---

## Essential Shortcuts (Learn These First!)

| Shortcut | Mode | Action |
|----------|------|--------|
| `Shift + Enter` | Both | Run cell, select below |
| `Ctrl + Enter` | Both | Run cell |
| `Alt + Enter` | Both | Run cell, insert below |
| `Esc` | Edit | Enter command mode |
| `Enter` | Command | Enter edit mode |
| `Ctrl + S` | Both | Save notebook |

## Navigation (Command Mode)

| Shortcut | Action |
|----------|--------|
| `↑` or `K` | Select cell above |
| `↓` or `J` | Select cell below |
| `Shift + ↑/↓` | Select multiple cells |
| `Ctrl + Home` | Go to first cell |
| `Ctrl + End` | Go to last cell |

## Cell Creation (Command Mode)

| Shortcut | Action |
|----------|--------|
| `A` | Insert cell above |
| `B` | Insert cell below |

## Cell Editing (Command Mode)

| Shortcut | Action |
|----------|--------|
| `X` | Cut cell |
| `C` | Copy cell |
| `V` | Paste cell below |
| `Shift + V` | Paste cell above |
| `D D` | Delete cell (press D twice) |
| `Z` | Undo cell deletion |
| `Shift + M` | Merge selected cells |

## Cell Type (Command Mode)

| Shortcut | Action |
|----------|--------|
| `M` | Change to markdown |
| `Y` | Change to code |
| `R` | Change to raw |
| `1-6` | Change to heading (markdown) |

## Running Cells

| Shortcut | Mode | Action |
|----------|------|--------|
| `Shift + Enter` | Both | Run cell, move to next |
| `Ctrl + Enter` | Both | Run cell, stay |
| `Alt + Enter` | Both | Run cell, insert below |
| `Ctrl + Shift + -` | Edit | Split cell at cursor |

## Edit Mode Shortcuts

| Shortcut | Action |
|----------|--------|
| `Tab` | Auto-complete or indent |
| `Shift + Tab` | Tooltip/documentation |
| `Ctrl + ]` | Indent |
| `Ctrl + [` | Dedent |
| `Ctrl + A` | Select all in cell |
| `Ctrl + Z` | Undo |
| `Ctrl + Y` | Redo |
| `Ctrl + /` | Toggle comment |

## Command Palette and Help

| Shortcut | Mode | Action |
|----------|------|--------|
| `Ctrl + Shift + P` | Both | Open command palette |
| `H` | Command | Show keyboard shortcuts |
| `Shift + ?` | Both | Show keyboard shortcuts |

## View Options (Command Mode)

| Shortcut | Action |
|----------|--------|
| `L` | Toggle line numbers |
| `Shift + L` | Toggle line numbers (all cells) |
| `O` | Toggle output |
| `Shift + O` | Toggle output scrolling |

## Find and Replace

| Shortcut | Mode | Action |
|----------|------|--------|
| `Ctrl + F` | Edit | Find |
| `Ctrl + H` | Edit | Find and replace |

## Kernel Operations (Command Mode)

| Shortcut | Action |
|----------|--------|
| `I I` | Interrupt kernel (press I twice) |
| `0 0` | Restart kernel (press 0 twice) |

## Multi-Cell Operations

| Shortcut | Action |
|----------|--------|
| `Shift + ↑/↓` | Select multiple cells |
| `Shift + M` | Merge selected cells |
| `Shift + J` or `Shift + ↓` | Extend selection down |
| `Shift + K` or `Shift + ↑` | Extend selection up |

## Pro Tips

### Workflow Tips
1. Use `B` repeatedly to quickly create multiple cells
2. Use `Shift + Enter` to run and advance through notebook
3. Use `DD` to quickly delete unwanted cells
4. Use `Shift + Tab` for quick documentation lookup

### Speed Combos
- Create and enter new cell: `B` then `Enter`
- Run and create new cell: `Alt + Enter`
- Copy cell multiple times: `C` then `V V V`
- Delete multiple cells: Select with `Shift + ↓` then `DD`

### Editing Tips
- `Ctrl + /`: Comment/uncomment multiple lines
- `Ctrl + D`: Delete entire line
- `Tab`: Indent or autocomplete
- `Shift + Tab`: Show function signature

## Custom Shortcuts

You can customize shortcuts:
1. Go to Settings → Advanced Settings Editor
2. Select Keyboard Shortcuts
3. Add custom shortcuts in JSON format

## Practice Challenge

Try completing these tasks using ONLY shortcuts:

1. Create 5 cells below (`BBBBB`)
2. Select all 5 (`Shift + ↓↓↓↓`)
3. Delete them (`DD`)
4. Undo deletion (`Z`)
5. Convert first to markdown (`M`)
6. Merge all (`Shift + M`)
7. Delete the merged cell (`DD`)

Time yourself - can you do it in under 10 seconds?
