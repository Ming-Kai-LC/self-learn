### File Status Indicators

| Symbol | Meaning |
|--------|----------|
| **M** | Modified - file has changes |
| **A** | Added - new file |
| **D** | Deleted - file removed |
| **R** | Renamed - file moved |
| **C** | Conflict - merge conflict |
| **U** | Untracked - not in Git yet |

### Understanding Changes vs Staged Changes

**CHANGES**: Files you've modified but haven't staged yet
- These are in your working directory
- Won't be included in next commit unless staged

**STAGED CHANGES**: Files ready to commit
- You've explicitly marked these for inclusion
- Will be included in next commit

Think of staging as creating a "shopping cart" of changes before "checking out" (committing).

## 4. Staging and Committing Changes

### Staging Files

**Stage individual file:**
1. Hover over file in CHANGES section
2. Click the `+` icon that appears
3. File moves to STAGED CHANGES

**Stage all files:**
- Hover over "CHANGES" header
- Click the `+` icon
- Or press `Ctrl+K Ctrl+Shift+S`

**Unstage a file:**
- In STAGED CHANGES, click the `-` icon
- File returns to CHANGES section

### Staging Partial Changes

Stage only specific lines from a file:
1. Open the diff view for the file
2. Select the lines you want to stage
3. Right-click â†’ "Stage Selected Ranges"
4. Or use the `+` icon next to line numbers

This is useful when you've made multiple unrelated changes in one file.

### Creating a Commit

**Steps:**
1. Stage the files you want to commit
2. In the message box at top, type commit message
3. Click the checkmark (âœ“) button or press `Ctrl+Enter`

**Commit message best practices:**
```
# Good commit messages:
"Add user authentication feature"
"Fix navigation bug in sidebar"
"Update README with installation steps"

# Bad commit messages:
"updates"
"fix"
"asdfasdf"
```

**Multi-line commit messages:**
- First line: Brief summary (50 chars or less)
- Blank line
- Detailed description

Example:
```
Add user authentication feature

- Implement login and signup forms
- Add JWT token validation
- Create user session management
```

### Commit All

**Commit without staging** (stages and commits in one step):
1. Enter commit message
2. Click the dropdown arrow next to checkmark
3. Select "Commit All"

Useful for simple changes, but staging first gives you more control.

## 5. Working with Branches

### What are Branches?

**Branches** let you work on different features or fixes independently without affecting the main code.

```
main:     A -- B -- C -- F -- G
                   \
feature:            D -- E
```

### Current Branch Indicator

Look at the **bottom-left** of status bar:
```
[â‡ main]  â† Shows current branch name
```

Click this to open branch menu.

### Creating a New Branch

**Method 1: Status Bar**
1. Click branch name in status bar
2. Select "Create new branch..."
3. Enter branch name (e.g., "feature/user-login")
4. Press Enter

**Method 2: Command Palette**
1. Press `Ctrl+Shift+P`
2. Type "Git: Create Branch"
3. Enter branch name

**Branch naming conventions:**
```
feature/user-authentication
bugfix/sidebar-navigation
hotfix/security-patch
experiment/new-ui
```

### Switching Branches

1. Click branch name in status bar
2. Select branch from dropdown list
3. VS Code switches immediately

**Shortcut**: `Ctrl+Shift+P` â†’ "Git: Checkout to..."

**Important**: Uncommitted changes will come with you when switching. Commit or stash them first!

### Merging Branches

Merge another branch into current branch:
1. Switch to the branch you want to merge INTO (e.g., main)
2. `Ctrl+Shift+P` â†’ "Git: Merge Branch"
3. Select the branch to merge FROM (e.g., feature/login)
4. VS Code merges the changes

### Deleting a Branch

After merging, delete the feature branch:
1. `Ctrl+Shift+P` â†’ "Git: Delete Branch"
2. Select branch to delete
3. Confirm deletion

**Note**: Can't delete the branch you're currently on.

## 7. Resolving Merge Conflicts

### What is a Merge Conflict?

A **conflict** occurs when:
- Two branches modify the same lines of code
- Git can't automatically determine which change to keep
- You must manually decide

### Identifying Conflicts

Files with conflicts are marked with `C` in Source Control:
```
MERGE CHANGES
 C  index.html  â† Conflict!
```

### Conflict Markers in Code

When you open a conflicted file:
```html
<<<<<<< HEAD (Current Change)
<h1>Welcome to My Site</h1>
=======
<h1>Welcome to Our Website</h1>
>>>>>>> feature-branch (Incoming Change)
```

- `<<<<<<< HEAD`: Your current branch's version
- `=======`: Separator
- `>>>>>>> branch-name`: The incoming branch's version

### Resolving with VS Code's Interface

VS Code adds clickable options above conflicts:

```
[Accept Current Change] [Accept Incoming Change] [Accept Both Changes] [Compare Changes]

<<<<<<< HEAD
<h1>Welcome to My Site</h1>
=======
<h1>Welcome to Our Website</h1>
>>>>>>> feature-branch
```

**Options:**
- **Accept Current Change**: Keep your version, discard incoming
- **Accept Incoming Change**: Use incoming version, discard yours
- **Accept Both Changes**: Keep both (one after another)
- **Compare Changes**: See side-by-side diff

### Manual Resolution

You can also manually edit:
1. Delete the conflict markers (`<<<<<<<`, `=======`, `>>>>>>>`)
2. Keep the code you want
3. Combine parts from both if needed

Example resolution:
```html
<!-- Combined the best of both: -->
<h1>Welcome to Our Site</h1>
```

### Completing the Merge

After resolving all conflicts:
1. Save the file(s)
2. Stage the resolved files (click `+`)
3. Commit with message like: "Merge feature-branch into main"
4. VS Code completes the merge

### Merge Editor (3-Way Merge)

For complex conflicts, use the merge editor:
1. When conflict appears, click "Resolve in Merge Editor"
2. See three panels:
   - Left: Current changes
   - Right: Incoming changes
   - Bottom: Result
3. Click checkboxes to accept changes
4. Edit result panel as needed
5. Click "Complete Merge"

## 10. Git Best Practices in VS Code

### Commit Frequency

**Commit often, but logically:**
- Make a commit when you complete a small, working piece of functionality
- Don't wait until end of day with 50 changes
- Don't commit every keystroke

**Good commit sizes:**
```
âœ“ "Add login form validation"
âœ“ "Fix navbar alignment bug"
âœ“ "Update API endpoint for users"
```

**Too large:**
```
âœ— "Add entire authentication system, refactor database, update all pages, fix 10 bugs"
```

### Review Before Committing

Always check your diffs before committing:
1. Click each changed file
2. Review the diff
3. Look for:
   - Debug code you forgot to remove
   - Commented-out code
   - Accidental changes
   - Console.log or print statements

### Use .gitignore

Create `.gitignore` to exclude files from Git:
```
# Python
*.pyc
__pycache__/
venv/
.env

# Node.js
node_modules/
package-lock.json

# IDE
.vscode/
.idea/

# OS
.DS_Store
Thumbs.db
```

### Branch Strategy

**Simple workflow for solo projects:**
- `main` branch: Stable, working code
- Feature branches: One per feature/fix
- Merge to main when complete

**Team workflow:**
- `main`: Production-ready code
- `develop`: Integration branch
- `feature/*`: New features
- `bugfix/*`: Bug fixes
- `hotfix/*`: Urgent fixes

### Sync Regularly

When working with others:
- **Pull before starting work** each day
- **Pull before pushing** your commits
- **Resolve conflicts early** before they grow
- **Communicate** when working on same files

### Keep Commits Focused

One commit should address one concern:
```
âœ“ Separate commits:
  - "Add user registration form"
  - "Fix typo in homepage header"
  - "Update dependencies"

âœ— Combined commit:
  - "Add registration, fix typo, update deps"
```

Makes it easier to:
- Understand history
- Revert specific changes
- Review code
