## 1. Understanding Settings Hierarchy

### Three Levels of Settings

VS Code has three layers of settings:

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚  1. Default Settings             â”‚  â† VS Code defaults
â”‚     (Built-in)                   â”‚    Lowest priority
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚  2. User Settings                â”‚  â† Your global preferences
â”‚     (Apply to all projects)      â”‚    Middle priority
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚  3. Workspace Settings           â”‚  â† Project-specific
â”‚     (Override user settings)     â”‚    Highest priority
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
```

**Priority**: Workspace > User > Default

### User Settings

**Location**: 
- Windows: `%APPDATA%\Code\User\settings.json`
- Mac: `~/Library/Application Support/Code/User/settings.json`
- Linux: `~/.config/Code/User/settings.json`

**When to use**:
- Personal preferences (font size, theme)
- Editor behavior you want everywhere
- Extension settings you always want

**Example:**
```json
{
  "editor.fontSize": 14,
  "editor.fontFamily": "Fira Code",
  "workbench.colorTheme": "One Dark Pro",
  "files.autoSave": "afterDelay"
}
```

### Workspace Settings

**Location**: `.vscode/settings.json` in project folder

**When to use**:
- Team conventions (tab size, formatting)
- Project-specific settings (Python interpreter)
- Different behavior per project

**Example:**
```json
{
  "editor.tabSize": 2,
  "python.linting.enabled": true,
  "python.formatting.provider": "black",
  "files.exclude": {
    "**/__pycache__": true,
    "**/*.pyc": true
  }
}
```

### Opening Settings

**GUI Method**:
- `Ctrl+,` (Comma) â†’ Opens Settings UI
- Search for settings
- Click to modify

**JSON Method**:
- `Ctrl+Shift+P` â†’ "Preferences: Open Settings (JSON)"
- Or click `{}` icon in Settings UI top-right

**Workspace Settings**:
- `Ctrl+Shift+P` â†’ "Preferences: Open Workspace Settings"
- Creates `.vscode/settings.json` if needed

## 2. Essential Editor Settings

### Font and Typography

```json
{
  // Font family
  "editor.fontFamily": "'Fira Code', 'Consolas', monospace",
  
  // Font size
  "editor.fontSize": 14,
  
  // Line height
  "editor.lineHeight": 1.5,
  
  // Letter spacing
  "editor.letterSpacing": 0.5,
  
  // Font ligatures (for Fira Code, JetBrains Mono, etc.)
  "editor.fontLigatures": true,
  
  // Font weight
  "editor.fontWeight": "400"
}
```

**Popular coding fonts**:
- Fira Code (free, ligatures)
- JetBrains Mono (free, ligatures)
- Cascadia Code (free, ligatures)
- Source Code Pro (free)
- Consolas (Windows built-in)

### Tab and Indentation

```json
{
  // Tab size
  "editor.tabSize": 4,
  
  // Use spaces instead of tabs
  "editor.insertSpaces": true,
  
  // Auto-detect indentation
  "editor.detectIndentation": true,
  
  // Show indentation guides
  "editor.renderIndentGuides": true,
  
  // Highlight active indent guide
  "editor.highlightActiveIndentGuide": true
}
```

### Line Numbers and Rulers

```json
{
  // Line numbers: 'on', 'off', 'relative'
  "editor.lineNumbers": "on",
  
  // Column rulers (vertical lines)
  "editor.rulers": [80, 120],
  
  // Render whitespace characters
  "editor.renderWhitespace": "selection",
  
  // Show line at end of file
  "files.insertFinalNewline": true,
  
  // Trim trailing whitespace on save
  "files.trimTrailingWhitespace": true
}
```

### Cursor and Selection

```json
{
  // Cursor style: 'line', 'block', 'underline'
  "editor.cursorStyle": "line",
  
  // Cursor blinking: 'blink', 'smooth', 'phase', 'expand', 'solid'
  "editor.cursorBlinking": "blink",
  
  // Cursor width
  "editor.cursorWidth": 2,
  
  // Smooth cursor animation
  "editor.cursorSmoothCaretAnimation": true,
  
  // Highlight current line
  "editor.renderLineHighlight": "all"
}
```

### Auto Save and Format

```json
{
  // Auto save: 'off', 'afterDelay', 'onFocusChange', 'onWindowChange'
  "files.autoSave": "afterDelay",
  
  // Auto save delay in milliseconds
  "files.autoSaveDelay": 1000,
  
  // Format on save
  "editor.formatOnSave": true,
  
  // Format on paste
  "editor.formatOnPaste": false,
  
  // Format on type
  "editor.formatOnType": false
}
```

## 3. Appearance Customization

### Color Themes

**Change theme**:
- `Ctrl+K Ctrl+T` â†’ Select color theme
- Or: `Ctrl+Shift+P` â†’ "Preferences: Color Theme"

**Built-in themes**:
- Dark+ (default dark)
- Light+ (default light)
- Monokai
- Solarized Dark/Light

**Popular theme extensions**:
- **One Dark Pro**: Atom's iconic theme
- **Dracula Official**: Popular dark theme
- **Night Owl**: Sarah Drasner's theme
- **Nord**: Arctic-inspired color palette
- **GitHub Theme**: Official GitHub themes
- **Monokai Pro**: Enhanced Monokai

**Install theme**:
1. `Ctrl+Shift+X` â†’ Search theme name
2. Install
3. `Ctrl+K Ctrl+T` â†’ Select installed theme

### File Icon Themes

**Change icon theme**:
- `Ctrl+Shift+P` â†’ "Preferences: File Icon Theme"

**Popular icon themes**:
- **Material Icon Theme**: Most popular
- **VSCode Icons**: Comprehensive icon set
- **Atom File Icons**: From Atom editor
- **Simple Icons**: Minimalist design

**Configure in settings**:
```json
{
  "workbench.iconTheme": "material-icon-theme",
  "material-icon-theme.folders.theme": "specific",
  "material-icon-theme.saturation": 1
}
```

### Workbench Appearance

```json
{
  // Activity bar position: 'left', 'right', 'hidden'
  "workbench.activityBar.location": "left",
  
  // Show breadcrumbs
  "breadcrumbs.enabled": true,
  
  // Sidebar position: 'left', 'right'
  "workbench.sideBar.location": "left",
  
  // Status bar visibility
  "workbench.statusBar.visible": true,
  
  // Tab settings
  "workbench.editor.showTabs": true,
  "workbench.editor.tabCloseButton": "right",
  "workbench.editor.tabSizing": "fit",
  
  // Minimap
  "editor.minimap.enabled": true,
  "editor.minimap.side": "right",
  "editor.minimap.showSlider": "mouseover",
  
  // Tree indentation
  "workbench.tree.indent": 12,
  
  // Compact folders (single child folders collapsed)
  "explorer.compactFolders": true
}
```

### Custom Color Customizations

Override specific colors:
```json
{
  "workbench.colorCustomizations": {
    // Editor colors
    "editor.background": "#1e1e1e",
    "editor.foreground": "#d4d4d4",
    "editor.lineHighlightBackground": "#2a2a2a",
    
    // Sidebar colors
    "sideBar.background": "#252526",
    "sideBar.foreground": "#cccccc",
    
    // Activity bar
    "activityBar.background": "#333333",
    "activityBar.foreground": "#ffffff",
    
    // Status bar
    "statusBar.background": "#007acc",
    "statusBar.foreground": "#ffffff"
  }
}
```

### Token Color Customizations

Customize syntax highlighting:
```json
{
  "editor.tokenColorCustomizations": {
    "comments": "#6A9955",
    "strings": "#ce9178",
    "numbers": "#b5cea8",
    "keywords": "#569cd6",
    "functions": "#dcdcaa"
  }
}
```

## 4. Language-Specific Settings

### Override Settings Per Language

Different settings for different languages:

```json
{
  // Global default
  "editor.tabSize": 4,
  
  // Python-specific
  "[python]": {
    "editor.tabSize": 4,
    "editor.insertSpaces": true,
    "editor.formatOnSave": true,
    "editor.defaultFormatter": "ms-python.black-formatter"
  },
  
  // JavaScript/TypeScript
  "[javascript]": {
    "editor.tabSize": 2,
    "editor.defaultFormatter": "esbenp.prettier-vscode"
  },
  "[typescript]": {
    "editor.tabSize": 2,
    "editor.defaultFormatter": "esbenp.prettier-vscode"
  },
  
  // HTML
  "[html]": {
    "editor.tabSize": 2,
    "editor.formatOnSave": true
  },
  
  // CSS
  "[css]": {
    "editor.tabSize": 2
  },
  
  // JSON
  "[json]": {
    "editor.tabSize": 2,
    "editor.formatOnSave": true
  },
  
  // Markdown
  "[markdown]": {
    "editor.wordWrap": "on",
    "editor.quickSuggestions": false
  }
}
```

### Python Configuration Example

```json
{
  "python.linting.enabled": true,
  "python.linting.pylintEnabled": true,
  "python.formatting.provider": "black",
  "python.linting.lintOnSave": true,
  "python.analysis.typeCheckingMode": "basic",
  "python.testing.pytestEnabled": true,
  
  "[python]": {
    "editor.rulers": [88],
    "editor.formatOnSave": true,
    "editor.codeActionsOnSave": {
      "source.organizeImports": true
    }
  }
}
```

### JavaScript/TypeScript Configuration

```json
{
  "javascript.updateImportsOnFileMove.enabled": "always",
  "typescript.updateImportsOnFileMove.enabled": "always",
  
  "[javascript]": {
    "editor.defaultFormatter": "esbenp.prettier-vscode",
    "editor.formatOnSave": true,
    "editor.codeActionsOnSave": {
      "source.fixAll.eslint": true
    }
  },
  
  "eslint.validate": [
    "javascript",
    "javascriptreact",
    "typescript",
    "typescriptreact"
  ]
}
```

## 5. Workspace Configuration

### Creating Workspace Settings

**Option 1: Manual creation**
1. Create `.vscode` folder in project root
2. Create `settings.json` inside it
3. Add project-specific settings

**Option 2: Through UI**
1. `Ctrl+Shift+P` â†’ "Preferences: Open Workspace Settings"
2. Modify settings (auto-creates `.vscode/settings.json`)

### Workspace Settings Example

**`.vscode/settings.json`** for a Python project:
```json
{
  // Python interpreter
  "python.defaultInterpreterPath": "${workspaceFolder}/venv/bin/python",
  
  // Linting and formatting
  "python.linting.enabled": true,
  "python.linting.pylintEnabled": true,
  "python.formatting.provider": "black",
  
  // Testing
  "python.testing.pytestEnabled": true,
  "python.testing.pytestArgs": [
    "tests"
  ],
  
  // Files to exclude
  "files.exclude": {
    "**/__pycache__": true,
    "**/*.pyc": true,
    "**/.pytest_cache": true,
    "**/venv": true
  },
  
  // Search exclusions
  "search.exclude": {
    "**/node_modules": true,
    "**/venv": true,
    "**/.venv": true
  },
  
  // Format on save
  "[python]": {
    "editor.formatOnSave": true,
    "editor.tabSize": 4
  }
}
```

### Recommended Extensions

**`.vscode/extensions.json`**:
```json
{
  "recommendations": [
    "ms-python.python",
    "ms-python.vscode-pylance",
    "ms-python.black-formatter",
    "ms-toolsai.jupyter",
    "eamodio.gitlens"
  ],
  "unwantedRecommendations": [
    "ms-vscode.csharp"
  ]
}
```

When teammates open the project, VS Code suggests installing recommended extensions.

### Multi-Root Workspaces

Work with multiple project folders simultaneously.

**Create multi-root workspace**:
1. `File` â†’ `Add Folder to Workspace`
2. Add multiple folders
3. `File` â†’ `Save Workspace As...`
4. Creates a `.code-workspace` file

**Example `.code-workspace` file**:
```json
{
  "folders": [
    {
      "name": "Frontend",
      "path": "./frontend"
    },
    {
      "name": "Backend",
      "path": "./backend"
    },
    {
      "name": "Shared",
      "path": "./shared"
    }
  ],
  "settings": {
    "editor.formatOnSave": true
  }
}
```

## 6. Performance Optimization

### File Watching

Reduce resource usage:
```json
{
  // Exclude files from file watcher
  "files.watcherExclude": {
    "**/.git/objects/**": true,
    "**/.git/subtree-cache/**": true,
    "**/node_modules/**": true,
    "**/.hg/store/**": true,
    "**/venv/**": true,
    "**/__pycache__/**": true
  },
  
  // Limit file watcher
  "files.watcherInclude": [
    "**/*.py",
    "**/*.js",
    "**/*.ts"
  ]
}
```

### Search Performance

```json
{
  // Exclude from search
  "search.exclude": {
    "**/node_modules": true,
    "**/bower_components": true,
    "**/*.code-search": true,
    "**/venv": true,
    "**/.venv": true,
    "**/__pycache__": true,
    "**/dist": true,
    "**/build": true
  },
  
  // Follow symbolic links
  "search.followSymlinks": false,
  
  // Use ripgrep for searching
  "search.useRipgrep": true
}
```

### Extension Performance

```json
{
  // Disable extension recommendations
  "extensions.ignoreRecommendations": false,
  
  // Auto update extensions
  "extensions.autoUpdate": true,
  
  // Auto check updates
  "extensions.autoCheckUpdates": true
}
```

**Check extension performance**:
- `Ctrl+Shift+P` â†’ "Developer: Show Running Extensions"
- See activation time and resource usage
- Disable slow extensions

### Editor Performance

```json
{
  // Disable minimap for large files
  "editor.minimap.enabled": false,
  
  // Limit suggestions
  "editor.suggest.maxVisibleSuggestions": 10,
  
  // Disable code lens
  "editor.codeLens": false,
  
  // Reduce bracket pair colorization
  "editor.bracketPairColorization.enabled": true,
  
  // Semantic highlighting
  "editor.semanticHighlighting.enabled": true,
  
  // Limit file size for syntax highlighting
  "editor.largeFileOptimizations": true
}
```

### Memory Settings

```json
{
  // Limit undo history
  "files.maxMemoryForLargeFilesMB": 4096,
  
  // Hot exit behavior
  "files.hotExit": "onExitAndWindowClose"
}
```

## 7. Settings Sync

### What is Settings Sync?

**Settings Sync** synchronizes your VS Code configuration across multiple machines using your Microsoft or GitHub account.

### What Gets Synced?

- Settings (settings.json)
- Keyboard shortcuts (keybindings.json)
- Extensions
- UI State (window layout, etc.)
- Snippets

### Enabling Settings Sync

**Method 1: Status Bar**
1. Click the gear icon in bottom-left
2. Select "Turn on Settings Sync"
3. Sign in with Microsoft or GitHub
4. Select what to sync

**Method 2: Command Palette**
1. `Ctrl+Shift+P` â†’ "Settings Sync: Turn On"
2. Follow prompts

### Sync Status

Look for sync icon in status bar:
```
ðŸ”„  â† Syncing
âœ“   â† Synced
âš    â† Sync error
```

### Configuring Sync

```json
{
  // Enable Settings Sync
  "settingsSync.keybindingsPerPlatform": true,
  
  // Ignored extensions (won't sync)
  "settingsSync.ignoredExtensions": [
    "local-extension-id"
  ],
  
  // Ignored settings
  "settingsSync.ignoredSettings": [
    "terminal.integrated.shell.windows"
  ]
}
```

### Manual Sync

Force sync:
- `Ctrl+Shift+P` â†’ "Settings Sync: Sync Now"

View sync activity:
- `Ctrl+Shift+P` â†’ "Settings Sync: Show Synced Data"

### Troubleshooting Sync

**If sync fails**:
1. Check internet connection
2. Sign out and sign in again
3. `Ctrl+Shift+P` â†’ "Settings Sync: Show Log"
4. Reset sync: "Settings Sync: Reset Local Data"

**Conflict resolution**:
- If conflicts occur, VS Code shows diff
- Choose local, remote, or merge
