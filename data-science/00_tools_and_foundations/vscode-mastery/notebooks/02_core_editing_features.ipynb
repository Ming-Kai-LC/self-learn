## Multi-Cursor Editing: Edit Multiple Places at Once

Multi-cursor editing is one of VS Code's most powerful features. It allows you to edit multiple locations simultaneously, dramatically speeding up repetitive editing tasks.

### Why Multi-Cursor Editing Matters

Imagine you need to rename a variable used 20 times in a file. Instead of find-replace (which might catch unwanted matches) or manually editing each occurrence, you can:
1. Select the variable
2. Press `Ctrl+D` repeatedly to select each occurrence
3. Type the new name once - all occurrences update simultaneously

### Creating Multiple Cursors

| Method | Shortcut | Description | Best For |
|--------|----------|-------------|----------|
| Box Selection | `Shift+Alt+Drag` | Column selection mode | Adding/removing same text on multiple lines |
| Add cursor above/below | `Ctrl+Alt+Up/Down` | Stack cursors vertically | Editing aligned code blocks |
| Add cursor at position | `Alt+Click` | Click anywhere to add cursor | Scattered edits |
| Select all occurrences | `Ctrl+Shift+L` | Select every match in file | Rename all instances |
| Add next occurrence | `Ctrl+D` | Select next match incrementally | Selective renaming |
| Undo last cursor | `Ctrl+U` | Remove last added cursor | Oops, too many! |

### Multi-Cursor Patterns

**Pattern 1: Column Editing**
```python
# Add 'self.' to multiple lines:
name = "Alice"      â†’  self.name = "Alice"
age = 30            â†’  self.age = 30
city = "NYC"        â†’  self.city = "NYC"
```
Method: `Shift+Alt+Drag` to select column before each variable

**Pattern 2: Rename Variable Selectively**
```python
# Rename some (not all) instances of 'data':
data = load_data()        â†’  user_data = load_data()
process(data)             â†’  process(user_data)
# but keep this one:
save_data(results)        â†’  save_data(results)  # unchanged
```
Method: Select first 'data', press `Ctrl+D` for each you want to rename

**Pattern 3: Stack Cursors**
```python
# Add commas to end of lines:
"apple"       â†’  "apple",
"banana"      â†’  "banana",
"cherry"      â†’  "cherry",
```
Method: `Ctrl+Alt+Down` to add cursors, then `End` to move to line ends, type `,`

## Advanced Text Selection

Beyond basic click-and-drag, VS Code offers sophisticated selection methods.

### Selection Shortcuts

| Action | Shortcut | Description |
|--------|----------|-------------|
| Select word | `Ctrl+D` | Select word under cursor |
| Select line | `Ctrl+L` | Select entire line |
| Expand selection | `Shift+Alt+Right` | Smart expand (word â†’ line â†’ block) |
| Shrink selection | `Shift+Alt+Left` | Smart shrink |
| Select to end of line | `Shift+End` | From cursor to line end |
| Select to start of line | `Shift+Home` | From cursor to line start |
| Select all | `Ctrl+A` | Select entire file |
| Select bracket content | `Ctrl+Shift+]` | Select all within brackets |

### Smart Selection (Expand/Shrink)

The expand/shrink selection feature understands code structure:

```python
def calculate_total(items):
    total = sum([item.price for item in items])
    return total
```

Place cursor in `item.price`:
1. `Shift+Alt+Right` â†’ selects `price`
2. `Shift+Alt+Right` â†’ selects `item.price`  
3. `Shift+Alt+Right` â†’ selects `item.price for item in items`
4. `Shift+Alt+Right` â†’ selects `[item.price for item in items]`
5. `Shift+Alt+Right` â†’ selects entire `sum(...)` expression
6. `Shift+Alt+Right` â†’ selects the line
7. `Shift+Alt+Right` â†’ selects the function

Then `Shift+Alt+Left` shrinks selection back down.

## Find and Replace

Powerful search and replace with regex support for complex patterns.

### Basic Find and Replace

**Find** (`Ctrl+F`):
- Opens find widget in current file
- Type search term
- `F3` / `Shift+F3` to navigate matches
- `Enter` to find next

**Replace** (`Ctrl+H`):
- Opens find and replace widget
- Enter find term and replace term
- `Ctrl+Shift+1`: Replace current match
- `Ctrl+Alt+Enter`: Replace all matches

### Find Options

Toggle these buttons in the find widget:
- `Alt+C`: Match case (Aa)
- `Alt+W`: Match whole word (Ab|)
- `Alt+R`: Use regular expression (.*)

### Regular Expression (Regex) Patterns

Enable regex with `Alt+R`, then use these patterns:

| Pattern | Matches | Example |
|---------|---------|----------|
| `\d` | Any digit (0-9) | `file\d` â†’ file1, file2 |
| `\d+` | One or more digits | `user\d+` â†’ user1, user123 |
| `\w` | Word character (a-z, A-Z, 0-9, _) | `\w+` â†’ any word |
| `\s` | Whitespace | `\s+` â†’ spaces/tabs |
| `.` | Any character | `a.c` â†’ abc, a1c, a*c |
| `.*` | Any characters (greedy) | `<.*>` â†’ entire tag |
| `.*?` | Any characters (non-greedy) | `<.*?>` â†’ one tag |
| `^` | Start of line | `^def` â†’ def at line start |
| `$` | End of line | `;$` â†’ semicolon at line end |
| `\|` | OR operator | `cat\|dog` â†’ cat or dog |
| `(...)` | Capture group | `(\w+)@` â†’ capture before @ |
| `[abc]` | Character class | `[aeiou]` â†’ any vowel |
| `[^abc]` | Negated class | `[^0-9]` â†’ not a digit |

### Capture Groups and Replacement

Use `()` to capture parts of the match, then `$1`, `$2`, etc. in replacement:

**Example 1: Swap names**
```
Find:    (\w+), (\w+)
Replace: $2 $1

"Smith, John" â†’ "John Smith"
"Doe, Jane" â†’ "Jane Doe"
```

**Example 2: Extract email domain**
```
Find:    \w+@(\w+\.\w+)
Replace: Domain: $1

"user@example.com" â†’ "Domain: example.com"
```

**Example 3: Add quotes around numbers**
```
Find:    (\d+)
Replace: "$1"

value = 42 â†’ value = "42"
```

## IntelliSense: Smart Code Completion

IntelliSense provides intelligent code completion based on language semantics, your code, and installed libraries.

### What IntelliSense Provides

1. **Auto-completion**: Suggests variables, functions, methods
2. **Parameter hints**: Shows function parameters as you type
3. **Quick info**: Displays documentation on hover
4. **Member lists**: Shows available methods/properties
5. **Import suggestions**: Auto-imports needed modules

### IntelliSense Triggers

| Shortcut | Action |
|----------|--------|
| `Ctrl+Space` | Trigger IntelliSense manually |
| `Ctrl+Shift+Space` | Trigger parameter hints |

| `Tab` or `Enter` | Accept suggestion |

## Code Folding: Collapse Code Blocks

Code folding lets you collapse sections of code to focus on what matters.

### What Can Be Folded?

- Functions and methods
- Classes
- If/else blocks
- For/while loops
- Try/except blocks
- Comment blocks
- Imports
- Custom regions (with markers)

### Folding Shortcuts

| Action | Shortcut | Description |
|--------|----------|-------------|
| Fold | `Ctrl+Shift+[` | Fold region at cursor |
| Unfold | `Ctrl+Shift+]` | Unfold region at cursor |
| Fold All | `Ctrl+K Ctrl+0` | Fold all regions |
| Unfold All | `Ctrl+K Ctrl+J` | Unfold all regions |
| Fold Level 1 | `Ctrl+K Ctrl+1` | Fold all level 1 (top-level functions/classes) |
| Fold Level 2 | `Ctrl+K Ctrl+2` | Fold all level 2 (nested blocks) |
| Toggle Fold | Click fold icon | Click the arrow in gutter |

### Folding Regions (Custom)

Create custom foldable regions:

**Python:**
```python
# region Setup
import os
import sys
# endregion
```

**JavaScript:**
```javascript
//#region Helpers
function helper1() { }
function helper2() { }
//#endregion
```

### When to Use Folding

1. **Hide imports**: Fold import statements to see code faster
2. **Focus on structure**: Fold all functions to see class outline
3. **Review code**: Fold completed sections, focus on what needs work
4. **Large files**: Collapse unrelated sections when working in one area

## Bracket Matching and Navigation

VS Code helps you work with brackets, parentheses, and braces.

### Bracket Pair Colorization

Enable built-in bracket colorization:
```json
{
    "editor.bracketPairColorization.enabled": true
}
```

Now matching brackets have the same color, making nesting clear:
```python
result = calculate([
    process({"key": value}),
    transform((x, y))
])
```

### Bracket Shortcuts

| Action | Shortcut | Description |
|--------|----------|-------------|
| Jump to bracket | `Ctrl+Shift+\` | Jump to matching bracket |
| Select to bracket | `Ctrl+Shift+]` | Select everything within brackets |

### Matching Bracket Highlighting

Place cursor next to a bracket:
- Matching bracket is highlighted
- Helpful in nested structures

### Auto-Closing Brackets

VS Code automatically closes brackets:
- Type `(` â†’ get `(|)` (cursor between)
- Type `[` â†’ get `[|]`
- Type `{` â†’ get `{|}`
- Type `"` â†’ get `"|"` (in strings)

Configure in settings:
```json
{
    "editor.autoClosingBrackets": "always",
    "editor.autoClosingQuotes": "always"
}
```

### Rename Symbol (F2)

The safest way to rename variables, functions, classes:

1. Place cursor on symbol
2. Press `F2`
3. Type new name
4. Press `Enter`
5. All references updated (even in other files!)

**Why use F2 instead of find/replace?**
- Understands scope (won't rename unrelated same-named variables)
- Updates all files in workspace
- Previews changes before applying
- Safer and smarter

### Extract Method

Convert code block into a reusable function:

```python
# Before:
total = 0
for item in items:
    total += item.price
tax = total * 0.08
final = total + tax

# Select the calculation lines, extract to method:
def calculate_total_with_tax(items):
    total = 0
    for item in items:
        total += item.price
    tax = total * 0.08
    return total + tax

final = calculate_total_with_tax(items)
```

### Code Actions Menu

Press `Ctrl+.` to see all available actions:
- Refactorings
- Quick fixes
- Source actions (organize imports, etc.)
- Generate code (getters/setters, etc.)
