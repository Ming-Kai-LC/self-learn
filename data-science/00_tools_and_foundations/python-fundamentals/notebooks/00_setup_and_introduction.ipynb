## 4. Comments: Notes in Your Code

Comments are notes for humans that Python ignores. They help explain what your code does.

### Single-Line Comments

Use the `#` symbol to start a comment:

````python
# This is a comment - Python will ignore this line
print("This line will run")  # You can also add comments after code
````

### Multi-Line Comments

For longer explanations, use triple quotes:

````python
"""
This is a multi-line comment.
You can write as much as you want here.
Python will ignore all of this.
"""

print("But this will run!")
````

**Best Practice**: Write comments to explain *why* you're doing something, not *what* you're doing. Good code should be readable without comments explaining every line.

## 5. Python Syntax Rules

Every programming language has rules. Here are Python's most important ones:

### 1. Indentation Matters!

Python uses indentation (spaces at the start of a line) to group code. This is different from many other languages.

````python
# This works - proper indentation (we'll learn more about this later)
x = 5
if x > 0:
    print("x is positive")  # This line is indented
````

### 2. Case Sensitivity

Python distinguishes between uppercase and lowercase letters:

````python
# These are three different things!
name = "Alice"
Name = "Bob"
NAME = "Charlie"

print(name)  # Alice
print(Name)  # Bob
print(NAME)  # Charlie
````

### 3. Strings Need Quotes

Text (called "strings") must be enclosed in quotes - single or double quotes work:

````python
print("This works")  # Double quotes
print("This works too")  # Single quotes
print("You can use 'quotes' inside quotes")  # Mix them
````

### 4. One Statement Per Line (Usually)

Each instruction typically goes on its own line:

````python
# Good practice - one statement per line
print("First line")
print("Second line")
````
