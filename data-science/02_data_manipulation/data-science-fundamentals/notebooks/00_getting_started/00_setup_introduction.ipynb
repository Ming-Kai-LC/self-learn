## 1. Welcome to Data Science!

### What is Data Science?

**Data Science** is the practice of extracting meaningful insights from data through:
- **Mathematics & Statistics**: Understanding patterns and making predictions
- **Programming**: Automating analysis and building tools
- **Domain Knowledge**: Asking the right questions

Think of a data scientist as a detective who:
1. **Collects evidence** (gathers data)
2. **Examines clues** (explores and cleans data)
3. **Tests theories** (statistical analysis)
4. **Solves the case** (derives insights and recommendations)

## 4. Your First Data Science Code

Let's write some simple data science code to get a taste of what's coming!

````python
# Import libraries (we'll learn these in detail later)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
````

````python
# Create some sample data
data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"],
    "Sales": [120, 135, 142, 158, 170],
}

# Convert to DataFrame (like an Excel spreadsheet)
df = pd.DataFrame(data)

print("Sales Data:")
print(df)
````

````python
# Calculate some statistics
print(f"Average daily sales: ${df['Sales'].mean():.2f}")
print(f"Best day: {df.loc[df['Sales'].idxmax(), 'Day']} (${df['Sales'].max()})")
print(f"Total weekly sales: ${df['Sales'].sum()}")
````

````python
# Create a visualization
plt.figure(figsize=(10, 6))
plt.plot(df["Day"], df["Sales"], marker="o", linewidth=2, markersize=8)
plt.title("Weekly Sales Trend", fontsize=16)
plt.xlabel("Day of Week", fontsize=12)
plt.ylabel("Sales ($)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nCongratulations! You just did data science! ðŸŽ‰")
````
