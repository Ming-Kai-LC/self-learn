### 2. Code Comments

#### Good Comments Explain WHY, Not WHAT

**Bad Comment** (states the obvious):
```python
# Increment i by 1
i += 1
```

**Good Comment** (explains reasoning):
```python
# Skip the first row because it contains column headers, not data
i += 1
```

#### When to Comment:
- âœ… **Complex algorithms**: Explain the approach
- âœ… **Non-obvious decisions**: Why this method over alternatives?
- âœ… **Workarounds**: Document known issues
- âœ… **TODOs**: Mark incomplete work
- âŒ **Obvious code**: Don't state what's clear from the code itself

````python
# Examples of good vs bad comments

print("=" * 80)
print("GOOD VS BAD COMMENTS")
print("=" * 80)

bad_examples = [
    ("âŒ Bad:", "# Calculate mean\nmean = data.mean()"),
    ("âŒ Bad:", "# Loop through items\nfor item in items:"),
    ("âŒ Bad:", "# Add 1 to counter\ncounter += 1"),
]

good_examples = [
    (
        "âœ… Good:",
        "# Use median instead of mean to reduce impact of outliers\nmedian = data.median()",
    ),
    (
        "âœ… Good:",
        "# Process items in order to maintain temporal relationships\nfor item in sorted(items, key=lambda x: x.timestamp):",
    ),
    (
        "âœ… Good:",
        "# Track successful API calls for rate limiting (max 100/hour)\nsuccess_count += 1",
    ),
]

print("\nğŸ“› BAD COMMENTS (State the obvious):")
print("-" * 80)
for label, example in bad_examples:
    print(f"\n{label}")
    print(example)

print("\n\nâœ¨ GOOD COMMENTS (Explain reasoning):")
print("-" * 80)
for label, example in good_examples:
    print(f"\n{label}")
    print(example)

print("\n" + "=" * 80)
print("ğŸ’¡ Rule: If someone asks 'Why?', you need a comment. If they ask 'What?', improve the code.")
````

### 3. Docstrings

Every function, class, and module should have a docstring!

#### Docstring Formats:
- **NumPy Style** (recommended for data science)
- Google Style
- reStructuredText (Sphinx)

#### NumPy Style Example:

````python
def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for a dataset.

    Uses the t-distribution for small samples (n < 30) and
    normal distribution for larger samples.

    Parameters
    ----------
    data : array-like
        Sample data for which to calculate the confidence interval.
        Must contain at least 2 values.
    confidence : float, default=0.95
        Confidence level (between 0 and 1).
        Common values: 0.90, 0.95, 0.99

    Returns
    -------
    tuple of float
        (lower_bound, upper_bound) of the confidence interval

    Raises
    ------
    ValueError
        If data has less than 2 values or confidence is not in (0, 1)

    Examples
    --------
    >>> data = [1, 2, 3, 4, 5]
    >>> lower, upper = calculate_confidence_interval(data)
    >>> print(f"95% CI: [{lower:.2f}, {upper:.2f}]")
    95% CI: [1.47, 4.53]

    Notes
    -----
    This function assumes the data is normally distributed. For
    non-normal data, consider bootstrap methods instead.

    References
    ----------
    .. [1] Student (1908). "The probable error of a mean".
           Biometrika 6(1): 1-25.

    See Also
    --------
    scipy.stats.t.interval : More flexible confidence interval calculation
    """
    from scipy import stats

    # Validate inputs
    if len(data) < 2:
        raise ValueError("Data must contain at least 2 values")
    if not 0 < confidence < 1:
        raise ValueError("Confidence must be between 0 and 1")

    # Calculate statistics
    mean = np.mean(data)
    se = stats.sem(data)

    # Use t-distribution
    interval = stats.t.interval(confidence, len(data) - 1, loc=mean, scale=se)

    return interval


# Example usage
np.random.seed(42)
sample_data = np.random.normal(100, 15, 50)
lower, upper = calculate_confidence_interval(sample_data)

print("âœ… WELL-DOCUMENTED FUNCTION")
print("=" * 80)
print(f"Sample mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
print("\nDocstring components:")
print("  âœ“ Brief description")
print("  âœ“ Parameters with types and defaults")
print("  âœ“ Return value description")
print("  âœ“ Possible errors")
print("  âœ“ Usage examples")
print("  âœ“ Important notes")
print("  âœ“ References to papers")
print("  âœ“ Related functions")
````

## Part 2: Version Control with Git

### What is Version Control?

Version control tracks changes to your code over time, allowing you to:
- ğŸ“œ **Track history**: See what changed, when, and why
- â®ï¸ **Revert changes**: Undo mistakes
- ğŸŒ¿ **Branch**: Experiment without breaking working code
- ğŸ¤ **Collaborate**: Work with others without conflicts

### Git Basics

#### The Git Workflow:

```
Working Directory  â†’  Staging Area  â†’  Repository
    (edit)            (git add)        (git commit)
      â†“                   â†“                 â†“
   file.py          file.py ready     file.py saved
                    to commit         in history
```

#### Essential Git Commands:

```bash
# Initialize a repository
git init

# Check status
git status

# Stage changes
git add filename.py        # Stage specific file
git add .                  # Stage all changes

# Commit changes
git commit -m "Your message"

# View history
git log
git log --oneline          # Compact view

# View changes
git diff                   # Unstaged changes
git diff --staged          # Staged changes

# Branching
git branch feature-name    # Create branch
git checkout feature-name  # Switch to branch
git checkout -b feature    # Create and switch

# Merge branches
git checkout main
git merge feature-name

# Remote repositories
git clone https://github.com/user/repo.git
git push origin main       # Push to remote
git pull origin main       # Pull from remote
```

````python
# Visualize a Git workflow

workflow_steps = [
    "1. Edit Code",
    "2. Stage Changes\n(git add)",
    "3. Commit\n(git commit)",
    "4. Push\n(git push)",
]

colors = ["lightcoral", "lightsalmon", "lightgreen", "lightblue"]

fig, ax = plt.subplots(figsize=(12, 3))

for i, (step, color) in enumerate(zip(workflow_steps, colors)):
    # Draw box
    ax.add_patch(plt.Rectangle((i * 2.5, 0), 2, 1, facecolor=color, edgecolor="black", linewidth=2))
    # Add text
    ax.text(i * 2.5 + 1, 0.5, step, ha="center", va="center", fontsize=11, fontweight="bold")

    # Draw arrow (except for last step)
    if i < len(workflow_steps) - 1:
        ax.annotate(
            "",
            xy=(i * 2.5 + 2.3, 0.5),
            xytext=(i * 2.5 + 2.0, 0.5),
            arrowprops=dict(arrowstyle="->", lw=2, color="black"),
        )

ax.set_xlim(-0.5, 10)
ax.set_ylim(-0.2, 1.5)
ax.axis("off")
ax.set_title("Git Workflow for Research Projects", fontsize=14, fontweight="bold", pad=20)

plt.tight_layout()
plt.savefig(os.path.join(output_dir, "git_workflow.png"), dpi=300, bbox_inches="tight")
plt.show()

print("\nğŸ’¡ Remember: Commit early, commit often!")
````

### Writing Good Commit Messages

Commit messages are documentation of **why** you made changes.

#### The Anatomy of a Good Commit Message:

```
Short summary (50 chars or less)

More detailed explanation if needed. Wrap at 72 characters.
Explain:
- WHY you made this change
- WHAT problem it solves
- Any side effects or limitations

Fixes #123
```

#### Examples:

**âŒ Bad:**
```
update
fix bug
changes
asdf
```

**âœ… Good:**
```
Add missing value imputation to preprocessing

Fix data quality issue where 15% of records had missing age values.
Uses median imputation for age as it's less sensitive to outliers
than mean imputation.

Closes #45
```

#### Conventional Commits:

A standardized format:

```
type(scope): subject

body
```

**Types:**
- `feat`: New feature
- `fix`: Bug fix
- `docs`: Documentation changes
- `style`: Code formatting (no logic changes)
- `refactor`: Code restructuring
- `test`: Adding tests
- `chore`: Maintenance tasks

**Examples:**
```
feat(preprocessing): add log transformation for skewed features
fix(model): correct feature scaling bug in pipeline
docs(readme): add installation instructions for Windows
```

````python
# Commit message quality checker


def evaluate_commit_message(message):
    """
    Evaluate the quality of a commit message.

    Parameters:
    -----------
    message : str
        The commit message to evaluate

    Returns:
    --------
    dict
        Quality score and feedback
    """
    score = 0
    feedback = []

    lines = message.strip().split("\n")
    subject = lines[0] if lines else ""

    # Check subject line
    if len(subject) > 0:
        score += 2
        feedback.append("âœ… Has subject line")
    else:
        feedback.append("âŒ Missing subject line")

    if 10 <= len(subject) <= 50:
        score += 2
        feedback.append("âœ… Subject length appropriate (10-50 chars)")
    elif len(subject) > 50:
        feedback.append("âš ï¸  Subject too long (>50 chars)")
    elif len(subject) > 0:
        feedback.append("âš ï¸  Subject too short (<10 chars)")

    # Check for vague words
    vague_words = ["update", "change", "fix", "stuff", "things", "various"]
    if not any(word in subject.lower() for word in vague_words):
        score += 2
        feedback.append("âœ… Subject is specific (no vague words)")
    else:
        feedback.append("âš ï¸  Subject contains vague words")

    # Check for imperative mood
    imperative_verbs = ["add", "fix", "remove", "update", "implement", "refactor", "improve"]
    if any(subject.lower().startswith(verb) for verb in imperative_verbs):
        score += 2
        feedback.append("âœ… Uses imperative mood")
    else:
        feedback.append("âš ï¸  Consider using imperative mood (e.g., 'Add', 'Fix')")

    # Check for body
    if len(lines) > 2 and lines[1].strip() == "":
        score += 2
        feedback.append("âœ… Has blank line before body")
        if len("\n".join(lines[2:])) > 20:
            score += 2
            feedback.append("âœ… Has detailed body")

    # Overall rating
    max_score = 10
    percentage = (score / max_score) * 100

    if percentage >= 80:
        rating = "Excellent"
    elif percentage >= 60:
        rating = "Good"
    elif percentage >= 40:
        rating = "Needs Improvement"
    else:
        rating = "Poor"

    return {
        "score": score,
        "max_score": max_score,
        "percentage": percentage,
        "rating": rating,
        "feedback": feedback,
    }


# Test examples
test_messages = [
    ("Bad", "update"),
    ("Better", "Fix preprocessing bug"),
    (
        "Good",
        """Add feature importance visualization

Implement SHAP values for model interpretability. This helps identify
which features most influence predictions, addressing reviewer feedback
from issue #12.""",
    ),
]

print("ğŸ“ COMMIT MESSAGE QUALITY EVALUATION")
print("=" * 80)

for label, message in test_messages:
    print(f"\n{label.upper()} EXAMPLE:")
    print("-" * 80)
    print(f"Message: {message[:50]}..." if len(message) > 50 else f"Message: {message}")

    result = evaluate_commit_message(message)
    print(
        f"\nScore: {result['score']}/{result['max_score']} ({result['percentage']:.0f}%) - {result['rating']}"
    )
    print("\nFeedback:")
    for item in result["feedback"]:
        print(f"  {item}")
````

### What to Version Control?

#### âœ… DO Version Control:
- Source code (.py, .R, .jl)
- Documentation (.md, .rst)
- Configuration files
- Requirements files
- Small data files (<10 MB)
- Analysis notebooks (.ipynb)
- Test files

#### âŒ DON'T Version Control:
- Large data files (>10 MB) - use DVC or LFS instead
- Generated files (outputs, figures)
- Compiled code (.pyc, .o)
- Virtual environment folders
- Secrets (.env, credentials.json)
- Temporary files
- OS-specific files (.DS_Store, Thumbs.db)

### The .gitignore File

A `.gitignore` file tells Git which files to ignore.

````python
# Generate a .gitignore template for data science


def generate_gitignore_template():
    """
    Generate a comprehensive .gitignore for data science projects.
    """
    template = """# Data Science .gitignore Template

# Python
__pycache__/
*.py[cod]
*$py.class
*.so
.Python

# Virtual Environments
venv/
env/
ENV/
.venv/
conda-env/

# Jupyter Notebooks
.ipynb_checkpoints/
*-checkpoint.ipynb

# Data Files (large files)
*.csv
*.xlsx
*.h5
*.hdf5
*.parquet
*.pkl
*.pickle
data/raw/
data/processed/

# Keep .gitkeep files
!.gitkeep

# Model Files
*.model
*.h5
*.pt
*.pth
*.ckpt
models/

# Outputs
outputs/
results/
*.png
*.jpg
*.pdf
figures/

# Logs
*.log
logs/

# Secrets
.env
credentials.json
*.key
*.pem

# IDE
.vscode/
.idea/
*.swp
*.swo
*~

# OS
.DS_Store
Thumbs.db
desktop.ini

# Temporary
*.tmp
*.bak
*.cache
"""
    return template


gitignore = generate_gitignore_template()
print("ğŸ“„ .GITIGNORE TEMPLATE")
print("=" * 80)
print(gitignore)

# Save template
with open(os.path.join(output_dir, "gitignore_template.txt"), "w") as f:
    f.write(gitignore)

print(f"\nâœ… .gitignore template saved to {output_dir}/gitignore_template.txt")
print("\nğŸ’¡ Copy this to your project root as '.gitignore'")
````

## Part 3: Jupyter Notebooks and Version Control

### The Challenge

Jupyter notebooks are stored as **JSON files** containing:
- Code cells
- Output cells (results, plots)
- Metadata (execution counts, timestamps)

This makes them **difficult to version control**:
- ğŸ˜ **Large diffs**: Outputs create huge changes
- ğŸ˜ **Merge conflicts**: Hard to resolve
- ğŸ˜ **Unclear changes**: JSON format is hard to read in diffs

### Solutions

#### 1. Clear Outputs Before Committing

In Jupyter:
- Cell â†’ All Output â†’ Clear

From command line:
```bash
jupyter nbconvert --clear-output --inplace notebook.ipynb
```

#### 2. Use nbstripout (Automatic)

Automatically strip outputs on commit:
```bash
pip install nbstripout
nbstripout --install
```

#### 3. Use nbdime for Better Diffs

```bash
pip install nbdime
nbdime config-git --enable --global
```

View notebook diff:
```bash
nbdiff notebook_old.ipynb notebook_new.ipynb
nbdiff-web  # Interactive web interface
```

#### 4. Convert to .py Files (Pairing)

Version control both .ipynb and .py:
```bash
jupyter nbconvert --to script notebook.ipynb
```

Or use Jupytext:
```bash
pip install jupytext
jupytext --set-formats ipynb,py notebook.ipynb
```

### Best Practices for Notebooks

1. âœ… **Keep notebooks focused**: One notebook = one analysis
2. âœ… **Clear outputs before commit**: Reduce diff size
3. âœ… **Restart & run all**: Ensure reproducibility
4. âœ… **Move reusable code to .py files**: Keep notebooks clean
5. âœ… **Add markdown explanations**: Document your thinking
6. âœ… **Name cells**: Use meaningful headers
7. âœ… **Version your data**: Track which data version you used

````python
# Notebook version control checklist


def create_notebook_git_checklist():
    """
    Create a checklist for notebook version control best practices.
    """
    checklist = {
        "ğŸ““ Before Committing": [
            "â˜ Restart kernel and run all cells",
            "â˜ Verify all cells execute without errors",
            "â˜ Clear all outputs (or use nbstripout)",
            "â˜ Check for sensitive data in outputs",
            "â˜ Remove debug/scratch cells",
        ],
        "ğŸ“ Notebook Quality": [
            "â˜ Add markdown headers for sections",
            "â˜ Explain WHY, not just WHAT",
            "â˜ Keep cells short and focused",
            "â˜ Move reusable code to .py modules",
            "â˜ Document data sources and versions",
        ],
        "ğŸ”§ Git Setup": [
            "â˜ Install nbstripout or nbdime",
            "â˜ Configure .gitignore for notebooks",
            "â˜ Consider pairing with .py files (Jupytext)",
        ],
        "ğŸ¤ Collaboration": [
            "â˜ Document notebook dependencies",
            "â˜ Specify data location and version",
            "â˜ Add README with execution instructions",
            "â˜ Use consistent naming conventions",
        ],
    }
    return checklist


checklist = create_notebook_git_checklist()

print("ğŸ““ NOTEBOOK VERSION CONTROL CHECKLIST")
print("=" * 80)
for category, items in checklist.items():
    print(f"\n{category}")
    for item in items:
        print(f"  {item}")

# Save as markdown
checklist_md = "# Notebook Version Control Checklist\n\n"
for category, items in checklist.items():
    checklist_md += f"## {category}\n\n"
    for item in items:
        checklist_md += f"- {item}\n"
    checklist_md += "\n"

with open(os.path.join(output_dir, "notebook_git_checklist.md"), "w") as f:
    f.write(checklist_md)

print(f"\nâœ… Checklist saved to {output_dir}/notebook_git_checklist.md")
````
