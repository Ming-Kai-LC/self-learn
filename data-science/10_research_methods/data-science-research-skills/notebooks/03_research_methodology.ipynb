## Part 1: Formulating Research Questions

The research question is the **foundation** of your entire project. A well-formulated question guides everything that follows.

### Characteristics of Good Research Questions

A good research question should be:

#### 1. **Clear and Specific**
- Everyone should understand exactly what you're asking
- No ambiguous terms
- Well-defined scope

**Bad**: "How does technology affect people?"
- What technology? Which people? Affect how?

**Good**: "How does smartphone screen time (>4 hours/day) affect sleep quality in college students aged 18-24?"
- Specific technology, population, and outcome

#### 2. **Focused (Not Too Broad)**
- Can be answered within your time and resource constraints
- One main question, not multiple questions

**Too Broad**: "What is the best machine learning algorithm?"
- Best for what? Under what conditions?

**Focused**: "Which algorithm (Random Forest, XGBoost, or Neural Network) performs best for predicting customer churn in e-commerce with imbalanced datasets?"
- Specific algorithms, specific task, specific context

#### 3. **Researchable**
- Can actually be investigated with available methods
- Data can be collected or accessed
- Ethical to study

**Not Researchable**: "What do aliens think about human AI?"
- Can't collect data

**Researchable**: "What do AI researchers think about the risks of advanced AI?"
- Can survey or interview researchers

#### 4. **Original and Meaningful**
- Adds something new to the field
- Matters to someone (academics, practitioners, society)
- Not already definitively answered

**Not Original**: "Does increasing sample size improve model performance?"
- Well-established fact

**Original**: "At what sample size do deep learning models for rare disease diagnosis reach acceptable performance?"
- Specific, practical question with implications

#### 5. **Complex (Not Yes/No)**
- Requires analysis, not just a simple answer
- Often starts with "How", "Why", "What factors", "To what extent"

**Too Simple**: "Do people use social media?"
- Yes. End of research.

**Complex**: "What factors influence social media usage patterns among Gen Z users?"
- Requires investigation and analysis

### The FINER Criteria

A framework for evaluating research questions:

- **F**easible: Can you actually do it?
- **I**nteresting: Do you and others care?
- **N**ovel: Is it new?
- **E**thical: Is it right to study?
- **R**elevant: Does it matter?

### Examples: Bad vs Good Research Questions

| Bad | Why Bad | Good | Why Good |
|-----|---------|------|----------|
| "Is AI good?" | Vague, subjective, too broad | "Does using AI-assisted diagnosis improve detection rates of diabetic retinopathy in primary care settings?" | Specific, measurable, focused |
| "Everything about ML" | Not a question, impossibly broad | "Which hyperparameter tuning strategies are most effective for CNNs on small medical imaging datasets?" | Specific task, specific context |
| "How to cure cancer?" | Too broad, not specific research | "Does combining feature X with standard markers improve early detection of lung cancer in high-risk populations?" | Specific intervention, specific outcome |
| "What's the future of data science?" | Speculative, not researchable | "What skills are most demanded in data science job postings in 2024 compared to 2020?" | Specific, data-collectible |
| "Are neural networks better?" | Better than what? For what? | "Do transformer models outperform LSTMs for time-series forecasting of stock prices with 1-day horizons?" | Specific comparison, specific task, specific timeframe |

````python
# ========================================
# Create Research Question Evaluation Tool
# ========================================


def evaluate_research_question(question):
    """
    Interactive tool to evaluate your research question using FINER criteria.
    """
    print("RESEARCH QUESTION EVALUATION TOOL")
    print("=" * 60)
    print(f"\nYour question: {question}\n")
    print("Rate each criterion from 1-5 (1=Poor, 5=Excellent):\n")

    criteria = {
        "Feasible": "Can you realistically complete this research with available resources?",
        "Interesting": "Is this question interesting to you and your field?",
        "Novel": "Does this add something new (not already well-established)?",
        "Ethical": "Is this research ethical to conduct?",
        "Relevant": "Will the answer matter to someone (policy, practice, theory)?",
    }

    scores = {}
    for criterion, description in criteria.items():
        print(f"{criterion}: {description}")
        # In practice, user would input score
        # For demonstration:
        score = np.random.randint(3, 6)  # Random 3-5
        scores[criterion] = score
        print(f"  Score: {score}/5\n")

    total = sum(scores.values())
    avg = total / len(scores)

    print("\nRESULTS:")
    print("=" * 60)
    print(f"Total Score: {total}/25")
    print(f"Average: {avg:.1f}/5")
    print()

    if avg >= 4.5:
        print("âœ… EXCELLENT - This is a strong research question!")
    elif avg >= 3.5:
        print("ðŸ‘ GOOD - With minor refinements, this could be excellent")
    elif avg >= 2.5:
        print("âš ï¸ NEEDS WORK - Consider revising several aspects")
    else:
        print("âŒ POOR - Significant revision needed")

    return scores


# Example
sample_question = "Does feature engineering improve customer churn prediction accuracy?"
scores = evaluate_research_question(sample_question)

# Visualize
fig, ax = plt.subplots(figsize=(10, 6))
criteria_names = list(scores.keys())
scores_values = list(scores.values())

bars = ax.barh(criteria_names, scores_values, color="steelblue", alpha=0.7)
ax.set_xlabel("Score (1-5)", fontweight="bold")
ax.set_title("FINER Criteria Evaluation", fontweight="bold", fontsize=13)
ax.set_xlim([0, 5])
ax.axvline(x=3, color="orange", linestyle="--", alpha=0.5, label="Threshold")

# Add score labels
for bar in bars:
    width = bar.get_width()
    ax.text(
        width + 0.1,
        bar.get_y() + bar.get_height() / 2,
        f"{width}/5",
        ha="left",
        va="center",
        fontweight="bold",
    )

plt.tight_layout()
plt.savefig(f"{output_dir}/finer_evaluation.png", dpi=150, bbox_inches="tight")
print(f"\nâœ… Evaluation chart saved to: {output_dir}/finer_evaluation.png")
plt.show()
````

## Part 3: Developing Hypotheses

Once you have a research question, you need a **hypothesis** - a testable prediction about the answer.

### What is a Hypothesis?

A **hypothesis** is a specific, testable prediction about the relationship between variables.

### Components of a Good Hypothesis

#### 1. Variables
- **Independent Variable (IV)**: What you manipulate or compare
- **Dependent Variable (DV)**: What you measure

**Example**: "Adding RFM features (IV) will improve churn prediction accuracy (DV)."

#### 2. Direction
- Specify the expected direction of the relationship
- Increase? Decrease? Difference?

**Vague**: "Feature engineering will affect model performance."
**Directional**: "Feature engineering will improve model performance by at least 10%."

#### 3. Testability
- Can be proven true or false with data
- Specific enough to measure

**Not Testable**: "Deep learning is the future."
**Testable**: "Deep learning models will achieve >90% accuracy on ImageNet by 2025."

### Hypothesis Structure

**Basic Format**:
"If [change in IV], then [change in DV] because [theoretical reason]."

**Example 1**: Stock Prediction
"If we use LSTM networks instead of ARIMA models, then prediction accuracy will improve by at least 10% because LSTMs can capture non-linear temporal patterns that ARIMA cannot."

**Example 2**: Customer Churn
"If we include behavioral features (RFM) in addition to demographic features, then churn prediction F1-score will increase by at least 15% because behavioral data better reflects customer engagement."

**Example 3**: Image Classification
"If we apply data augmentation to training images, then model accuracy on small datasets (< 1000 images) will improve by 5-10% because augmentation increases effective training data diversity."

### Null vs Alternative Hypothesis

In statistical testing, you always have two hypotheses:

**Null Hypothesis (Hâ‚€)**:
- States there is NO effect or NO difference
- What you try to disprove
- Example: "There is no difference in accuracy between LSTM and ARIMA."

**Alternative Hypothesis (Hâ‚ or Hâ‚)**:
- States there IS an effect or difference
- What you're trying to support
- Example: "LSTM achieves higher accuracy than ARIMA."

**Statistical testing attempts to reject the null hypothesis**

## Part 4: Types of Research Approaches

Different research questions require different approaches.

### Quantitative Research

**What**: Numbers, statistics, objective measurements

**When to use**:
- Testing hypotheses
- Measuring relationships between variables
- Generalizing to populations

**Characteristics**:
- Large sample sizes
- Statistical analysis
- Objective measurements
- Structured data collection

**Examples**:
- "Does algorithm X outperform algorithm Y?"
- "What percentage of users churn within 3 months?"
- "Is there a correlation between features A and B?"

**Common Methods**:
- Experiments
- Surveys (with closed questions)
- Statistical analysis of existing data

### Qualitative Research

**What**: Words, themes, subjective understanding

**When to use**:
- Exploring new phenomena
- Understanding experiences
- Generating hypotheses

**Characteristics**:
- Small, purposive samples
- In-depth analysis
- Subjective interpretation
- Flexible data collection

**Examples**:
- "How do data scientists decide which algorithm to use?"
- "What challenges do users face with AI interfaces?"
- "Why do customers abandon their carts?"

**Common Methods**:
- Interviews
- Focus groups
- Observations
- Content analysis

### Mixed Methods Research

**What**: Combines quantitative and qualitative

**When to use**:
- Need both breadth (quant) and depth (qual)
- Want to explain quantitative findings
- Complex research questions

**Example Study**:
1. **Quantitative**: Survey 1000 users about satisfaction (numbers)
2. **Qualitative**: Interview 20 users to understand WHY they're satisfied/dissatisfied (insights)
3. **Integration**: Use interviews to explain survey findings

### Comparison Table

| Aspect | Quantitative | Qualitative | Mixed Methods |
|--------|-------------|-------------|---------------|
| **Goal** | Test hypotheses | Explore phenomena | Comprehensive understanding |
| **Data** | Numbers | Words, images | Both |
| **Sample** | Large, random | Small, purposive | Both |
| **Analysis** | Statistical | Thematic | Both |
| **Strength** | Generalizability | Depth | Completeness |
| **Weakness** | May miss nuance | Hard to generalize | Time-consuming |

## Part 5: Common Pitfalls and How to Avoid Them

### Pitfall 1: Question Too Broad

**Problem**: "How can we improve machine learning?"

**Solution**: Be specific
- What aspect of ML?
- For what application?
- Compared to what?

**Better**: "How can ensembling improve accuracy for small tabular datasets in healthcare?"

### Pitfall 2: Question Too Narrow

**Problem**: "Does changing learning rate from 0.001 to 0.002 in a 3-layer network trained on exactly 1000 samples of iris dataset improve accuracy?"

**Solution**: Ensure some generalizability

**Better**: "How does learning rate affect convergence speed in neural networks for small classification datasets?"

### Pitfall 3: Confirming What's Already Known

**Problem**: "Does more data improve model performance?"
- Already well-established

**Solution**: Add novel angle

**Better**: "At what data size do transformers outperform CNNs for medical image classification?"

### Pitfall 4: Unfocused Hypothesis

**Problem**: "Neural networks work better."
- Better than what?
- For what task?
- How much better?

**Solution**: Specify everything

**Better**: "ResNet-50 achieves 5-10% higher accuracy than VGG-16 for skin lesion classification on datasets with < 5000 images."

### Pitfall 5: Non-Testable Hypothesis

**Problem**: "AI will revolutionize healthcare."
- Too vague
- Can't measure "revolutionize"

**Solution**: Make it measurable

**Better**: "AI-assisted diagnosis systems will reduce diagnostic errors by 15% in radiology within 5 years."

### Pitfall 6: Ignoring Feasibility

**Problem**: "I'll analyze all tweets ever posted to understand sentiment."
- Billions of tweets
- Computational limits
- Time constraints

**Solution**: Be realistic

**Better**: "I'll analyze 100,000 tweets from Q1 2024 about topic X to understand sentiment patterns."
