## What is MCP?

The **Model Context Protocol (MCP)** is an open standard created by Anthropic that enables AI assistants like Claude to securely connect to external data sources and tools.

Think of MCP as a **universal adapter** that lets Claude:
- Access your GitHub repositories
- Query databases directly
- Read and write to cloud storage
- Interact with business tools (Slack, Jira, etc.)
- Connect to any custom data source you build

## Creating Custom MCP Servers

You can build custom MCP servers to integrate any system or API!

### MCP Server Anatomy

An MCP server provides:
1. **Resources:** Data Claude can read
2. **Tools:** Functions Claude can call
3. **Prompts:** Reusable prompt templates (optional)

### Building a Simple MCP Server

#### Example: Weather API Server

**server.js**
```javascript
#!/usr/bin/env node
import { Server } from '@modelcontextprotocol/sdk/server/index.js';
import { StdioServerTransport } from '@modelcontextprotocol/sdk/server/stdio.js';
import { CallToolRequestSchema, ListToolsRequestSchema } from '@modelcontextprotocol/sdk/types.js';

// Create server instance
const server = new Server(
  {
    name: 'weather-server',
    version: '1.0.0',
  },
  {
    capabilities: {
      tools: {},
    },
  }
);

// Define available tools
server.setRequestHandler(ListToolsRequestSchema, async () => {
  return {
    tools: [
      {
        name: 'get_weather',
        description: 'Get current weather for a city',
        inputSchema: {
          type: 'object',
          properties: {
            city: {
              type: 'string',
              description: 'City name',
            },
          },
          required: ['city'],
        },
      },
    ],
  };
});

// Handle tool calls
server.setRequestHandler(CallToolRequestSchema, async (request) => {
  if (request.params.name === 'get_weather') {
    const { city } = request.params.arguments;
    
    // Call weather API
    const response = await fetch(
      `https://api.openweathermap.org/data/2.5/weather?q=${city}&appid=${process.env.WEATHER_API_KEY}`
    );
    const data = await response.json();
    
    return {
      content: [
        {
          type: 'text',
          text: JSON.stringify(data, null, 2),
        },
      ],
    };
  }
  
  throw new Error('Unknown tool');
});

// Start server
async function main() {
  const transport = new StdioServerTransport();
  await server.connect(transport);
}

main().catch(console.error);
```

**package.json**
```json
{
  "name": "mcp-weather-server",
  "version": "1.0.0",
  "type": "module",
  "bin": {
    "mcp-weather-server": "./server.js"
  },
  "dependencies": {
    "@modelcontextprotocol/sdk": "^0.5.0"
  }
}
```

**Configuration**
```json
{
  "mcpServers": {
    "weather": {
      "command": "node",
      "args": ["/path/to/weather-server/server.js"],
      "env": {
        "WEATHER_API_KEY": "your-api-key"
      }
    }
  }
}
```

## Advanced Custom Server: REST API Wrapper

### Example: Stripe Payment Server

```javascript
#!/usr/bin/env node
import { Server } from '@modelcontextprotocol/sdk/server/index.js';
import { StdioServerTransport } from '@modelcontextprotocol/sdk/server/stdio.js';
import Stripe from 'stripe';

const stripe = new Stripe(process.env.STRIPE_API_KEY);

const server = new Server(
  {
    name: 'stripe-server',
    version: '1.0.0',
  },
  {
    capabilities: {
      tools: {},
      resources: {},
    },
  }
);

// Define tools
server.setRequestHandler(ListToolsRequestSchema, async () => {
  return {
    tools: [
      {
        name: 'list_recent_payments',
        description: 'List recent payment transactions',
        inputSchema: {
          type: 'object',
          properties: {
            limit: {
              type: 'number',
              description: 'Number of payments to retrieve',
              default: 10,
            },
          },
        },
      },
      {
        name: 'get_customer',
        description: 'Get customer details by ID',
        inputSchema: {
          type: 'object',
          properties: {
            customer_id: {
              type: 'string',
              description: 'Stripe customer ID',
            },
          },
          required: ['customer_id'],
        },
      },
      {
        name: 'refund_payment',
        description: 'Process a refund for a payment',
        inputSchema: {
          type: 'object',
          properties: {
            payment_intent_id: {
              type: 'string',
              description: 'Payment intent ID to refund',
            },
            amount: {
              type: 'number',
              description: 'Amount to refund (optional, full refund if not specified)',
            },
          },
          required: ['payment_intent_id'],
        },
      },
    ],
  };
});

// Handle tool execution
server.setRequestHandler(CallToolRequestSchema, async (request) => {
  const { name, arguments: args } = request.params;
  
  try {
    if (name === 'list_recent_payments') {
      const payments = await stripe.paymentIntents.list({
        limit: args.limit || 10,
      });
      
      return {
        content: [
          {
            type: 'text',
            text: JSON.stringify(payments.data, null, 2),
          },
        ],
      };
    }
    
    if (name === 'get_customer') {
      const customer = await stripe.customers.retrieve(args.customer_id);
      
      return {
        content: [
          {
            type: 'text',
            text: JSON.stringify(customer, null, 2),
          },
        ],
      };
    }
    
    if (name === 'refund_payment') {
      const refund = await stripe.refunds.create({
        payment_intent: args.payment_intent_id,
        amount: args.amount,
      });
      
      return {
        content: [
          {
            type: 'text',
            text: `Refund processed: ${JSON.stringify(refund, null, 2)}`,
          },
        ],
      };
    }
    
    throw new Error(`Unknown tool: ${name}`);
  } catch (error) {
    return {
      content: [
        {
          type: 'text',
          text: `Error: ${error.message}`,
        },
      ],
      isError: true,
    };
  }
});

async function main() {
  const transport = new StdioServerTransport();
  await server.connect(transport);
}

main().catch(console.error);
```

## Security Considerations

### 1. Credential Management

**âŒ Bad Practice:**
```json
{
  "mcpServers": {
    "github": {
      "env": {
        "GITHUB_TOKEN": "ghp_hardcoded_token_in_config"
      }
    }
  }
}
```

**âœ… Good Practice:**
```json
{
  "mcpServers": {
    "github": {
      "env": {
        "GITHUB_TOKEN": "${GITHUB_TOKEN}"
      }
    }
  }
}
```

Set environment variables:
```bash
# In ~/.bashrc or ~/.zshrc
export GITHUB_TOKEN="ghp_your_token"
export POSTGRES_PASSWORD="secure_password"
```

Or use a secrets manager:
```bash
# Using 1Password CLI
export GITHUB_TOKEN=$(op read "op://Personal/GitHub/token")
```

### 2. Principle of Least Privilege

**GitHub token scopes:**
```
âœ… Only needed scopes:
  - repo (if private repos)
  - public_repo (if only public)
  - read:org (if reading org data)

âŒ Don't grant:
  - admin:org (unless absolutely necessary)
  - delete_repo (rarely needed)
```

**Database permissions:**
```sql
-- Create read-only user for Claude
CREATE USER claude_readonly WITH PASSWORD 'secure_password';
GRANT CONNECT ON DATABASE mydb TO claude_readonly;
GRANT SELECT ON ALL TABLES IN SCHEMA public TO claude_readonly;
-- No INSERT, UPDATE, DELETE privileges
```

### 3. Network Security

**Database connections:**
```json
{
  "mcpServers": {
    "postgres": {
      "env": {
        "POSTGRES_CONNECTION_STRING": "postgresql://user:pass@localhost:5432/db?sslmode=require"
      }
    }
  }
}
```

Always use:
- SSL/TLS connections (`sslmode=require`)
- VPN for remote databases
- SSH tunnels when appropriate

### 5. Rate Limiting

Implement rate limiting in custom servers:

```javascript
import rateLimit from 'express-rate-limit';

const limiter = {
  tokens: 100,
  refillRate: 10, // per minute
  lastRefill: Date.now(),
};

function checkRateLimit() {
  const now = Date.now();
  const elapsed = (now - limiter.lastRefill) / 60000;
  limiter.tokens = Math.min(100, limiter.tokens + elapsed * limiter.refillRate);
  limiter.lastRefill = now;
  
  if (limiter.tokens < 1) {
    throw new Error('Rate limit exceeded');
  }
  
  limiter.tokens -= 1;
}
```

### 6. Input Validation

Always validate inputs in custom servers:

```javascript
server.setRequestHandler(CallToolRequestSchema, async (request) => {
  const { arguments: args } = request.params;
  
  // Validate required fields
  if (!args.user_id || typeof args.user_id !== 'string') {
    throw new Error('Invalid user_id');
  }
  
  // Sanitize inputs
  const sanitizedUserId = args.user_id.replace(/[^a-zA-Z0-9-]/g, '');
  
  // Prevent SQL injection
  const query = 'SELECT * FROM users WHERE id = $1';
  const result = await db.query(query, [sanitizedUserId]);
});
```

### 6. Caching

**Cache expensive operations:**

```javascript
const cache = new Map();
const CACHE_TTL = 5 * 60 * 1000; // 5 minutes

async function getCachedData(key, fetchFn) {
  const cached = cache.get(key);
  
  if (cached && Date.now() - cached.timestamp < CACHE_TTL) {
    return cached.data;
  }
  
  const data = await fetchFn();
  cache.set(key, { data, timestamp: Date.now() });
  
  return data;
}
```

### Resources vs Tools

**Resources:** Data Claude can read
```javascript
server.setRequestHandler(ListResourcesRequestSchema, async () => {
  return {
    resources: [
      {
        uri: 'file:///logs/app.log',
        name: 'Application Logs',
        mimeType: 'text/plain',
      },
    ],
  };
});

server.setRequestHandler(ReadResourceRequestSchema, async (request) => {
  const uri = request.params.uri;
  const content = await fs.readFile(uri, 'utf-8');
  return {
    contents: [
      {
        uri,
        mimeType: 'text/plain',
        text: content,
      },
    ],
  };
});
```

### Prompts (Reusable Templates)

```javascript
server.setRequestHandler(ListPromptsRequestSchema, async () => {
  return {
    prompts: [
      {
        name: 'bug_report',
        description: 'Template for filing bug reports',
        arguments: [
          {
            name: 'title',
            description: 'Bug title',
            required: true,
          },
        ],
      },
    ],
  };
});

server.setRequestHandler(GetPromptRequestSchema, async (request) => {
  if (request.params.name === 'bug_report') {
    return {
      messages: [
        {
          role: 'user',
          content: {
            type: 'text',
            text: `Create a detailed bug report for: ${request.params.arguments.title}`,
          },
        },
      ],
    };
  }
});
```
