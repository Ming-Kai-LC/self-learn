## TodoWrite Tool

Track multi-step tasks with a structured todo list.

### When to Use

- Complex tasks with 3+ steps
- Multiple user requests
- Non-trivial implementations
- User explicitly requests todo tracking

### When NOT to Use

- Single, simple tasks
- Trivial operations
- Purely informational requests

### Task Structure

```python
TodoWrite(
    todos=[
        {
            "content": "Create user authentication module",
            "activeForm": "Creating user authentication module",
            "status": "in_progress"
        },
        {
            "content": "Write unit tests for auth",
            "activeForm": "Writing unit tests for auth",
            "status": "pending"
        },
        {
            "content": "Update documentation",
            "activeForm": "Updating documentation",
            "status": "pending"
        }
    ]
)
```

### Task States

- `pending`: Not yet started
- `in_progress`: Currently working (only ONE at a time)
- `completed`: Successfully finished

### Critical Rules

1. Each task needs `content` AND `activeForm`
2. Only ONE task in_progress at a time
3. Complete tasks immediately when done
4. Only mark completed when FULLY accomplished
5. Update list after each significant step

### Example Workflow

```python
# Step 1: Create initial todo list
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "in_progress"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "pending"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "pending"}
    ]
)

# ... implement feature ...

# Step 2: After implementing feature
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "completed"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "in_progress"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "pending"}
    ]
)

# ... write tests ...

# Step 3: After writing tests
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "completed"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "completed"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "in_progress"}
    ]
)
```

**Pro Tip:** TodoWrite helps you stay organized and shows progress to users.

## NotebookEdit Tool

Edit Jupyter notebooks (.ipynb files) by cell.

### Parameters

- `notebook_path` (required): Absolute path to .ipynb file
- `new_source` (required): New cell content
- `cell_id` (optional): ID of cell to edit/insert after
- `cell_type` (optional): "code" or "markdown"
- `edit_mode` (optional): "replace", "insert", or "delete"

### Examples

```python
# Replace cell content
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-3",
    new_source="import pandas as pd\nimport numpy as np"
)

# Insert new markdown cell
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-5",
    cell_type="markdown",
    edit_mode="insert",
    new_source="## Data Cleaning\n\nSteps to clean the dataset:"
)

# Insert new code cell
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-10",
    cell_type="code",
    edit_mode="insert",
    new_source="df = pd.read_csv('data.csv')\ndf.head()"
)

# Delete cell
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-7",
    edit_mode="delete",
    new_source=""  # Still required but ignored
)
```

### Edit Modes

- `replace`: Replace cell content (default)
- `insert`: Add new cell after specified cell_id
- `delete`: Remove cell

### Best Practices

- Read notebook first to see cell IDs
- Use meaningful cell content
- Organize with markdown cells
- Test code cells after editing

### Workflow Example

```python
# Step 1: Read notebook to see structure
Read(file_path="/home/user/project/analysis.ipynb")

# Step 2: Add section header
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-5",
    cell_type="markdown",
    edit_mode="insert",
    new_source="## Feature Engineering"
)

# Step 3: Add code
NotebookEdit(
    notebook_path="/home/user/project/analysis.ipynb",
    cell_id="cell-6",  # The cell we just inserted
    cell_type="code",
    edit_mode="insert",
    new_source="df['new_feature'] = df['col1'] * df['col2']"
)

# Step 4: Verify changes
Read(file_path="/home/user/project/analysis.ipynb")
```

## Background Process Management

```python
# Read output from background process
BashOutput(
    bash_id="shell-123",  # Get from /bashes command
)

# Filter output with regex
BashOutput(
    bash_id="shell-123",
    filter="ERROR|WARNING"
)

# Monitor specific patterns
BashOutput(
    bash_id="shell-123",
    filter="Server started|Error|Failed"
)
```

## Advanced Pattern 1: Parallel Tool Execution

Run independent tools simultaneously for speed.

### The Pattern

```python
# Find all relevant files at once
Glob(pattern="**/*.py")
Glob(pattern="**/*.js")
Glob(pattern="**/*.md")

# Search multiple patterns
Grep(pattern="TODO", output_mode="files_with_matches")
Grep(pattern="FIXME", output_mode="files_with_matches")
Grep(pattern="HACK", output_mode="files_with_matches")

# Read multiple files
Read(file_path="/home/user/project/src/app.py")
Read(file_path="/home/user/project/src/utils.py")
Read(file_path="/home/user/project/tests/test_app.py")
```

### When to Use

- All operations are independent
- No shared state between calls
- All likely to succeed
- Need results quickly

### When NOT to Use

- Operations depend on each other
- Need results from one to inform another
- Writing to same file
- Sequential logic required

### Performance Benefit

Instead of:
- Sequential: 3 operations × 100ms = 300ms

Parallel:
- Parallel: max(100ms, 100ms, 100ms) = 100ms

**3x faster!**

## Advanced Pattern 2: Progressive Search

Start broad, then narrow down for efficient discovery.

### The Pattern

```python
# Step 1: Find files with matches
Grep(
    pattern="class \\w+",
    type="py",
    output_mode="files_with_matches"
)
# Result: 150 files found

# Step 2: Narrow to specific pattern
Grep(
    pattern="class User\\w*",
    type="py",
    output_mode="files_with_matches"
)
# Result: 12 files found

# Step 3: Get actual content
Grep(
    pattern="class User\\w*",
    type="py",
    output_mode="content",
    **{"-A": 10}
)

# Step 4: Read specific files
Read(file_path="/home/user/project/src/models/user.py")
```

### Benefits

- Faster initial results
- Avoid overwhelming output
- Better understanding of codebase
- More targeted analysis
- Less data to process

### Example: Finding API Endpoints

```python
# 1. Find all route files
Grep(pattern="@app\\.route|@router", output_mode="files_with_matches")

# 2. Narrow to POST endpoints
Grep(pattern="POST.*@app\\.route", output_mode="files_with_matches")

# 3. Get full endpoint definitions
Grep(pattern="@app\\.route.*POST", output_mode="content", **{"-A": 5})

# 4. Read specific route file
Read(file_path="/home/user/project/api/auth_routes.py")
```

## Tool Permissions and Security

### File Access

**Read Tool:**
- Can access any file on the system
- Respects file permissions
- Safe for reading sensitive files
- No modification capability

**Write Tool:**
- Can create/overwrite files
- Requires confirmation for existing files
- Creates parent directories if needed
- Cannot delete files

**Edit Tool:**
- Can modify existing files
- Must read first (safety check)
- Exact string matching prevents accidents
- Cannot create new files

### Command Execution

**Bash Tool:**
- Runs with your user permissions
- Can execute any shell command
- No sudo prompts (would fail)
- Review commands before running

### Security Best Practices

1. **Review generated commands**
   - Check before executing destructive operations
   - Verify paths are correct
   - Understand what each command does

2. **Don't commit secrets**
   - Use .gitignore for sensitive files
   - Use environment variables
   - Never hardcode API keys

3. **Use environment variables**
   ```python
   # ❌ Bad
   API_KEY = "sk-1234567890abcdef"
   
   # ✅ Good
   import os
   API_KEY = os.getenv("API_KEY")
   ```

4. **Be cautious with destructive operations**
   - `rm -rf`
   - `git push --force`
   - `chmod -R 777`
   - `DROP DATABASE`

5. **Validate file paths**
   ```python
   # Check path before deleting
   Bash(command="ls -la /path/to/delete")  # Verify first
   Bash(command="rm -rf /path/to/delete")  # Then delete
   ```

6. **Use hooks for security checks**
   - Pre-commit hooks to check for secrets
   - Pre-push hooks to run tests
   - Git hooks to enforce standards

## Dangerous Operations to Avoid

### High Risk Commands

```bash
# ❌ DON'T: Delete everything
rm -rf /

# ❌ DON'T: Force push to main
git push --force origin main

# ❌ DON'T: Execute untrusted code
curl http://untrusted.com/script.sh | bash

# ❌ DON'T: Change system permissions
chmod 777 -R /

# ❌ DON'T: Disable safety checks
git commit --no-verify
```

### Safe Alternatives

```bash
# ✅ DO: Delete specific directory
rm -rf ./build

# ✅ DO: Regular push
git push origin feature-branch

# ✅ DO: Review before executing
curl http://trusted.com/script.sh > script.sh
# Review script.sh
bash script.sh

# ✅ DO: Specific permissions
chmod 755 ./scripts/build.sh

# ✅ DO: Run hooks
git commit  # Let hooks run
```

### Verification Workflow

```python
# Before destructive operations, verify:

# 1. Check what will be affected
Bash(command="git status")

# 2. Dry run if possible
Bash(command="git clean -n")  # -n = dry run

# 3. Actually execute
Bash(command="git clean -f")  # -f = force
```

## Performance Optimization

### 1. Use Parallel Execution

```python
# ❌ Slow: Sequential (300ms total)
Read(file_path="/home/user/project/app.py")     # 100ms
Read(file_path="/home/user/project/utils.py")   # 100ms
Read(file_path="/home/user/project/config.py")  # 100ms

# ✅ Fast: Parallel (100ms total)
# Call all at once in same message
Read(file_path="/home/user/project/app.py")
Read(file_path="/home/user/project/utils.py")
Read(file_path="/home/user/project/config.py")
```

### 2. Use Efficient Output Modes

```python
# ❌ Slow: Get full content when you just need file list
Grep(pattern="TODO", output_mode="content")

# ✅ Fast: Get only what you need
Grep(pattern="TODO", output_mode="files_with_matches")
```

### 3. Limit Results

```python
# ❌ Slow: Returns thousands of results
Grep(pattern="import", output_mode="content")

# ✅ Fast: Limit to what you need
Grep(
    pattern="import",
    output_mode="content",
    head_limit=50
)
```

### 4. Use Specific Patterns

```python
# ❌ Slow: Too broad
Grep(pattern="function")

# ✅ Fast: Specific
Grep(
    pattern="function handleSubmit",
    type="js"
)
```

### 5. Filter by File Type

```python
# ❌ Slow: Search all files
Grep(pattern="class")

# ✅ Fast: Search only Python files
Grep(pattern="class", type="py")
```

### 6. Progressive Narrowing

```python
# Start broad (fast)
Grep(pattern="User", output_mode="files_with_matches")
# Result: 100 files

# Narrow down
Grep(pattern="class User", output_mode="files_with_matches")
# Result: 15 files

# Get specific
Grep(pattern="class UserAuth", output_mode="content")
# Result: 3 matches with context
```

### 7. Avoid Multiline When Possible

```python
# ❌ Slower: Multiline mode
Grep(
    pattern="class[\\s\\S]*?\\{",
    multiline=True
)

# ✅ Faster: Single-line pattern
Grep(pattern="class \\w+")
```

### 8. Batch Independent Operations

```python
# ❌ Slow: One at a time
# Message 1:
Glob(pattern="**/*.py")
# Message 2:
Glob(pattern="**/*.js")
# Message 3:
Glob(pattern="**/*.md")

# ✅ Fast: All in one message
Glob(pattern="**/*.py")
Glob(pattern="**/*.js")
Glob(pattern="**/*.md")
```
