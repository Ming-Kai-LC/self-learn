# Project Setup

Master the art of setting up projects for Claude Code - from initial configuration to team collaboration and production-ready setups.

## What Is Project Setup?

Project setup in Claude Code involves configuring your project to maximize Claude's effectiveness:

- **Configuration files** - Define how Claude interacts with your project
- **Permissions and tools** - Control what Claude can access and modify
- **Project structure** - Organize files for optimal Claude understanding
- **Automation** - Set up hooks, skills, and commands for your workflow
- **Team collaboration** - Share configurations across your team
- **Environment-specific settings** - Different configs for dev, staging, production

### Why Proper Setup Matters

A well-configured project enables:

| Benefit | Impact |
|---------|--------|
| **Faster onboarding** | New team members productive immediately |
| **Consistent quality** | Automated checks and standards |
| **Enhanced safety** | Controlled permissions prevent accidents |
| **Better context** | Claude understands your project structure |
| **Team alignment** | Shared configurations ensure consistency |
| **Productivity boost** | Automated workflows save time |

## Core Configuration Files

### The `.claude` Directory

All Claude Code configuration lives in the `.claude` directory at your project root:

```
my-project/
├── .claude/
│   ├── settings.json         # Main configuration
│   ├── CLAUDE.md             # Project context for Claude
│   ├── commands/             # Custom slash commands
│   │   └── deploy.md
│   └── skills/               # Project-specific skills
│       └── code-review/
│           └── SKILL.md
├── .gitignore
├── package.json
└── src/
```

### settings.json - Main Configuration

The `settings.json` file controls Claude's behavior in your project.

**Basic configuration:**
```json
{
  "projectName": "My Awesome Project",
  "description": "A web application for managing tasks",
  
  "allowedTools": [
    "Read",
    "Write",
    "Edit",
    "Bash",
    "Grep",
    "Glob"
  ],
  
  "hooks": {
    "session-start": "npm install",
    "pre-commit": "npm test"
  },
  
  "ignorePaths": [
    "node_modules/**",
    "dist/**",
    "*.log"
  ]
}
```

**Complete configuration options:**
```json
{
  // Project metadata
  "projectName": "string",
  "description": "string",
  "version": "1.0.0",
  
  // Tool permissions
  "allowedTools": ["Read", "Write", "Edit", "Bash", "Grep", "Glob"],
  "disallowedTools": ["WebSearch"],
  
  // Paths
  "includePaths": ["src/**", "tests/**"],
  "ignorePaths": ["node_modules/**", "dist/**", ".git/**"],
  
  // Hooks
  "hooks": {
    "session-start": "command",
    "session-end": "command",
    "pre-commit": "command",
    "post-commit": "command",
    "before-bash": "command",
    "after-bash": "command",
    "before-write": "command",
    "after-write": "command"
  },
  
  // Behavior
  "autoCommit": false,
  "confirmDestructiveActions": true,
  
  // Context
  "contextFiles": ["README.md", "docs/ARCHITECTURE.md"],
  "maxContextFiles": 50,
  
  // Model preferences
  "preferredModel": "claude-sonnet-4-5",
  
  // Custom settings
  "customSettings": {
    "testCommand": "npm test",
    "lintCommand": "npm run lint",
    "buildCommand": "npm run build"
  }
}
```

### CLAUDE.md - Project Context

The `CLAUDE.md` file provides context about your project to Claude. This is automatically read when Claude starts working on your project.

**Example CLAUDE.md:**
```markdown
# My Awesome Project

A modern web application for task management built with React and Node.js.

## Architecture

- **Frontend:** React 18 with TypeScript
- **Backend:** Node.js with Express
- **Database:** PostgreSQL with Prisma ORM
- **Testing:** Jest and React Testing Library
- **Deployment:** Docker containers on AWS ECS

## Project Structure

```
src/
├── client/           # React frontend
│   ├── components/   # Reusable UI components
│   ├── pages/        # Page components
│   ├── hooks/        # Custom React hooks
│   └── utils/        # Utility functions
├── server/           # Express backend
│   ├── routes/       # API routes
│   ├── controllers/  # Request handlers
│   ├── services/     # Business logic
│   └── middleware/   # Express middleware
└── shared/           # Shared types and utilities
```

## Coding Standards

### TypeScript
- Use strict mode
- Explicit return types for functions
- Prefer interfaces over types
- No `any` types allowed

### React
- Functional components with hooks
- Props interfaces in same file
- One component per file
- CSS modules for styling

### API Design
- RESTful endpoints
- Versioned API (/api/v1/...)
- JSON responses with consistent format:
  ```json
  {
    "success": true,
    "data": {},
    "meta": {}
  }
  ```

### Testing
- Unit tests for all business logic
- Integration tests for API endpoints
- E2E tests for critical user flows
- Minimum 80% code coverage

## Git Workflow

- Main branch: `main` (protected)
- Feature branches: `feature/description`
- Bug fixes: `fix/description`
- Commit format: `type(scope): message`
  - Types: feat, fix, docs, style, refactor, test, chore

## Development Workflow

1. Create feature branch from `main`
2. Implement feature with tests
3. Run `npm run lint` and `npm test`
4. Create PR with description
5. Code review required
6. Merge to main after approval

## Environment Variables

Required for development:
```
DATABASE_URL=postgresql://...
JWT_SECRET=...
AWS_REGION=us-east-1
NODE_ENV=development
```

## Common Commands

```bash
npm run dev          # Start development servers
npm test             # Run all tests
npm run lint         # Lint code
npm run build        # Build for production
npm run migrate      # Run database migrations
```

## Important Notes

- Never commit secrets or API keys
- Always run tests before committing
- Update this file when architecture changes
- See ARCHITECTURE.md for detailed design docs
```

## Recommended Project Structure

### Web Application (Full-Stack)

```

**Restrictive (Production/Review):**
```json
{
  "allowedTools": [
    "Read",
    "Grep",
    "Glob"
  ],
  "disallowedTools": [
    "Write",
    "Edit",
    "Bash"
  ]
}
```

**Balanced (Most Projects):**
```json
{
  "allowedTools": [
    "Read",
    "Write",
    "Edit",
    "Bash",
    "Grep",
    "Glob"
  ],
  "confirmDestructiveActions": true
}
```

### Path Restrictions

**Include specific paths:**
```json
{
  "includePaths": [
    "src/**",
    "tests/**",
    "docs/**",
    "*.md"
  ]
}
```

**Ignore sensitive paths:**
```json
{
  "ignorePaths": [
    "node_modules/**",
    "dist/**",
    "build/**",
    ".git/**",
    "*.log",
    ".env*",
    "secrets/**",
    "credentials/**",
    "*.key",
    "*.pem"
  ]
}
```

### File Type Restrictions

**Allow only specific file types:**
```json
{
  "allowedFileTypes": [
    ".js",
    ".ts",
    ".jsx",
    ".tsx",
    ".json",
    ".md"
  ]
}
```

**Block dangerous file types:**
```json
{
  "disallowedFileTypes": [
    ".exe",
    ".dll",
    ".sh",
    ".bat"
  ]
}
```

### Read-Only Mode

For code review or analysis only:
```json
{
  "readOnly": true,
  "allowedTools": [
    "Read",
    "Grep",
    "Glob"
  ]
}
```

## Team Setup and Collaboration

### Sharing Configuration

**Commit `.claude` directory to git:**
```bash
# .gitignore - DON'T ignore .claude/
node_modules/
dist/
.env
# .claude/  ← Don't add this line!
```

**Everyone gets the same setup:**
```bash
git clone <repo>
cd my-project
# .claude/ automatically available!
claude
```

### User-Level Overrides

Team members can override project settings:

**Project settings (committed):**
```json
// .claude/settings.json
{
  "hooks": {
    "pre-commit": "npm test"
  }
}
```

**User override (local only):**
```json
// ~/.claude/settings.json
{
  "hooks": {
    "pre-commit": "npm test -- --watchAll=false --bail"
  },
  "preferredModel": "claude-sonnet-4-5"
}
```

### Complete Workflow Setup

**1. Session Hooks:**
```json
{
  "hooks": {
    "session-start": "git fetch && npm install && docker-compose up -d",
    "session-end": "docker-compose down"
  }
}
```

**2. Quality Hooks:**
```json
{
  "hooks": {
    "after-write": "prettier --write $FILE_PATH",
    "pre-commit": "npm run lint && npm test"
  }
}
```

**3. Project Skills:**
```
.claude/skills/
├── code-review/
│   └── SKILL.md          # Team's review checklist
├── component-gen/
│   ├── SKILL.md
│   └── templates/
│       ├── Component.tsx
│       ├── Component.test.tsx
│       └── Component.module.css
└── api-endpoint/
    ├── SKILL.md
    └── templates/
        ├── route.ts
        ├── controller.ts
        └── test.ts
```

**4. Custom Commands:**
```
.claude/commands/
├── test.md           # Run specific test suites
├── deploy.md         # Deploy to environments
├── db-migrate.md     # Run migrations
├── docs.md           # Generate documentation
└── review-pr.md      # Review pull requests
```

**Example command - test.md:**
```markdown
---
description: Run tests with options for unit, integration, or e2e tests
---

Run the project's test suite.

Ask the user which tests to run:
1. All tests
2. Unit tests only
3. Integration tests only
4. E2E tests only
5. Specific test file

Then execute the appropriate command:
- All: `npm test`
- Unit: `npm test -- --testPathPattern=unit`
- Integration: `npm test -- --testPathPattern=integration`
- E2E: `npm run test:e2e`
- Specific: `npm test -- <file-path>`

After tests complete:
- Show summary (passed, failed, coverage)
- If failures, offer to help debug
- Suggest next steps
```

**Example command - deploy.md:**
```markdown
---
description: Deploy application to staging or production environments
---

Deploy the application following our deployment process.

## Pre-deployment Checks

1. Verify clean git status
2. Run full test suite
3. Check for uncommitted changes
4. Ensure on correct branch

## Deployment Steps

Ask which environment:
- **Staging:** `npm run deploy:staging`
- **Production:** `npm run deploy:prod` (requires confirmation)

For production:
1. Create git tag
2. Show confirmation prompt
3. Deploy only if confirmed
4. Monitor deployment logs

## Post-deployment

1. Verify deployment successful
2. Run health checks
3. Update team in Slack
```

## Environment-Specific Configurations

### Development Environment

**settings.dev.json:**
```json
{
  "projectName": "My Project (Development)",
  
  "allowedTools": [
    "Read",
    "Write",
    "Edit",
    "Bash",
    "Grep",
    "Glob",
    "WebSearch"
  ],
  
  "hooks": {
    "session-start": "npm install && docker-compose up -d && npm run dev &",
    "after-write": "if [[ $FILE_PATH == *.ts || $FILE_PATH == *.tsx ]]; then eslint --fix $FILE_PATH; fi",
    "pre-commit": "npm run lint && npm test -- --bail",
    "session-end": "docker-compose down"
  },
  
  "ignorePaths": [
    "node_modules/**",
    "dist/**",
    "*.log"
  ],
  
  "customSettings": {
    "autoReload": true,
    "verboseLogging": true
  }
}
```

### Staging Environment

**settings.staging.json:**
```json
{
  "projectName": "My Project (Staging)",
  
  "allowedTools": [
    "Read",
    "Write",
    "Edit",
    "Bash",
    "Grep",
    "Glob"
  ],
  
  "hooks": {
    "pre-commit": "npm run lint && npm test && npm run build",
    "post-commit": "git push origin staging"
  },
  
  "confirmDestructiveActions": true,
  
  "customSettings": {
    "requireTests": true,
    "requireReview": false
  }
}
```

### Production Environment

**settings.prod.json:**
```json
{
  "projectName": "My Project (PRODUCTION)",
  
  "readOnly": true,
  
  "allowedTools": [
    "Read",
    "Grep",
    "Glob"
  ],
  
  "disallowedTools": [
    "Write",
    "Edit",
    "Bash"
  ],
  
  "ignorePaths": [
    "**/*"
  ],
  
  "includePaths": [
    "logs/**",
    "*.log"
  ],
  
  "customSettings": {
    "readOnlyMode": true,
    "auditLog": true
  }
}
```

### Switching Environments

**Using environment variables:**
```bash
# Development
export CLAUDE_ENV=dev
claude --config .claude/settings.dev.json

# Staging
export CLAUDE_ENV=staging
claude --config .claude/settings.staging.json

# Production (read-only)
export CLAUDE_ENV=prod
claude --config .claude/settings.prod.json
```

**Using scripts:**
```json
// package.json
{
  "scripts": {
    "claude:dev": "claude --config .claude/settings.dev.json",
    "claude:staging": "claude --config .claude/settings.staging.json",
    "claude:prod": "claude --config .claude/settings.prod.json"
  }
}
```

## Best Practices for Project Organization

### 1. Clear Project Context

**Comprehensive CLAUDE.md:**
```markdown
# Project Name

## What This Project Does
Clear, concise description

## Architecture
- Tech stack
- Key dependencies
- Architecture patterns

## Project Structure
Explain directory organization

## Coding Standards
Your team's standards

## Common Tasks
How to accomplish common goals

## Important Notes
Gotchas, warnings, critical info
```

### 2. Consistent File Organization

```
✅ Good organization:
src/
├── components/        # Clear purpose
│   ├── Button/
│   │   ├── Button.tsx
│   │   ├── Button.test.tsx
│   │   └── Button.module.css
│   └── Input/
│       ├── Input.tsx
│       ├── Input.test.tsx
│       └── Input.module.css
└── utils/            # Clear purpose
    ├── date.ts
    └── format.ts

❌ Poor organization:
src/
├── stuff/            # Vague
├── misc/             # Unclear
├── temp/             # Temporary code stays
└── old/              # Old code lingers
```

### 3. Comprehensive .gitignore

```bash
# Dependencies
node_modules/
vendor/

# Build outputs
dist/
build/
*.out

# Environment
.env
.env.local
.env.*.local

# IDE
.vscode/
.idea/
*.swp
*.swo

# OS
.DS_Store
Thumbs.db

# Logs
*.log
logs/

# Testing
coverage/
.nyc_output/

# Don't ignore .claude/ - keep it!
# .claude/  ← Don't add this
```

## Troubleshooting

### Configuration Not Loading

**Problem:** Claude doesn't recognize configuration

**Checklist:**
```bash
✓ File is at .claude/settings.json (correct location)
✓ JSON syntax is valid (check with `jq . .claude/settings.json`)
✓ File has correct permissions (readable)
✓ Restart Claude session
```

**Test:**
```bash
# Validate JSON
cat .claude/settings.json | jq .

# Check file exists
ls -la .claude/settings.json

# Check permissions
chmod 644 .claude/settings.json
```

### Hooks Not Executing

**Problem:** Hooks don't run

**Solutions:**

**1. Check hook name:**
```json
// ❌ Wrong
"sessionstart": "npm install"

// ✅ Correct
"session-start": "npm install"
```

**2. Verify command exists:**
```bash
# Test command manually
which npm
npm install
```

**3. Check command syntax:**
```json
// ❌ Invalid shell syntax
"pre-commit": "npm test &"

// ✅ Valid syntax
"pre-commit": "npm test"
```

**4. Add logging:**
```json
{
  "hooks": {
    "session-start": "echo 'Hook running...' && npm install"
  }
}
```

### CLAUDE.md Not Read

**Problem:** Claude doesn't seem to know project context

**Solutions:**

**1. Check file location:**
```bash
ls -la .claude/CLAUDE.md
```

**2. Verify in contextFiles:**
```json
{
  "contextFiles": [
    ".claude/CLAUDE.md"
  ]
}
```

**3. Explicitly reference:**
```
"Read the project context in .claude/CLAUDE.md and then..."
```

### Paths Not Ignored

**Problem:** Claude can see files that should be ignored

**Solutions:**

**1. Check glob pattern:**
```json
{
  "ignorePaths": [
    "node_modules/**",  // ✅ Correct (recursive)
    "node_modules",     // ❌ Only ignores the directory itself
    "*.log"             // ✅ Correct (all log files)
  ]
}
```

**2. Use both ignore and include:**
```json
{
  "includePaths": ["src/**", "tests/**"],
  "ignorePaths": ["src/generated/**"]
}
```

### Command Not Found

**Problem:** `/mycommand` not recognized

**Checklist:**
```bash
✓ File is at .claude/commands/mycommand.md
✓ File has frontmatter with description
✓ Restart Claude session
```

**Test:**
```bash
# List all commands
# (Use /help or /commands in Claude)

# Check file
cat .claude/commands/mycommand.md
```

### Skill Not Available

**Problem:** Skill doesn't appear in `/skills`

**Solutions:**

**1. Check structure:**
```bash
# ✅ Correct
.claude/skills/skill-name/SKILL.md

# ❌ Wrong
.claude/skills/SKILL.md
.claude/skills/skill-name.md
```

**2. Validate frontmatter:**
```markdown
---
name: skill-name
description: When to use this skill
---
```

**3. Check case sensitivity:**
```bash
# File must be named SKILL.md (all caps)
ls .claude/skills/*/SKILL.md
```

### Permission Denied

**Problem:** Claude can't write/execute despite settings

**Solutions:**

**1. Check allowedTools:**
```json
{
  "allowedTools": [
    "Write",  // Needed for file creation
    "Edit",   // Needed for file modification
    "Bash"    // Needed for commands
  ]
}
```

**2. Check file permissions:**
```bash
# Make files writable
chmod 644 src/**/*.js

# Make directories writable
chmod 755 src/
```

**3. Check path restrictions:**
```json
{
  "includePaths": ["src/**"],  // Claude can only access src/
  "ignorePaths": ["src/generated/**"]  // But not src/generated/
}
```

### Environment Variables Not Available

**Problem:** Hooks can't access environment variables

**Solutions:**

**1. Source .env in hook:**
```json
{
  "hooks": {
    "session-start": "source .env && npm run dev"
  }
}
```

**2. Use absolute paths:**
```json
{
  "hooks": {
    "session-start": "export PATH=$PATH:/usr/local/bin && npm install"
  }
}
```

**3. Use script file:**
```json
{
  "hooks": {
    "session-start": "./scripts/setup.sh"
  }
}
```

## Key Takeaways

### 1. Configuration is Key

Proper setup enables:
- **Faster onboarding** - New team members productive immediately
- **Consistent quality** - Automated checks ensure standards
- **Enhanced safety** - Controlled permissions prevent accidents
- **Better context** - Claude understands your project

### 2. Essential Files

```
.claude/
├── settings.json      # Main configuration (REQUIRED)
├── CLAUDE.md          # Project context (HIGHLY RECOMMENDED)
├── commands/          # Custom commands (OPTIONAL)
└── skills/            # Project skills (OPTIONAL)
```

### 3. Configuration Structure

```json
{
  // Metadata
  "projectName": "string",
  "description": "string",
  
  // Permissions
  "allowedTools": ["Read", "Write", "Edit", "Bash"],
  "ignorePaths": ["node_modules/**", "dist/**"],
  
  // Automation
  "hooks": {
    "session-start": "setup command",
    "pre-commit": "quality checks"
  },
  
  // Context
  "contextFiles": ["README.md", ".claude/CLAUDE.md"]
}
```

### 4. CLAUDE.md Best Practices

Include:
- ✅ Project overview and purpose
- ✅ Architecture and tech stack
- ✅ Project structure explanation
- ✅ Coding standards and conventions
- ✅ Common commands and workflows
- ✅ Important notes and gotchas

### 5. Permission Levels

| Environment | Tools | Use Case |
|-------------|-------|----------|
| **Development** | All tools | Full development access |
| **Review** | Read, Grep, Glob | Code review only |
| **Production** | Read only | Monitoring/debugging |

### 6. Team Collaboration

**DO:**
- ✅ Commit `.claude/` to git
- ✅ Document setup in README
- ✅ Share skills and commands
- ✅ Use consistent standards

**DON'T:**
- ❌ Include secrets in settings
- ❌ Use absolute paths
- ❌ Assume team has same tools
- ❌ Forget to document changes

### 7. Hook Strategy

```json
{
  "hooks": {
    // Setup
    "session-start": "Install and start services",
    
    // Quality
    "after-write": "Auto-format",
    "pre-commit": "Lint and test",
    
    // Cleanup
    "session-end": "Stop services"
  }
}
```
