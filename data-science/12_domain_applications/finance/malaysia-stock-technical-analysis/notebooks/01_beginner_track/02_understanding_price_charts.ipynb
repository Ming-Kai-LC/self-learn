## Part 1: What Are Candlestick Charts?

### üìú A Brief History

Candlestick charts were invented by **Munehisa Homma**, a Japanese rice trader in the 1700s. Over 300 years later, they're still the most popular way to visualize stock prices!

**Why?** Because they show WAY more information than simple line charts in a single glance.

### üéØ Why Use Candlesticks Instead of Line Charts?

**Line Chart:**
- Shows only closing prices
- Misses important intraday information
- Can't show battle between buyers and sellers

**Candlestick Chart:**
- Shows Open, High, Low, AND Close
- Reveals the "story" of each trading day
- Shows who won: buyers or sellers

Think of it this way:
- **Line chart** = Seeing only the final score of a football match
- **Candlestick** = Watching the entire game, seeing all the action!

---

### üìä Anatomy of a Candlestick

Each candlestick represents **one trading period** (could be a day, hour, week, etc.) and has three parts:

#### 1. **The Body** (Rectangle)
- The thick rectangular part
- Shows the range between **Open** and **Close**
- **Color matters:**
  - **Green/White body**: Close > Open (Price went UP üìà - Bullish!)
  - **Red/Black body**: Close < Open (Price went DOWN üìâ - Bearish!)

#### 2. **Upper Shadow / Upper Wick** (Line above body)
- The thin line extending above the body
- Shows how high the price went during the day
- Represents the **High** price
- *What it tells us:* Buyers pushed price up, but couldn't hold it

#### 3. **Lower Shadow / Lower Wick** (Line below body)
- The thin line extending below the body
- Shows how low the price went during the day
- Represents the **Low** price
- *What it tells us:* Sellers pushed price down, but buyers fought back

### üîç Reading the Shadows (Wicks)

**Long Upper Shadow:**
- Buyers pushed price high but couldn't hold it
- Sellers are strong
- *Often bearish signal*

**Long Lower Shadow:**
- Sellers pushed price low but buyers fought back
- Buyers are strong
- *Often bullish signal*

**No Shadows (Marubozu):**
- Strong conviction in one direction
- Price opened at one extreme and closed at the other
- Very powerful signal!

## Part 3: Support and Resistance Levels

**Support and Resistance** are THE MOST important concepts in technical analysis. Master these, and you'll already be ahead of most beginners!

### üéØ What Are These Levels?

Think of stock prices like a bouncing ball in a room:

**Support Level** = The FLOOR
- A price level where the stock tends to **find buying interest**
- Price has difficulty falling below this level
- When price approaches support, **buyers step in** to catch the falling price
- Like a trampoline - price bounces back UP

**Resistance Level** = The CEILING
- A price level where the stock tends to **find selling pressure**
- Price has difficulty rising above this level  
- When price approaches resistance, **sellers step in** to take profits
- Like hitting your head on the ceiling - price bounces back DOWN

---

### üìñ Real-World Example

Imagine UUE stock:
- Has been bouncing between RM 0.50 and RM 0.60 for months
- Every time it drops to RM 0.50, buyers say "Great price!" and buy
- Every time it rises to RM 0.60, sellers say "Time to sell!" and take profits

Result:
- **RM 0.50 = Support** (floor - price bounces UP from here)
- **RM 0.60 = Resistance** (ceiling - price bounces DOWN from here)

---

### üß† The Psychology Behind It

**Why do Support and Resistance exist?**

**At Support (RM 0.50):**
- Previous buyers who bought at RM 0.50: "This is a good price, let me buy more!"
- Previous sellers who sold at RM 0.50: "I regret selling, let me buy back!"
- New buyers: "Everyone seems to buy at RM 0.50, I should too!"

**At Resistance (RM 0.60):**
- Previous buyers at RM 0.50: "20% profit! Time to sell!"
- Previous buyers at RM 0.60 who are stuck: "Finally breaking even, let me sell!"
- Traders: "It always falls from RM 0.60, let me sell before it drops!"

It's all about **human psychology and memory**!

---

### üí∞ Why They Matter (EXTREMELY Important!)

#### 1. **Trading Decisions**
- **Buy near support** ‚Üí Low risk, high reward (floor is nearby)
- **Sell near resistance** ‚Üí Take profits before price falls
- **Avoid buying near resistance** ‚Üí High risk (ceiling nearby)

#### 2. **Breakouts = Big Moves**
- When price **breaks ABOVE resistance** ‚Üí üöÄ Often shoots much higher!  
  - *Why?* All the sellers who were waiting are done, only buyers left
  
- When price **breaks BELOW support** ‚Üí üìâ Often drops much further!
  - *Why?* All the buyers who were waiting give up, panic selling begins

#### 3. **Risk Management**  
- **Set stop-losses below support** ‚Üí If support breaks, you exit with small loss
- **Set profit targets at resistance** ‚Üí Lock in gains before price reverses

#### 4. **Role Reversal (Mind-Blowing Concept!)**
- **Old resistance becomes NEW support** when broken upward
  - *Example*: Stock breaks above RM 0.60 resistance, then RM 0.60 becomes the new floor!
  
- **Old support becomes NEW resistance** when broken downward
  - *Example*: Stock breaks below RM 0.50 support, then RM 0.50 becomes the new ceiling!

---

### ü§î Think About It

Look at this scenario:

```
Current Price: RM 0.55
Support: RM 0.50 (bounced here 3 times in past 2 months)
Resistance: RM 0.60 (rejected here 3 times in past 2 months)
```

**Question 1:** If you buy at RM 0.55, where would you place your stop-loss?
- **Answer:** Just below RM 0.50 (maybe RM 0.49) - if support breaks, you're wrong, exit!

**Question 2:** If you buy at RM 0.55, where's your first profit target?
- **Answer:** Just below RM 0.60 (maybe RM 0.59) - take profits before hitting resistance!

**Question 3:** If price breaks ABOVE RM 0.60 with high volume, what should you do?
- **Answer:** This is a bullish breakout! Price might go much higher. Consider buying or holding!

---

### üí° How to Identify Support and Resistance

Look for **price levels where**:

1. **Multiple Bounces** ‚Üí Price touched it 2-3+ times
2. **Strong Reactions** ‚Üí Price bounced sharply (not slowly)
3. **Round Numbers** ‚Üí Humans like RM 0.50, RM 1.00, RM 5.00 (psychological levels)
4. **Recent Highs/Lows** ‚Üí Markets have memory, recent extremes matter
5. **High Volume** ‚Üí Many participants = stronger level

---

### ‚ö†Ô∏è Important Warnings

**Support and Resistance are ZONES, not exact prices!**
- Don't expect support to be exactly RM 0.500000
- Think of it as RM 0.48 - 0.52 (a zone around RM 0.50)
- Give some wiggle room!

**They're not magic - they CAN break!**
- When they break, they often break HARD
- Always use stop-losses to protect yourself
- Never assume "it will definitely bounce here"

**The more times a level is tested, the WEAKER it becomes**
- Like tapping on glass repeatedly - eventually it breaks
- If price tests resistance 5 times, the 6th time it might break through!

---

Now let's identify support and resistance for our UUE stock! üîç

````python
# ========================================
# Calculate Support and Resistance Levels
# ========================================

if "data" not in locals() or len(data) == 0:
    print("‚ö†Ô∏è Please run the previous cells first to fetch stock data.")
else:
    # ========================================
    # METHOD 1: Find Extreme Levels
    # ========================================

    # The simplest method: Find the highest high and lowest low
    # These are the "absolute" levels - price couldn't go beyond these

    resistance = data["High"].max()  # Highest price in 3 months
    # .max() finds the maximum value

    support = data["Low"].min()  # Lowest price in 3 months
    # .min() finds the minimum value

    # ========================================
    # METHOD 2: Find Intermediate Levels
    # ========================================

    # We also want to find levels BETWEEN the extremes
    # These are more "realistic" levels where price bounces multiple times

    # Get the top 10 highest prices
    # .nlargest(10) sorts and gives us the 10 highest values
    sorted_highs = data["High"].nlargest(10)

    # Get the bottom 10 lowest prices
    # .nsmallest(10) sorts and gives us the 10 lowest values
    sorted_lows = data["Low"].nsmallest(10)

    # Find the average of these top 10 and bottom 10
    # This gives us a "cluster" of where prices tend to peak/bottom
    # .mean() calculates the average
    resistance_2 = sorted_highs.mean()  # Average of top 10 highs
    support_2 = sorted_lows.mean()  # Average of bottom 10 lows

    # ========================================
    # DISPLAY THE LEVELS
    # ========================================

    print("üìä Support and Resistance Levels\n")
    print("=" * 60)

    # Show resistance levels (ceiling - where price struggles to go above)
    print(f"\nüî¥ RESISTANCE LEVELS (Selling Pressure):")
    print(f"   Level 1 (Strong):   RM {resistance:.2f}")
    print(f"   Level 2 (Moderate): RM {resistance_2:.2f}")

    # Show support levels (floor - where price struggles to go below)
    print(f"\nüü¢ SUPPORT LEVELS (Buying Interest):")
    print(f"   Level 1 (Moderate): RM {support_2:.2f}")
    print(f"   Level 2 (Strong):   RM {support:.2f}")

    # ========================================
    # WHERE IS PRICE NOW?
    # ========================================

    # Get the most recent closing price
    # .iloc[-1] means "get the last row"
    current_price = data["Close"].iloc[-1]
    print(f"\nüìç CURRENT PRICE: RM {current_price:.2f}")

    # ========================================
    # CALCULATE DISTANCES
    # ========================================

    # How far is the current price from resistance?
    # Formula: ((Target - Current) / Current) √ó 100
    # This gives us the % move needed to reach resistance
    dist_to_resistance = ((resistance - current_price) / current_price) * 100

    # How far is support below us?
    # This tells us our "downside risk"
    dist_to_support = ((current_price - support) / current_price) * 100

    # ========================================
    # SHOW DISTANCES AND INTERPRETATION
    # ========================================

    print(f"\nüìè Distance to Resistance: {dist_to_resistance:.1f}% above current price")
    print(f"   ‚Üí Price needs to rise {dist_to_resistance:.1f}% to hit resistance")

    print(f"\nüìè Distance to Support: {dist_to_support:.1f}% below current price")
    print(f"   ‚Üí If price falls {dist_to_support:.1f}%, it hits support")

    # ========================================
    # RISK/REWARD ANALYSIS
    # ========================================

    print(f"\nüí° Risk/Reward Analysis:")

    # If closer to support than resistance ‚Üí Lower risk to buy
    # If closer to resistance than support ‚Üí Higher risk to buy

    if current_price < (support + (resistance - support) * 0.4):
        # Price is in the lower 40% of the range (closer to support)
        print(f"   ‚úÖ Price is closer to SUPPORT")
        print(f"   ‚úÖ Relatively SAFE zone to consider buying")
        print(f"   ‚úÖ Good risk/reward ratio (floor nearby, ceiling far away)")
    elif current_price > (support + (resistance - support) * 0.6):
        # Price is in the upper 40% of the range (closer to resistance)
        print(f"   ‚ö†Ô∏è  Price is closer to RESISTANCE")
        print(f"   ‚ö†Ô∏è  RISKY zone to buy (ceiling nearby)")
        print(f"   ‚ö†Ô∏è  Consider waiting for a pullback toward support")
    else:
        # Price is in the middle 20% of the range
        print(f"   üìä Price is in the MIDDLE zone")
        print(f"   üìä Neutral area - watch for direction")
        print(f"   üìä Wait for clearer signals (move toward support or resistance)")

# üìù What you should see:
# - Two resistance levels (strong and moderate)
# - Two support levels (strong and moderate)
# - Current price position
# - Distance percentages to each level
# - Risk/reward assessment
#
# ü§î How to interpret:
# - Resistance Level 1 = Absolute highest in 3 months (very strong)
# - Resistance Level 2 = Cluster of recent highs (moderate)
# - Support Level 2 = Cluster of recent lows (moderate)
# - Support Level 1 = Absolute lowest in 3 months (very strong)
#
# üí° Trading implications:
# - If price is near support ‚Üí Good time to BUY (low risk)
# - If price is near resistance ‚Üí Good time to SELL (take profits)
# - If price is in middle ‚Üí WAIT for clearer signals
#
# ‚ö†Ô∏è Remember:
# - These are ZONES, not exact prices
# - Support/resistance can break!
# - Always use stop-losses to protect yourself
````

## Part 4: Basic Candlestick Patterns

Now that you can read individual candlesticks, let's learn to recognize **patterns** that predict future price movements!

### üéØ Why Patterns Matter

Certain candlestick shapes appear repeatedly before major price moves. Recognizing them gives you an edge!

Think of it like reading body language:
- Someone crossing their arms ‚Üí Defensive, closed off
- A Doji candlestick ‚Üí Market is indecisive, trend might change

**These patterns have been used for 300+ years because they WORK!**

### üìä Single Candlestick Patterns

Let's learn the 4 most important patterns every trader should know:

---

#### 1. **Doji** - The Indecision Candle

**What it looks like:**
```
   High ‚îÄ‚îÄ‚îê
          ‚îÇ ‚Üê Upper shadow
   O/C ‚îÄ‚îÄ‚îº‚îÄ‚îÄ ‚Üê Very small body (Open ‚âà Close)
          ‚îÇ ‚Üê Lower shadow
   Low ‚îÄ‚îÄ‚îò
```

**Characteristics:**
- Open ‚âà Close (body is tiny or non-existent)
- Can have long or short shadows
- Looks like a cross or plus sign (+)

**What it means:**
- **Indecision**: Neither buyers nor sellers won
- Market opened, battle ensued all day, ended at same price
- Like a tug-of-war ending in a tie

**When it matters:**
- **In an uptrend** ‚Üí Warning! Buyers losing steam, possible reversal down
- **In a downtrend** ‚Üí Warning! Sellers losing steam, possible reversal up
- **In sideways market** ‚Üí Just more sideways movement

**Real-world example:**
- Stock in strong uptrend for weeks
- Suddenly a Doji appears
- Next few days: Price starts falling (trend reversal!)

---

#### 2. **Hammer** - The Bullish Reversal

**What it looks like:**
```
   High ‚îÄ‚îÄ‚î¨‚îÄ‚îê
          ‚îÇ ‚îÇ ‚Üê Small body at TOP
   O/C ‚îÄ‚îÄ‚îî‚îÄ‚îò
          ‚îÇ
          ‚îÇ ‚Üê LONG lower shadow (2-3x body)
          ‚îÇ
   Low ‚îÄ‚îÄ‚îò
```

**Characteristics:**
- Small body at the TOP of the candle
- Long lower shadow (at least 2x body length)
- Little to no upper shadow
- Looks like a hammer üî®

**What it means:**
- **Sellers tried to push price down** (created long lower shadow)
- **Buyers fought back strongly** and pushed price back up
- Buyers won the battle! üí™

**When it matters:**
- **MUST appear after a downtrend** (at support level)
- Signals: "The selling is over, buyers are stepping in!"
- Often marks the bottom before price goes up

**Trading strategy:**
- Wait for next candle to CONFIRM (should be green/bullish)
- Buy when confirmation appears
- Set stop-loss below the hammer's low

**Real-world example:**
- Stock falling from RM 1.00 to RM 0.50
- Hammer appears at RM 0.50 (support)
- Next day: Green candle confirms reversal
- Stock rallies back to RM 0.70 (40% gain!)

---

#### 3. **Shooting Star** - The Bearish Reversal

**What it looks like:**
```
   High ‚îÄ‚îÄ‚îê
          ‚îÇ
          ‚îÇ ‚Üê LONG upper shadow (2-3x body)
          ‚îÇ
   O/C ‚îÄ‚îÄ‚î¨‚îÄ‚îê
          ‚îÇ ‚îÇ ‚Üê Small body at BOTTOM
   Low ‚îÄ‚îÄ‚îî‚îÄ‚îò
```

**Characteristics:**
- Small body at the BOTTOM of the candle
- Long upper shadow (at least 2x body length)
- Little to no lower shadow
- Looks like a shooting star üå†

**What it means:**
- **Buyers tried to push price up** (created long upper shadow)
- **Sellers fought back strongly** and pushed price back down
- Sellers won the battle! üìâ

**When it matters:**
- **MUST appear after an uptrend** (at resistance level)
- Signals: "The buying is over, sellers are taking control!"
- Often marks the top before price falls

**Trading strategy:**
- Wait for next candle to CONFIRM (should be red/bearish)
- Sell or take profits when confirmation appears
- If shorting, set stop-loss above the shooting star's high

**Real-world example:**
- Stock rising from RM 0.50 to RM 1.00
- Shooting star appears at RM 1.00 (resistance)
- Next day: Red candle confirms reversal
- Stock falls back to RM 0.70 (30% drop from top!)

---

#### 4. **Marubozu** - The Strong Conviction Candle

**What it looks like:**

**Bullish Marubozu (Green):**
```
   High/Close ‚îÄ‚îÄ‚î¨‚îÄ‚îê
                ‚îÇ ‚îÇ
                ‚îÇ ‚îÇ ‚Üê LARGE body
                ‚îÇ ‚îÇ ‚Üê NO shadows!
   Low/Open ‚îÄ‚îÄ‚îÄ‚îÄ‚îî‚îÄ‚îò
```

**Bearish Marubozu (Red):**
```
   High/Open ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îê
                ‚îÇ ‚îÇ
                ‚îÇ ‚îÇ ‚Üê LARGE body
                ‚îÇ ‚îÇ ‚Üê NO shadows!
   Low/Close ‚îÄ‚îÄ‚îÄ‚îî‚îÄ‚îò
```

**Characteristics:**
- Large body with little to no shadows
- Opens at one extreme, closes at the other
- "Marubozu" = "Bald" or "Shaved head" in Japanese

**What it means:**
- **EXTREME conviction** in one direction
- No back-and-forth, just straight movement
- Either strong buying or strong selling

**When it matters:**
- **Green Marubozu** ‚Üí Very bullish! Strong buying pressure
  - Buyers dominated from open to close
  - Often continues upward next day
  
- **Red Marubozu** ‚Üí Very bearish! Strong selling pressure
  - Sellers dominated from open to close
  - Often continues downward next day

**Trading strategy:**
- Green Marubozu after consolidation ‚Üí Buy signal
- Red Marubozu after rally ‚Üí Sell signal
- These are STRONG signals with high reliability

**Real-world example:**
- Stock consolidating around RM 0.50 for weeks
- Suddenly: Green Marubozu appears (RM 0.50 ‚Üí RM 0.58 in one day!)
- This signals strong institutional buying
- Stock continues to RM 0.70 over next few weeks

**2. Confirmation is Key**
- Don't trade on the pattern candle itself
- Wait for next candle to confirm
- Confirmation = price moving in expected direction

**3. Volume Validates**
- Pattern with HIGH volume = More reliable
- Pattern with LOW volume = Less reliable
- Volume shows conviction!

**4. Practice Makes Perfect**
- The more you look at charts, the better you get

- Like recognizing faces in a crowd

Let's now write code to automatically detect these patterns in our data! üîç

````python
# ========================================
# Automatic Pattern Detection
# ========================================

if "data" not in locals() or len(data) == 0:
    print("‚ö†Ô∏è Please run the previous cells first to fetch stock data.")
else:
    # ========================================
    # STEP 1: Calculate Candle Components
    # ========================================

    # For pattern detection, we need to measure parts of each candle

    # Body size = distance between Open and Close
    # abs() makes it always positive (we just care about size, not direction)
    data["Body"] = abs(data["Close"] - data["Open"])

    # Upper Shadow = distance from top of body to High
    # max(axis=1) finds the higher of Open or Close for each row
    # Then: High - (top of body) = upper shadow length
    data["Upper_Shadow"] = data["High"] - data[["Open", "Close"]].max(axis=1)

    # Lower Shadow = distance from Low to bottom of body
    # min(axis=1) finds the lower of Open or Close for each row
    # Then: (bottom of body) - Low = lower shadow length
    data["Lower_Shadow"] = data[["Open", "Close"]].min(axis=1) - data["Low"]

    # Total Range = Full candle height (High to Low)
    data["Total_Range"] = data["High"] - data["Low"]

    # ========================================
    # STEP 2: Identify DOJI Patterns
    # ========================================

    # A Doji has a very small body compared to total range
    # Rule: Body is less than 10% of the total range
    # Why 10%? It's a commonly used threshold in technical analysis

    data["Is_Doji"] = data["Body"] < (data["Total_Range"] * 0.1)

    # Explanation:
    # - If Body = 0.01 and Total_Range = 0.20, then 0.01 < 0.02 ‚Üí True (Doji)
    # - If Body = 0.05 and Total_Range = 0.20, then 0.05 > 0.02 ‚Üí False (Not a Doji)

    # ========================================
    # STEP 3: Identify HAMMER Patterns
    # ========================================

    # A Hammer has:
    # 1. Long lower shadow (at least 2x the body size)
    # 2. Small or no upper shadow (less than half of body)

    data["Is_Hammer"] = (data["Lower_Shadow"] > data["Body"] * 2) & (  # Long lower shadow
        data["Upper_Shadow"] < data["Body"] * 0.5
    )  # Small upper shadow

    # The & operator means AND - both conditions must be true

    # Explanation:
    # - Lower shadow must be 2x+ body length (rejection of lower prices)
    # - Upper shadow must be small (minimal testing of higher prices)
    # - This creates the "hammer" shape üî®

    # ========================================
    # STEP 4: Identify SHOOTING STAR Patterns
    # ========================================

    # A Shooting Star is the opposite of a Hammer:
    # 1. Long upper shadow (at least 2x the body size)
    # 2. Small or no lower shadow (less than half of body)

    data["Is_Shooting_Star"] = (data["Upper_Shadow"] > data["Body"] * 2) & (  # Long upper shadow
        data["Lower_Shadow"] < data["Body"] * 0.5
    )  # Small lower shadow

    # Explanation:
    # - Upper shadow must be 2x+ body length (rejection of higher prices)
    # - Lower shadow must be small (minimal testing of lower prices)
    # - This creates the "shooting star" shape üå†

    # ========================================
    # STEP 5: Count Pattern Occurrences
    # ========================================

    # .sum() counts how many True values in each column
    # True = 1, False = 0, so summing gives us the count

    doji_count = data["Is_Doji"].sum()
    hammer_count = data["Is_Hammer"].sum()
    shooting_star_count = data["Is_Shooting_Star"].sum()

    # ========================================
    # STEP 6: Display Results
    # ========================================

    print("üïØÔ∏è Candlestick Pattern Analysis\n")
    print("=" * 60)
    print(f"üìä Analysis Period: {len(data)} trading days")
    print(
        f"üìÖ Date Range: {data.index[0].strftime('%Y-%m-%d')} to {data.index[-1].strftime('%Y-%m-%d')}"
    )

    print(f"\nüìà Patterns Found:")
    print(f"   ‚Ä¢ Doji (Indecision):        {doji_count} occurrences")
    print(f"   ‚Ä¢ Hammer (Bullish):         {hammer_count} occurrences")
    print(f"   ‚Ä¢ Shooting Star (Bearish):  {shooting_star_count} occurrences")

    # ========================================
    # STEP 7: Show Recent Pattern Dates
    # ========================================

    # Show the most recent occurrences of each pattern
    # This helps you see if any patterns appeared recently!

    if doji_count > 0:
        # Get the last 3 Doji patterns (or fewer if less than 3 exist)
        # .tail(3) gets the last 3 rows where Is_Doji is True
        recent_doji = data[data["Is_Doji"]].tail(3)

        print(f"\nüîç Most Recent Doji Patterns:")
        for date in recent_doji.index:
            price = recent_doji.loc[date, "Close"]
            print(f"   üìç {date.strftime('%Y-%m-%d')} - Close: RM {price:.2f}")
            print(f"      ‚Üí Indecision signal - watch for trend change")

    if hammer_count > 0:
        recent_hammer = data[data["Is_Hammer"]].tail(3)

        print(f"\nüîç Most Recent Hammer Patterns:")
        for date in recent_hammer.index:
            price = recent_hammer.loc[date, "Close"]
            print(f"   üìç {date.strftime('%Y-%m-%d')} - Close: RM {price:.2f}")
            print(f"      ‚Üí Potential bullish reversal - check if at support")

    if shooting_star_count > 0:
        recent_star = data[data["Is_Shooting_Star"]].tail(3)

        print(f"\nüîç Most Recent Shooting Star Patterns:")
        for date in recent_star.index:
            price = recent_star.loc[date, "Close"]
            print(f"   üìç {date.strftime('%Y-%m-%d')} - Close: RM {price:.2f}")
            print(f"      ‚Üí Potential bearish reversal - check if at resistance")

    # ========================================
    # STEP 8: Pattern Frequency Analysis
    # ========================================

    total_patterns = doji_count + hammer_count + shooting_star_count

    if total_patterns > 0:
        print(f"\nüí° Pattern Insights:")
        print(f"   ‚Ä¢ Total patterns found: {total_patterns}")
        print(f"   ‚Ä¢ Pattern frequency: {(total_patterns/len(data)*100):.1f}% of trading days")

        if doji_count > hammer_count + shooting_star_count:
            print(f"   ‚Ä¢ Market shows INDECISION (many Dojis)")
            print(f"   ‚Ä¢ Likely consolidating or unsure of direction")
        elif hammer_count > shooting_star_count * 1.5:
            print(f"   ‚Ä¢ More Hammers than Shooting Stars")
            print(f"   ‚Ä¢ Suggests bulls attempting reversals from lows")
        elif shooting_star_count > hammer_count * 1.5:
            print(f"   ‚Ä¢ More Shooting Stars than Hammers")
            print(f"   ‚Ä¢ Suggests bears attempting reversals from highs")
        else:
            print(f"   ‚Ä¢ Balanced pattern distribution")
    else:
        print(f"\nüí° No major patterns detected in this period")
        print(f"   ‚Ä¢ This could mean strong trending market (less reversals)")
        print(f"   ‚Ä¢ Or patterns are too subtle for our thresholds")

# üìù What you should see:
# - Count of each pattern type found
# - Dates when patterns appeared
# - Interpretation of what patterns mean
# - Pattern frequency statistics
#
# ü§î How to use this:
# - Recent Doji ‚Üí Watch for trend change
# - Recent Hammer at support ‚Üí Potential buy opportunity
# - Recent Shooting Star at resistance ‚Üí Potential sell opportunity
# - Many Dojis ‚Üí Market is indecisive
#
# üí° Important Notes:
# - These are SIMPLIFIED pattern detection rules
# - Professional traders use more sophisticated algorithms
# - Always verify patterns visually on the chart
# - Context matters! Check WHERE patterns appear (support/resistance)
#
# ‚ö†Ô∏è Limitations:
# - Our thresholds (10%, 2x, etc.) are standard but not perfect
# - Some patterns might be missed
# - Some detected "patterns" might be false signals
# - Always combine with other analysis!
````

Imagine looking at a mountain:

| Time Frame | Each Candle = | Best For | Trader Type | Holding Period |
|------------|---------------|----------|-------------|----------------|
| **1 minute** | 1 minute of trading | Scalping | Day traders | Minutes |

### üéì Understanding Each Time Frame

#### **Short Time Frames (1min - 15min)**

**Characteristics:**
- Lots of "noise" (random price movements)
- Many fake signals and false breakouts
- Need to watch screen constantly
- High stress, fast decisions

**Good for:**
- Day traders who close all positions before market closes

- People who can dedicate full attention during market hours

**Challenges:**
- Very stressful and time-consuming
- Transaction costs add up quickly
- Requires experience and discipline
- Easy to overtrade and lose money

**üí° Beginner advice**: AVOID these unless you have time and experience!

---

#### **Medium Time Frames (1 hour - Daily)**

**Characteristics:**
- Balance between detail and overview
- Clearer trends than short time frames
- Less noise, more reliable signals
- Check charts 1-2 times per day

**Good for:**
- Swing traders holding positions for days/weeks
- Part-time traders with day jobs
- People learning technical analysis
- Most retail traders

**Advantages:**
- Less stressful than short time frames
- Lower transaction costs (fewer trades)
- Trends are clearer and more reliable
- Better work-life balance

**üí° Beginner advice**: START HERE! Daily charts are perfect for learning.

---

#### **Long Time Frames (Weekly - Monthly)**

**Characteristics:**
- Very clear long-term trends
- Minimal noise
- Slow-moving, patient approach
- Check charts weekly or monthly

**Good for:**
- Long-term investors
- Busy professionals
- Retirement accounts
- People who hate watching charts

**Advantages:**
- Least stressful
- Very low transaction costs
- Tax benefits (long-term capital gains)
- More time to think and research

**Challenges:**
- Miss short-term opportunities
- Slower to react to changes
- Need patience (positions held for months/years)

**üí° Beginner advice**: Great for retirement investing!

---

### üîç Multiple Time Frame Analysis (Pro Strategy)

Professional traders DON'T just look at one time frame. They use **multiple time frames** together!

**The 3-Timeframe Rule:**

#### **1. Higher Time Frame** (e.g., Weekly/Daily) - "The Boss"
- **Purpose**: Identify the OVERALL trend
- **What to ask**: Is this stock in an uptrend, downtrend, or sideways?
- **Rule**: This is your "boss" - only trade in its direction!

#### **2. Medium Time Frame** (e.g., Daily/4-hour) - "The Manager"
- **Purpose**: Find ENTRY opportunities within the trend
- **What to ask**: Where are good entry points in the trend?
- **Rule**: Look for pullbacks or breakouts here

#### **3. Lower Time Frame** (e.g., 1-hour/15-min) - "The Worker"
- **Purpose**: FINE-TUNE exact entry and exit points
- **What to ask**: What's the precise price to enter right now?
- **Rule**: Use this for timing, not direction!

---

### üìñ Real-World Example

Let's say you're analyzing UUE stock:

**Step 1: Check Weekly Chart (The Boss)**
- Trend: Strong uptrend for past 3 months
- **Decision**: Only look for BUY opportunities (trend is up)

**Step 2: Check Daily Chart (The Manager)**
- Current: Stock pulled back to RM 0.50 support
- Pattern: Hammer appeared yesterday
- **Decision**: This looks like a good entry zone!

**Step 3: Check 1-Hour Chart (The Worker)**
- Current: Price bouncing off RM 0.50
- Entry: Wait for green candle above RM 0.51
- **Decision**: Buy at RM 0.51 with stop at RM 0.49

**Result:**
- Weekly says: "Uptrend - buy"
- Daily says: "Here's the opportunity - support level"
- Hourly says: "Enter now at RM 0.51"

**This is how pros trade!**

---

### üéØ The Golden Rules

#### **Rule #1: Trade in the Direction of the Higher Time Frame**
- Weekly chart shows uptrend ‚Üí Only BUY on daily chart
- Weekly chart shows downtrend ‚Üí Only SELL on daily chart
- **Never fight the higher time frame trend!**

#### **Rule #2: Higher Time Frames are More Reliable**
- Daily chart > 1-hour chart > 15-minute chart
- The longer the time frame, the stronger the signal
- Short time frames lie often, long time frames rarely do

#### **Rule #3: Use Lower Time Frames for Entry, Not Direction**
- Don't decide "buy or sell" based on 5-minute chart
- Use daily/weekly for direction
- Use 5-min/15-min for precise entry timing

---

### ü§î Think About It

**Scenario:** You see these signals:

- **Weekly Chart**: Strong downtrend
- **Daily Chart**: Hammer pattern at support
- **Should you buy?**

**Answer:** NO! 
- The weekly "boss" says downtrend
- The daily hammer is a trap (false signal)
- This is called a "counter-trend trade" - very risky!
- **Wait for weekly trend to change first**

---

### üí° Time Frame Selection Guide

**How to choose YOUR time frame:**

1. **How much time can you dedicate?**
   - Less than 1 hour/day ‚Üí Daily charts or higher
   - 2-3 hours/day ‚Üí 1-hour or daily charts
   - Full day ‚Üí Any time frame

2. **How patient are you?**

````python
# ========================================
# Multiple Time Frame Comparison
# ========================================

if "stock" not in locals():
    print("‚ö†Ô∏è Please run the previous cells first to create the stock object.")
else:
    try:
        print("üìä Fetching data for multiple time frames...\n")

        # ========================================
        # STEP 1: Fetch Different Time Periods
        # ========================================

        # Fetch 1 month of daily data
        # Each candle = 1 trading day
        # Total candles: ~20-22 (markets closed weekends/holidays)
        data_1m = stock.history(period="1mo")

        # Fetch 6 months of daily data
        # Each candle = 1 trading day
        # Total candles: ~120-130
        data_6m = stock.history(period="6mo")

        # Fetch 1 year of daily data
        # Each candle = 1 trading day
        # Total candles: ~250-260
        data_1y = stock.history(period="1y")

        # ========================================
        # STEP 2: Check if We Got All Data
        # ========================================

        if len(data_1m) == 0 or len(data_6m) == 0 or len(data_1y) == 0:
            print("‚ö†Ô∏è Unable to fetch data for all time frames.")
            print("   Please check your internet connection and try again.")
        else:
            # Success! We have data for all three time frames
            print(f"‚úÖ 1 Month:  {len(data_1m)} trading days")
            print(f"‚úÖ 6 Months: {len(data_6m)} trading days")
            print(f"‚úÖ 1 Year:   {len(data_1y)} trading days")
            print(f"\nCreating multi-timeframe chart...\n")

            # ========================================
            # STEP 3: Create Subplot Layout
            # ========================================

            # make_subplots creates a figure with multiple chart panels
            # rows=3, cols=1 means 3 charts stacked vertically
            fig = make_subplots(
                rows=3,
                cols=1,  # 3 charts, one above the other
                # Titles for each subplot
                subplot_titles=(
                    "1 Month View (Short-term)",
                    "6 Months View (Medium-term)",
                    "1 Year View (Long-term)",
                ),
                # Space between subplots (8% of total height)
                vertical_spacing=0.08,
                # Heights of each row (all equal at 33%)
                row_heights=[0.33, 0.33, 0.34],
            )

            # ========================================
            # STEP 4: Add 1 Month Candlestick Chart
            # ========================================

            # Add candlestick chart to row 1, column 1
            fig.add_trace(
                go.Candlestick(
                    x=data_1m.index,  # Dates for X-axis
                    open=data_1m["Open"],  # Opening prices
                    high=data_1m["High"],  # High prices
                    low=data_1m["Low"],  # Low prices
                    close=data_1m["Close"],  # Closing prices
                    name="1M",  # Legend name
                ),
                row=1,
                col=1,  # Position: first row, first column
            )

            # ========================================
            # STEP 5: Add 6 Months Candlestick Chart
            # ========================================

            # Add candlestick chart to row 2, column 1
            fig.add_trace(
                go.Candlestick(
                    x=data_6m.index,
                    open=data_6m["Open"],
                    high=data_6m["High"],
                    low=data_6m["Low"],
                    close=data_6m["Close"],
                    name="6M",
                ),
                row=2,
                col=1,  # Position: second row, first column
            )

            # ========================================
            # STEP 6: Add 1 Year Candlestick Chart
            # ========================================

            # Add candlestick chart to row 3, column 1
            fig.add_trace(
                go.Candlestick(
                    x=data_1y.index,
                    open=data_1y["Open"],
                    high=data_1y["High"],
                    low=data_1y["Low"],
                    close=data_1y["Close"],
                    name="1Y",
                ),
                row=3,
                col=1,  # Position: third row, first column
            )

            # ========================================
            # STEP 7: Customize Overall Layout
            # ========================================

            fig.update_layout(
                title=f"{ticker} - Multiple Time Frame Analysis",
                height=1000,  # Total height in pixels (tall enough to see all 3)
                showlegend=False,  # Hide legend (not needed with titles)
                template="plotly_white",  # Clean white background
            )

            # Turn off range sliders for all subplots
            # (they take up space and we don't need them)
            fig.update_xaxes(rangeslider_visible=False)

            # ========================================
            # STEP 8: Display the Chart
            # ========================================

            fig.show()

            # ========================================
            # STEP 9: Analysis and Insights
            # ========================================

            print("‚úÖ Multiple time frame chart created!")

            print("\nüí° What to observe:")
            print("   üìä TOP CHART (1 Month):")
            print("      ‚Ä¢ Short-term view - recent price action")
            print("      ‚Ä¢ Notice daily fluctuations and noise")
            print("      ‚Ä¢ Good for: Entry/exit timing")

            print("\n   üìä MIDDLE CHART (6 Months):")
            print("      ‚Ä¢ Medium-term view - clearer trend")
            print("      ‚Ä¢ Less noise than 1-month chart")
            print("      ‚Ä¢ Good for: Swing trading, identifying trend")

            print("\n   üìä BOTTOM CHART (1 Year):")
            print("      ‚Ä¢ Long-term view - big picture")
            print("      ‚Ä¢ Major support/resistance levels visible")
            print("      ‚Ä¢ Good for: Overall trend direction")

            # ========================================
            # STEP 10: Calculate Trends for Each Period
            # ========================================

            print("\nüìà Trend Analysis:")

            # 1 Month trend
            change_1m = (
                (data_1m["Close"].iloc[-1] - data_1m["Close"].iloc[0])
                / data_1m["Close"].iloc[0]
                * 100
            )
            trend_1m = "UP üìà" if change_1m > 0 else "DOWN üìâ" if change_1m < 0 else "FLAT ‚û°Ô∏è"

            # 6 Months trend
            change_6m = (
                (data_6m["Close"].iloc[-1] - data_6m["Close"].iloc[0])
                / data_6m["Close"].iloc[0]
                * 100
            )
            trend_6m = "UP üìà" if change_6m > 0 else "DOWN üìâ" if change_6m < 0 else "FLAT ‚û°Ô∏è"

            # 1 Year trend
            change_1y = (
                (data_1y["Close"].iloc[-1] - data_1y["Close"].iloc[0])
                / data_1y["Close"].iloc[0]
                * 100
            )
            trend_1y = "UP üìà" if change_1y > 0 else "DOWN üìâ" if change_1y < 0 else "FLAT ‚û°Ô∏è"

            # Display trends
            print(f"   ‚Ä¢ 1 Month:  {trend_1m}  ({change_1m:+.1f}%)")
            print(f"   ‚Ä¢ 6 Months: {trend_6m}  ({change_6m:+.1f}%)")
            print(f"   ‚Ä¢ 1 Year:   {trend_1y}  ({change_1y:+.1f}%)")

            # ========================================
            # STEP 11: Trend Alignment Analysis
            # ========================================

            print("\nüéØ Trend Alignment:")

            # Check if all trends are in the same direction
            all_up = change_1m > 0 and change_6m > 0 and change_1y > 0
            all_down = change_1m < 0 and change_6m < 0 and change_1y < 0

            if all_up:
                print("   ‚úÖ ALL time frames show UPTREND")
                print("   ‚úÖ Strong bullish alignment - consider buying!")
                print("   ‚úÖ Short, medium, AND long-term traders are all buying")
            elif all_down:
                print("   ‚ö†Ô∏è  ALL time frames show DOWNTREND")
                print("   ‚ö†Ô∏è  Strong bearish alignment - avoid buying!")
                print("   ‚ö†Ô∏è  Short, medium, AND long-term traders are all selling")
            else:
                print("   ‚ö†Ô∏è  Time frames show DIFFERENT trends")
                print("   ‚ö†Ô∏è  Mixed signals - market is transitioning")
                print("   üí° Recommendation: Wait for clearer alignment")

                # Show which are up vs down
                if change_1y > 0:
                    print(f"   üìä Long-term: UP (follow this direction!)")
                else:
                    print(f"   üìä Long-term: DOWN (follow this direction!)")

            print("\nüéì Remember:")
            print("   ‚Ä¢ The LONGER time frame is the 'boss'")
            print("   ‚Ä¢ Never fight the 1-year trend!")
            print("   ‚Ä¢ Use shorter time frames for timing only")

    except Exception as e:
        print(f"‚ö†Ô∏è Error creating multi-timeframe chart: {e}")
        print("\nüîß Troubleshooting:")
        print("   1. Check internet connection")
        print("   2. Make sure ticker is valid")
        print("   3. Try running previous cells again")

# üìù What you should see:
# - Three candlestick charts stacked vertically
# - Top: 1 month (detailed, more candles close together)
# - Middle: 6 months (medium view)
# - Bottom: 1 year (big picture, candles further apart)
# - Trend analysis showing % change for each period
# - Alignment analysis (all up, all down, or mixed)
#
# ü§î How to interpret:
# - If all three trend same direction ‚Üí STRONG signal
# - If trends conflict ‚Üí Market is transitioning, be careful
# - Longer time frames override shorter ones
#
# üí° Pro tip:
# - Zoom into each chart to see details
# - Compare where current price is in each view
# - Look for support/resistance across all time frames
#
# ‚ö†Ô∏è Common observations:
# - 1-month might be down while 1-year is up (healthy pullback)
# - 1-year up + 6-month down = potential buying opportunity
# - All down = stay away or wait for reversal
````

## ‚ö†Ô∏è Common Mistakes to Avoid

Learn from these mistakes that trap 90% of beginners:

### 1. **Buying at Resistance** ‚ùå
- **Mistake**: "Stock looks good, I'll buy now!" (at RM 0.60 resistance)
- **Result**: Price immediately drops to RM 0.55
- **‚úÖ Correct**: Wait for breakout above resistance OR buy near support

### 2. **Ignoring Higher Time Frames** ‚ùå
- **Mistake**: "15-minute chart shows hammer, I'll buy!"
- **Reality**: Weekly chart shows strong downtrend
- **Result**: Small bounce then continued fall
- **‚úÖ Correct**: Check weekly FIRST, then daily, then hourly

### 3. **Trading Without Confirmation** ‚ùå
- **Mistake**: Buying on the pattern candle itself
- **Example**: See hammer, buy immediately
- **Result**: Next candle is red, pattern fails
- **‚úÖ Correct**: Wait for next green candle to confirm

### 4. **No Stop-Loss** ‚ùå
- **Mistake**: "I'll just hold if it goes down, it will come back"
- **Reality**: Stock continues falling, -50% loss
- **‚úÖ Correct**: ALWAYS set stop-loss below support before buying

### 5. **Forcing Trades** ‚ùå
- **Mistake**: "I must trade today!" (when there's no clear setup)
- **Result**: Enter at bad prices, lose money
- **‚úÖ Correct**: Wait for proper setups. Sometimes cash is a position!

### 6. **Chasing Breakouts** ‚ùå
- **Mistake**: Stock breaks resistance, buy at the peak
- **Result**: False breakout, price falls back down
- **‚úÖ Correct**: Wait for pullback to old resistance (now support)

### 7. **Overtrading** ‚ùå
- **Mistake**: Trading every day, multiple times
- **Result**: Fees eat profits, stressed out, poor decisions
- **‚úÖ Correct**: Quality > Quantity. Wait for A+ setups

### 8. **Ignoring Volume** ‚ùå
- **Mistake**: "Price is breaking out!" (on low volume)
- **Result**: Weak move, quickly reverses
- **‚úÖ Correct**: Breakouts need HIGH volume to be reliable

### Problem: Pattern Detection Shows Zero Patterns

**This is actually NORMAL!** It means:
- Strong trending market (no reversals)
- Patterns exist but don't meet our thresholds
- Time period too short

**Solutions:**
- Try longer time period: `period="6mo"` or `period="1y"`
- Adjust thresholds in pattern detection code (make them less strict)
- Manually look at charts - automated detection isn't perfect

---

### Problem: Support/Resistance Levels Seem Wrong

**Remember:**
- These are AUTOMATED calculations (simplified)
- Real S/R identification requires visual inspection
- Our code finds extremes, not necessarily best levels

**Solutions:**
- Use the interactive chart to visually identify levels
- Look for MULTIPLE bounces at same price
- Round numbers (RM 0.50, 1.00) are often better than calculated levels

---

### Problem: Can't Decide if Pattern is Valid

**Checklist:**
1. ‚úÖ Does it match the pattern definition? (body/shadow ratios)
2. ‚úÖ Is it at support (Hammer) or resistance (Shooting Star)?
3. ‚úÖ Was there confirmation next candle?
4. ‚úÖ Was volume above average?

**If any answer is NO, skip the trade!**

---

### Problem: Time Frames Show Conflicting Signals

**This is NORMAL during trend transitions!**

**What to do:**
1. **Follow the highest time frame** (weekly > daily > hourly)
2. **Wait for alignment** - don't force trades
3. **If weekly says UP but daily says DOWN**:
   - Healthy pullback in uptrend
   - Wait for daily to turn back up
   - Then buy at support
