````python
# Verify n8n is running

import subprocess
import urllib.request
import socket

def check_n8n_running(port=5678):
    """
    Check if n8n is accessible on the specified port.
    """
    print("=" * 60)
    print("N8N ACCESSIBILITY CHECK")
    print("=" * 60)
    
    # Check if port is listening
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(2)
    
    try:
        result = sock.connect_ex(('localhost', port))
        sock.close()
        
        if result == 0:
            print(f"âœ… Port {port} is OPEN")
            print(f"âœ… n8n appears to be running")
            print(f"\nðŸŒ Access n8n at: http://localhost:{port}")
            print("\nðŸ’¡ Next steps:")
            print("   1. Open your web browser")
            print(f"   2. Navigate to http://localhost:{port}")
            print("   3. Create your owner account")
            return True
        else:
            print(f"âŒ Port {port} is CLOSED")
            print(f"\nâš ï¸  n8n does not appear to be running")
            print("\nTroubleshooting:")
            print("   1. Verify you started n8n (npm: 'n8n' or Docker: 'docker ps')")
            print("   2. Check for error messages in the terminal")
            print(f"   3. Try a different port if {port} is in use")
            print("   4. Check Windows Firewall settings")
            return False
            
    except Exception as e:
        print(f"âŒ Error checking port: {e}")
        return False
    
    print("=" * 60)

# Run the check
check_n8n_running(5678)
````

## Step 4: Creating Your First Workflow

Let's create a simple workflow that:
1. Triggers manually
2. Makes an HTTP request to get random user data
3. Displays the result

- Click the HTTP Request node
- Set these parameters:
  - **Method**: GET

- Drag from the **Manual** node's output (small circle on right)
- Drop on the **HTTP Request** node's input (small circle on left)
- You should see a line connecting them

#### 5. Test the Workflow
- Click **"Execute Workflow"** button (top bar)
- Watch as the nodes execute (they'll turn green)
- Click the **HTTP Request** node to see the response data

#### 6. Save the Workflow
- Click the workflow name ("My workflow") at top
- Rename to: "My First Workflow"
- Press Enter to save

## Step 5: Understanding Workflow Execution

### How Workflows Execute

n8n workflows execute **node by node**, following connections:

1. **Trigger fires**: Manual click, schedule, webhook, etc.
2. **Data flows**: Each node receives data from previous node
3. **Processing**: Node performs its action (HTTP request, database query, etc.)
4. **Output**: Results passed to next connected node
5. **Completion**: Workflow ends when no more nodes to execute

### Execution Modes

#### Manual Execution
- Click "Execute Workflow" button
- Used for testing
- Shows results in real-time
- Workflow doesn't need to be "Active"

#### Production Execution
- Toggle "Active" switch ON
- Workflow runs automatically when triggered
- Results saved in "Executions" history
- Used for real automation

### Reading Execution Data

When you click a node after execution:
- **INPUT tab**: Data received from previous node
- **OUTPUT tab**: Data produced by this node
- **JSON view**: Raw data structure
- **Table view**: Formatted data table

## Step 6: Testing Webhooks with --tunnel Flag

### What is --tunnel?

The `--tunnel` flag creates a **temporary public URL** that forwards to your local n8n instance:

- Allows testing webhooks from external services
- No need to expose your computer to the internet
- URL looks like: `https://random-name.n8n.cloud`

### Starting n8n with --tunnel

**For npm installation:**

```bash
n8n start --tunnel
```

**For Docker:**

```powershell
docker run -d `
  --name n8n `
  -p 5678:5678 `
  -v n8n_data:/home/node/.n8n `
  docker.n8n.io/n8nio/n8n `
  start --tunnel
```

### When to Use --tunnel

âœ… **Good for:**
- Testing webhook workflows during development
- Trying out integrations before production
- Learning how webhooks work

âŒ **NOT for:**
- Production deployments (unreliable)
- Long-term use (tunnel can disconnect)
- Sensitive data (traffic routes through n8n's servers)

### Important Warnings

- Tunnel URL changes each time you restart
- Connection can be unstable
- Data passes through n8n cloud (privacy concern)
- Only for development/testing

## Step 7: Configuring Basic Authentication

If you didn't set up authentication during installation, you should do it now.

### Why Authentication Matters

Without authentication:
- Anyone on your network can access n8n
- They can view/edit/delete workflows
- They can access stored credentials
- They can execute workflows consuming resources

**Never run n8n without authentication if it's network-accessible.**

### Setting Up Basic Auth (npm)

Stop n8n (Ctrl+C), then restart with environment variables:

**PowerShell:**
```powershell
$env:N8N_BASIC_AUTH_ACTIVE="true"
$env:N8N_BASIC_AUTH_USER="admin"
$env:N8N_BASIC_AUTH_PASSWORD="YourStrongPassword123!"
n8n
```

**Command Prompt:**
```batch
set N8N_BASIC_AUTH_ACTIVE=true
set N8N_BASIC_AUTH_USER=admin
set N8N_BASIC_AUTH_PASSWORD=YourStrongPassword123!
n8n
```

### Setting Up Basic Auth (Docker Compose)

Edit your `docker-compose.yml` and add:

```yaml
environment:
  - N8N_BASIC_AUTH_ACTIVE=true
  - N8N_BASIC_AUTH_USER=admin
  - N8N_BASIC_AUTH_PASSWORD=YourStrongPassword123!
```

Then restart:
```powershell
docker compose restart
```

### Testing Authentication

1. Access http://localhost:5678
2. Browser should show login prompt
3. Enter your username and password
4. Should see n8n interface

**If you can access without login, authentication is NOT working.**
