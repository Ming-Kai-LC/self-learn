## Introduction: Why Single Timeframe Trading Fails

**Most losing traders focus on a single timeframe** - usually the daily chart. This is a critical mistake.

### The Problem with Single Timeframe

Imagine you see a bullish signal on the daily chart:
- âœ… Golden cross (SMA 50 crosses above SMA 200)
- âœ… RSI turning up from oversold
- âœ… Volume increasing

You enter a long position... and it immediately fails!

**Why?** You didn't check the weekly chart, which shows:
- âŒ Strong downtrend for months
- âŒ Price at major resistance
- âŒ Death cross on weekly timeframe

**You were trading against the higher timeframe trend!**

### The Power of Multi-timeframe Analysis

**Multi-timeframe analysis** solves this problem:
1. **Higher timeframes show the big picture** (Monthly, Weekly)
2. **Lower timeframes show entry/exit precision** (Daily, Hourly)
3. **Alignment increases win rate** by 20-30%

### Malaysian Market Context

For **Bursa Malaysia swing trading** (3-15 day holds):
- **Monthly chart**: Determine bull or bear market
- **Weekly chart**: Find intermediate trends and major levels
- **Daily chart**: Identify entry and exit points

**Golden Rule**: **Trade in the direction of the higher timeframe trend!**

## 1. Timeframe Hierarchy

### Understanding Timeframe Relationships

Think of timeframes like a **telescope with different zoom levels**:

| Timeframe | Zoom Level | Shows | Typical Traders | Malaysian Market |
|-----------|------------|-------|----------------|------------------|
| **Monthly** | ðŸ”­ Wide angle | Long-term trend (years) | Investors | Bull/bear market |
| **Weekly** | ðŸ” Medium zoom | Intermediate trend (months) | Position traders | Sector rotation |
| **Daily** | ðŸ“Š Close-up | Short-term trend (weeks) | Swing traders | 3-15 day holds |
| **Hourly** | ðŸ”¬ Micro | Intraday moves (hours) | Day traders | Entry/exit timing |

### Timeframe Rules

1. **Higher timeframe = Higher authority**
   - Monthly trend > Weekly trend > Daily trend
   
2. **Use 3-4 timeframes**
   - Too few: Miss the big picture
   - Too many: Analysis paralysis
   
3. **Rule of 4-6x**
   - Each timeframe should be 4-6x the previous
   - Example: Daily (1 day) â†’ Weekly (5 days) â†’ Monthly (20 days)

### Top-Down Analysis Process

**Always analyze from higher to lower timeframes:**

```
STEP 1: Monthly Chart  â†’ Identify long-term trend (bull/bear/sideways)
          â†“
STEP 2: Weekly Chart   â†’ Find intermediate trend and major S/R levels
          â†“
STEP 3: Daily Chart    â†’ Locate entry setups (MA crosses, RSI, etc.)
          â†“
STEP 4: Execute Trade  â†’ Only if all timeframes align!
```

Let's download data for multiple timeframes!

````python
# Download data for multiple timeframes
# We'll use Public Bank (1295.KL) - stable banking stock

ticker = '1295.KL'

# Download extended history for monthly/weekly analysis
start_date = '2020-01-01'
end_date = '2024-12-31'

print(f"Downloading {ticker} (Public Bank) data for multi-timeframe analysis...\n")

# Daily data
daily_data = yf.download(ticker, start=start_date, end=end_date, interval='1d', progress=False)

# Weekly data (yfinance: '1wk')
weekly_data = yf.download(ticker, start=start_date, end=end_date, interval='1wk', progress=False)

# Monthly data (yfinance: '1mo')
monthly_data = yf.download(ticker, start=start_date, end=end_date, interval='1mo', progress=False)

print("âœ… Data Downloaded:")
print(f"   Daily: {len(daily_data)} bars")
print(f"   Weekly: {len(weekly_data)} bars")
print(f"   Monthly: {len(monthly_data)} bars")
print(f"\nDate range: {daily_data.index[0].strftime('%Y-%m-%d')} to {daily_data.index[-1].strftime('%Y-%m-%d')}")
````

## 2. Monthly Chart Analysis (Long-term Trend)

### Purpose of Monthly Charts

The **monthly chart** answers the most important question:

**"Are we in a bull market or bear market?"**

### What to Look For

1. **Trend Direction**:
   - Uptrend: Higher highs and higher lows
   - Downtrend: Lower highs and lower lows
   - Sideways: Ranging between support and resistance

2. **Major Support/Resistance**:
   - Multi-year highs and lows
   - Psychological levels (round numbers)

3. **Moving Averages**:
   - SMA 12 (1-year)
   - SMA 24 (2-year)
   
### Trading Rules Based on Monthly Trend

| Monthly Trend | Trading Approach |
|---------------|------------------|
| **Bull Market** | âœ… Look for pullback entries to go LONG |
| **Bear Market** | âŒ Avoid longs, consider shorts or stay out |
| **Sideways** | ðŸ“Š Range trading: Buy support, sell resistance |

Let's analyze the monthly chart!

````python
# Calculate monthly moving averages
monthly_data['SMA_12'] = monthly_data['Close'].rolling(window=12).mean()
monthly_data['SMA_24'] = monthly_data['Close'].rolling(window=24).mean()

# Determine monthly trend
latest_close = monthly_data['Close'].iloc[-1]
sma12 = monthly_data['SMA_12'].iloc[-1]
sma24 = monthly_data['SMA_24'].iloc[-1]

if latest_close > sma12 > sma24:
    monthly_trend = "ðŸŸ¢ BULL MARKET"
elif latest_close < sma12 < sma24:
    monthly_trend = "ðŸ”´ BEAR MARKET"
else:
    monthly_trend = "âšª SIDEWAYS MARKET"

# Visualize monthly chart
plt.figure(figsize=(16, 8))
plt.plot(monthly_data.index, monthly_data['Close'], linewidth=3, 
         label='Monthly Close', color='black', marker='o', markersize=5)
plt.plot(monthly_data.index, monthly_data['SMA_12'], linewidth=2.5, 
         label='SMA 12 (1-year)', color='blue', alpha=0.8)
plt.plot(monthly_data.index, monthly_data['SMA_24'], linewidth=2.5, 
         label='SMA 24 (2-year)', color='red', alpha=0.8)

plt.title(f'Public Bank (1295.KL) - MONTHLY Chart\nCurrent Trend: {monthly_trend}', 
          fontsize=18, fontweight='bold')
plt.xlabel('Date', fontsize=13)
plt.ylabel('Price (RM)', fontsize=13)
plt.legend(loc='best', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nðŸ“Š MONTHLY TIMEFRAME ANALYSIS:")
print("=" * 60)
print(f"Current Trend: {monthly_trend}")
print(f"\nCurrent Price: RM{latest_close:.2f}")
print(f"SMA 12 (1-year): RM{sma12:.2f}")
print(f"SMA 24 (2-year): RM{sma24:.2f}")
print(f"\nðŸ’¡ Trading Implication:")
if "BULL" in monthly_trend:
    print("   âœ… Favor LONG positions on pullbacks")
    print("   âœ… Buy when price dips to support")
elif "BEAR" in monthly_trend:
    print("   âŒ Avoid LONG positions")
    print("   âš ï¸  Consider staying in cash or short-selling")
else:
    print("   ðŸ“Š Range trading: Buy support, sell resistance")
    print("   âš ï¸  Avoid breakout trades (likely false breakouts)")
````

## 3. Weekly Chart Analysis (Intermediate Trend)

### Purpose of Weekly Charts

The **weekly chart** shows:
- Intermediate trends (lasting weeks to months)
- Major support and resistance zones
- Better entry timing than monthly

### What to Look For

1. **Intermediate Trend**:
   - SMA 10 (10 weeks â‰ˆ 2.5 months)
   - SMA 20 (20 weeks â‰ˆ 5 months)

2. **Major Support/Resistance**:
   - Previous swing highs/lows
   - Moving average levels

3. **Alignment with Monthly**:
   - Weekly uptrend + Monthly uptrend = âœ… Strong alignment
   - Weekly downtrend + Monthly uptrend = âš ï¸ Correction (pullback)
   - Weekly downtrend + Monthly downtrend = âŒ Strong downtrend

Let's analyze the weekly chart!

````python
# Calculate weekly moving averages
weekly_data['SMA_10'] = weekly_data['Close'].rolling(window=10).mean()
weekly_data['SMA_20'] = weekly_data['Close'].rolling(window=20).mean()

# Determine weekly trend
latest_close_w = weekly_data['Close'].iloc[-1]
sma10_w = weekly_data['SMA_10'].iloc[-1]
sma20_w = weekly_data['SMA_20'].iloc[-1]

if latest_close_w > sma10_w > sma20_w:
    weekly_trend = "ðŸŸ¢ UPTREND"
elif latest_close_w < sma10_w < sma20_w:
    weekly_trend = "ðŸ”´ DOWNTREND"
else:
    weekly_trend = "âšª SIDEWAYS"

# Visualize weekly chart
plt.figure(figsize=(16, 8))
plt.plot(weekly_data.index, weekly_data['Close'], linewidth=2.5, 
         label='Weekly Close', color='black', marker='o', markersize=4)
plt.plot(weekly_data.index, weekly_data['SMA_10'], linewidth=2, 
         label='SMA 10 (10 weeks)', color='blue', alpha=0.8)
plt.plot(weekly_data.index, weekly_data['SMA_20'], linewidth=2, 
         label='SMA 20 (20 weeks)', color='red', alpha=0.8)

plt.title(f'Public Bank (1295.KL) - WEEKLY Chart\nCurrent Trend: {weekly_trend}', 
          fontsize=18, fontweight='bold')
plt.xlabel('Date', fontsize=13)
plt.ylabel('Price (RM)', fontsize=13)
plt.legend(loc='best', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nðŸ“Š WEEKLY TIMEFRAME ANALYSIS:")
print("=" * 60)
print(f"Current Trend: {weekly_trend}")
print(f"\nCurrent Price: RM{latest_close_w:.2f}")
print(f"SMA 10 (10 weeks): RM{sma10_w:.2f}")
print(f"SMA 20 (20 weeks): RM{sma20_w:.2f}")

# Check alignment with monthly
print(f"\nðŸ” TIMEFRAME ALIGNMENT CHECK:")
print(f"   Monthly: {monthly_trend}")
print(f"   Weekly:  {weekly_trend}")

if "BULL" in monthly_trend and "UPTREND" in weekly_trend:
    alignment = "âœ… STRONG BULLISH ALIGNMENT"
    print(f"\n   {alignment}")
    print("   ðŸ’¡ Excellent conditions for LONG entries!")
elif "BEAR" in monthly_trend and "DOWNTREND" in weekly_trend:
    alignment = "âŒ STRONG BEARISH ALIGNMENT"
    print(f"\n   {alignment}")
    print("   âš ï¸  Avoid LONG positions - strong downtrend")
else:
    alignment = "âš ï¸  MIXED SIGNALS"
    print(f"\n   {alignment}")
    print("   ðŸ“Š Wait for clearer alignment or trade more cautiously")
````

## 4. Daily Chart Analysis (Entry/Exit Precision)

### Purpose of Daily Charts

The **daily chart** is where you:
- Find specific entry setups
- Identify exact entry and exit points
- Apply technical indicators (RSI, MACD)

### Trading Logic

**Only trade daily signals that align with higher timeframes!**

Example:
- Monthly: Bull market âœ…
- Weekly: Uptrend âœ…
- Daily: Pullback to SMA 20, RSI oversold âœ…
- **Action**: BUY (high-probability setup)

vs.

- Monthly: Bear market âŒ
- Weekly: Downtrend âŒ
- Daily: Bullish pattern âš ï¸
- **Action**: DO NOT BUY (counter-trend, low probability)

Let's analyze the daily chart!

````python
# Calculate daily moving averages and indicators
daily_data['SMA_20'] = daily_data['Close'].rolling(window=20).mean()
daily_data['SMA_50'] = daily_data['Close'].rolling(window=50).mean()

# Calculate RSI
def calculate_rsi(data, period=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

daily_data['RSI'] = calculate_rsi(daily_data)

# Determine daily trend
latest_close_d = daily_data['Close'].iloc[-1]
sma20_d = daily_data['SMA_20'].iloc[-1]
sma50_d = daily_data['SMA_50'].iloc[-1]
rsi_d = daily_data['RSI'].iloc[-1]

if latest_close_d > sma20_d > sma50_d:
    daily_trend = "ðŸŸ¢ UPTREND"
elif latest_close_d < sma20_d < sma50_d:
    daily_trend = "ðŸ”´ DOWNTREND"
else:
    daily_trend = "âšª SIDEWAYS"

# Visualize daily chart with indicators
fig, axes = plt.subplots(2, 1, figsize=(16, 12), sharex=True, 
                         gridspec_kw={'height_ratios': [3, 1]})

# Top: Price and MAs
axes[0].plot(daily_data.index, daily_data['Close'], linewidth=2, 
             label='Daily Close', color='black', alpha=0.8)
axes[0].plot(daily_data.index, daily_data['SMA_20'], linewidth=2, 
             label='SMA 20', color='blue', alpha=0.7)
axes[0].plot(daily_data.index, daily_data['SMA_50'], linewidth=2, 
             label='SMA 50', color='red', alpha=0.7)
axes[0].set_title(f'Public Bank (1295.KL) - DAILY Chart\nCurrent Trend: {daily_trend}', 
                  fontsize=18, fontweight='bold')
axes[0].set_ylabel('Price (RM)', fontsize=12)
axes[0].legend(loc='best', fontsize=11)
axes[0].grid(True, alpha=0.3)

# Bottom: RSI
axes[1].plot(daily_data.index, daily_data['RSI'], linewidth=2, 
             label='RSI (14)', color='purple', alpha=0.8)
axes[1].axhline(y=70, color='red', linestyle='--', alpha=0.5, label='Overbought (70)')
axes[1].axhline(y=30, color='green', linestyle='--', alpha=0.5, label='Oversold (30)')
axes[1].set_xlabel('Date', fontsize=12)
axes[1].set_ylabel('RSI', fontsize=12)
axes[1].legend(loc='best', fontsize=10)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(0, 100)

plt.tight_layout()
plt.show()

print(f"\nðŸ“Š DAILY TIMEFRAME ANALYSIS:")
print("=" * 60)
print(f"Current Trend: {daily_trend}")
print(f"\nCurrent Price: RM{latest_close_d:.2f}")
print(f"SMA 20: RM{sma20_d:.2f}")
print(f"SMA 50: RM{sma50_d:.2f}")
print(f"RSI: {rsi_d:.1f}")

# RSI interpretation
if rsi_d > 70:
    rsi_status = "ðŸ”´ OVERBOUGHT - Potential pullback"
elif rsi_d < 30:
    rsi_status = "ðŸŸ¢ OVERSOLD - Potential bounce"
else:
    rsi_status = "âšª NEUTRAL"
print(f"RSI Status: {rsi_status}")
````

## 5. Complete Multi-timeframe Decision Matrix

### The Trading Decision Process

Now let's combine all three timeframes to make a trading decision:

### High-Probability Trade Setups

| Monthly | Weekly | Daily | Action | Probability |
|---------|--------|-------|--------|-------------|
| ðŸŸ¢ Bull | ðŸŸ¢ Up | ðŸŸ¢ Up | **STRONG BUY** | Very High |
| ðŸŸ¢ Bull | ðŸŸ¢ Up | ðŸ”´ Down | **BUY pullback** | High |
| ðŸŸ¢ Bull | âšª Side | ðŸŸ¢ Up | **CAUTIOUS BUY** | Medium |
| ðŸ”´ Bear | ðŸ”´ Down | ðŸ”´ Down | **STRONG SELL/AVOID** | Very High |
| ðŸ”´ Bear | ðŸ”´ Down | ðŸŸ¢ Up | **DO NOT BUY** | Low |
| âšª Side | âšª Side | ðŸŸ¢ Up | **Range trading** | Medium |

### Malaysian Swing Trading Strategy

For **3-15 day swing trades** in Bursa Malaysia:

**ENTRY CRITERIA** (All must be true):
1. âœ… Monthly: Bull market OR sideways with support holding
2. âœ… Weekly: Uptrend OR pullback in uptrend
3. âœ… Daily: Bullish setup (MA cross, RSI oversold bounce, chart pattern)
4. âœ… Volume: Above average on breakout/entry day

**EXIT CRITERIA** (Any one is true):
1. âŒ Daily: Closes below SMA 20
2. âŒ RSI: Exceeds 70 (overbought)
3. âŒ Target: Reaches resistance level
4. âŒ Time: Holding period exceeds 15 days without progress

Let's create a function to evaluate multi-timeframe alignment!

````python
def analyze_multitimeframe(monthly_trend, weekly_trend, daily_trend, rsi):
    """
    Analyze multi-timeframe alignment and provide trading recommendation.
    
    Args:
        monthly_trend: Monthly trend status
        weekly_trend: Weekly trend status
        daily_trend: Daily trend status
        rsi: Current RSI value
    
    Returns:
        Dictionary with recommendation and reasoning
    """
    
    # Bullish alignment
    if "BULL" in monthly_trend and "UP" in weekly_trend and "UP" in daily_trend:
        return {
            'action': 'ðŸŸ¢ STRONG BUY SIGNAL',
            'confidence': '95%',
            'reasoning': 'All timeframes aligned bullish - highest probability setup',
            'strategy': 'Enter on any minor pullback or continuation pattern'
        }
    
    # Pullback buy opportunity
    elif "BULL" in monthly_trend and "UP" in weekly_trend and "DOWN" in daily_trend and rsi < 40:
        return {
            'action': 'ðŸŸ¢ BUY PULLBACK',
            'confidence': '80%',
            'reasoning': 'Healthy pullback in uptrend with oversold RSI',
            'strategy': 'Enter when RSI turns up from oversold and price bounces from SMA 20/50'
        }
    
    # Bearish alignment
    elif "BEAR" in monthly_trend and "DOWN" in weekly_trend:
        return {
            'action': 'ðŸ”´ AVOID LONG POSITIONS',
            'confidence': '90%',
            'reasoning': 'Higher timeframes bearish - strong downtrend',
            'strategy': 'Stay in cash or consider short positions (if experienced)'
        }
    
    # Counter-trend signal
    elif ("BEAR" in monthly_trend or "DOWN" in weekly_trend) and "UP" in daily_trend:
        return {
            'action': 'âš ï¸  CAUTION - COUNTER-TREND',
            'confidence': '30%',
            'reasoning': 'Daily bullish but higher timeframes bearish - likely false signal',
            'strategy': 'Skip this trade - trading against higher timeframe trend'
        }
    
    # Mixed signals
    else:
        return {
            'action': 'âšª NEUTRAL - WAIT',
            'confidence': '50%',
            'reasoning': 'Mixed signals across timeframes - no clear direction',
            'strategy': 'Wait for clearer alignment or use smaller position size'
        }

# Get trading recommendation
recommendation = analyze_multitimeframe(monthly_trend, weekly_trend, daily_trend, rsi_d)

print("\n" + "=" * 80)
print("ðŸ“Š MULTI-TIMEFRAME TRADING RECOMMENDATION")
print("=" * 80)
print(f"\nðŸ” TIMEFRAME STATUS:")
print(f"   Monthly: {monthly_trend}")
print(f"   Weekly:  {weekly_trend}")
print(f"   Daily:   {daily_trend}")
print(f"   RSI:     {rsi_d:.1f}")
print(f"\n{recommendation['action']}")
print(f"Confidence Level: {recommendation['confidence']}")
print(f"\nReasoning: {recommendation['reasoning']}")
print(f"\nStrategy: {recommendation['strategy']}")
print("\n" + "=" * 80)
````
