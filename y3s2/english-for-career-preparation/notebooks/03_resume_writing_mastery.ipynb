## 1. Setup and Imports

````python
# Import required libraries
import warnings
warnings.filterwarnings('ignore')

from gtts import gTTS
import IPython.display as ipd
from pathlib import Path
import pandas as pd

def generate_audio(text, filename='audio', slow=False, lang='en'):
    """
    Generate audio from text for pronunciation practice.
    """
    audio_dir = Path('../audio')
    audio_dir.mkdir(exist_ok=True)
    
    tts = gTTS(text=text, lang=lang, slow=slow)
    output_path = audio_dir / f"{filename}.mp3"
    tts.save(str(output_path))
    
    return ipd.Audio(str(output_path), autoplay=False)

print("âœ… Setup complete! Ready to master resume writing.")
````

#### The CAR Method (Context-Action-Result)

Every bullet point should tell a mini-story:

1. **Context**: What was the situation/challenge?
2. **Action**: What did YOU do?
3. **Result**: What was the measurable outcome?

#### Bad vs Good Examples:

| âŒ Weak (Duty-focused) | âœ… Strong (Achievement-focused) |

### Power Action Verbs by Category

Start every bullet point with a **strong action verb**:

#### Leadership:
- Led, Directed, Managed, Supervised, Coordinated, Spearheaded, Orchestrated

#### Achievement:
- Achieved, Exceeded, Surpassed, Accomplished, Attained, Delivered, Completed

#### Improvement:
- Improved, Enhanced, Optimized, Streamlined, Strengthened, Upgraded, Revitalized

#### Creation:
- Developed, Created, Designed, Built, Established, Implemented, Launched

#### Analysis:
- Analyzed, Evaluated, Assessed, Investigated, Researched, Identified, Diagnosed

#### Communication:
- Presented, Communicated, Negotiated, Collaborated, Facilitated, Consulted

#### Technical:

- Percentages: "Increased sales by 30%"

## 3. ATS Optimization: Getting Past the Robots

**75% of resumes are rejected by ATS** (Applicant Tracking Systems) before a human ever sees them!

### What is ATS?

ATS is software that:
- Scans resumes for keywords from the job description
- Ranks candidates by keyword match percentage
- Filters out resumes with poor formatting
- Presents only top candidates to human reviewers

### ATS-Friendly Formatting Rules

âœ… **Do:**
- Use standard section headings: "Work Experience", "Education", "Skills"
- Use simple, clean fonts: Arial, Calibri, Georgia, Times New Roman (10-12pt)
- Save as .docx or .pdf (check job posting for preferred format)
- Use standard bullet points (â€¢)
- Left-align all text
- Use single-column layout
- Include phone number with country code (+60)

âŒ **Don't:**
- Use tables, text boxes, or columns
- Add headers and footers (ATS may not read them)
- Use images, graphics, or logos
- Use fancy fonts or colors
- Use creative section names ("My Journey" instead of "Work Experience")
- Submit as .jpg, .png, or .pages files

### Keyword Optimization Strategy

#### Step 1: Analyze the Job Description

Look for:
- **Required skills**: "Must have experience with Python"
- **Required qualifications**: "Bachelor's degree in Computer Science"
- **Action words**: "Develop", "Manage", "Lead"
- **Industry terms**: "Agile methodology", "Full-stack development"

#### Step 2: Match Keywords Naturally

**Job Description Says:**
```
"We are seeking a Software Developer with experience in Python, 
React, and PostgreSQL. The ideal candidate will develop RESTful 
APIs and work in an Agile environment."
```

**Your Resume Should Include:**
```
â€¢ Developed RESTful APIs using Python and Flask, serving 10,000+ requests daily
â€¢ Built responsive web applications using React and Redux
â€¢ Designed and optimized PostgreSQL databases with 1M+ records
â€¢ Collaborated with cross-functional teams using Agile methodology
```

âš ï¸ **Warning**: Don't just copy-paste keywords! Integrate them naturally into your achievements.

````python
# ATS keyword matching example
print("ğŸ“Š ATS KEYWORD MATCHING DEMONSTRATION")
print("="*70)

job_description = """
We are seeking a Data Analyst with experience in SQL, Python, and 
data visualization. The ideal candidate will analyze large datasets, 
create dashboards, and provide actionable insights to stakeholders.
"""

# Extract key terms
keywords = [
    "Data Analyst",
    "SQL",
    "Python",
    "data visualization",
    "analyze",
    "datasets",
    "dashboards",
    "insights",
    "stakeholders"
]

print("JOB DESCRIPTION:")
print(job_description)

print("\nKEY TERMS TO INCLUDE IN RESUME:")
print("-" * 70)
for i, keyword in enumerate(keywords, 1):
    print(f"{i}. {keyword}")

print("\n\nHOW TO USE KEYWORDS NATURALLY:")
print("-" * 70)

examples = [
    "â€¢ Analyzed customer behavior datasets using Python and SQL, identifying 3 key trends that increased retention by 15%",
    "â€¢ Created interactive dashboards in Tableau visualizing sales data for C-level stakeholders",
    "â€¢ Provided actionable insights from data visualization reports, leading to RM 500K cost savings"
]

for example in examples:
    print(f"\n{example}")

print("\n" + "="*70)
print("ğŸ’¡ Notice how keywords are integrated naturally with achievements!")
print("="*70)
````

## 4. Tailoring Your Resume for Different Jobs

**Never use the same resume for every job!** Tailor each resume to match the specific job.

### The 3-Step Tailoring Process

#### Step 1: Create a Master Resume

Include EVERYTHING:
- All jobs and internships
- All projects
- All skills
- All achievements

This is your "database" to pull from.

#### Step 2: Analyze Each Job Posting

Highlight:
- Required skills
- Preferred skills
- Key responsibilities
- Company values/culture

#### Step 3: Customize Your Resume

- **Professional Summary**: Rewrite to match job requirements
- **Work Experience**: Emphasize relevant achievements, de-emphasize irrelevant ones
- **Skills**: Reorder to prioritize job-required skills
- **Projects**: Include projects using required technologies

## 5. Common Resume Mistakes That Lead to Rejection

### Mistake 1: Spelling and Grammar Errors

**Impact**: 58% of resumes with typos are immediately rejected.

âŒ **Examples:**
- "Experiance in Python programming"
- "Recieved Employee of the Month award"
- "Worked closely with client's to deliver projects" (incorrect apostrophe)

âœ… **Solution:**
- Use spell-check (Grammarly, MS Word)
- Read resume aloud to catch errors
- Ask someone else to proofread
- Wait 24 hours and proofread again with fresh eyes

### Mistake 2: Including Irrelevant Information

âŒ **Don't Include:**
- "Hobbies: Reading, watching movies, playing games"
- Irrelevant work experience (unless you have nothing else)
- "References available upon request" (this is assumed)
- Personal information: age, marital status, religion, photo (unless specifically requested)

### Mistake 3: Using Passive Language

| âŒ Weak (Passive) | âœ… Strong (Active) |
|-----------------|------------------|
| "Was responsible for managing team" | "Managed team of 8 developers" |
| "Involved in website redesign" | "Redesigned company website using React" |
| "Helped to improve sales" | "Increased sales by 25% through targeted campaigns" |
| "Assisted with customer service" | "Resolved 50+ customer inquiries daily with 98% satisfaction" |

### Mistake 4: Resume Too Long or Too Short

**Ideal Length:**
- **Fresh Graduates**: 1 page (maximum)
- **2-10 years experience**: 1-2 pages
- **10+ years experience**: 2 pages (maximum)

### Mistake 5: Poor Formatting and Visual Hierarchy

âŒ **Bad:**
- Everything same font size
- No white space, text cramped
- Inconsistent bullet points
- Random bold and italics

âœ… **Good:**
- Name largest (18-24pt)
- Section headers bold and slightly larger (12-14pt)
- Body text consistent (10-11pt)
- Consistent spacing between sections
- Margins: 0.5-1 inch on all sides

### Mistake 6: Listing Duties Instead of Achievements

âŒ **Duty-focused (Weak):**
```
â€¢ Managed social media accounts
â€¢ Posted content regularly
â€¢ Responded to customer messages
```

âœ… **Achievement-focused (Strong):**
```

â€¢ Created 100+ social media posts that generated 500K+ impressions and 15K+ engagements
```

## 8. Summary and Key Takeaways

### Resume Structure Essentials

1. **Contact Information**: Professional email, phone, LinkedIn, location
2. **Professional Summary**: 2-3 sentences highlighting value proposition
3. **Work Experience**: Achievement-focused with metrics (CAR method)
4. **Education**: Degree, university, GPA (if 3.5+), relevant coursework
5. **Skills**: Categorized and tailored to job requirements
6. **Projects**: Essential for students/developers to show practical skills

### The CAR Method for Achievements

- **Context**: What was the situation?
- **Action**: What did you do?
- **Result**: What measurable outcome did you achieve?

### ATS Optimization Checklist

- [ ] Use standard section headings
- [ ] Include keywords from job description
- [ ] Use simple formatting (no tables, columns, text boxes)
- [ ] Save as .docx or .pdf (check job posting)
- [ ] Use standard fonts (Arial, Calibri, Georgia)
- [ ] Avoid images, graphics, headers, footers

### Common Mistakes to Avoid

- âŒ Spelling and grammar errors
- âŒ Unprofessional email address
- âŒ Duty-focused instead of achievement-focused
- âŒ Missing metrics and numbers
- âŒ Using same resume for every job
- âŒ Too long or too short
- âŒ Poor formatting and visual hierarchy

### Final Resume Checklist

Before submitting your resume:

- [ ] Proofread 3+ times (use Grammarly, ask someone else)
- [ ] All bullet points start with strong action verbs
- [ ] Included metrics and quantifiable results
- [ ] Tailored to specific job posting
- [ ] Matched keywords from job description
- [ ] Consistent formatting throughout
- [ ] Contact information is accurate
- [ ] Saved with professional filename: FirstName_LastName_Resume.pdf
- [ ] File size under 2MB
- [ ] Correct length (1 page for fresh grad, 1-2 for experienced)
