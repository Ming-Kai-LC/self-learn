## 1. Setup and Imports

````python
# Import required libraries
import warnings
warnings.filterwarnings('ignore')

from gtts import gTTS
import IPython.display as ipd
from pathlib import Path
import pandas as pd

def generate_audio(text, filename='audio', slow=False, lang='en'):
    """
    Generate audio from text for pronunciation practice.
    """
    audio_dir = Path('../audio')
    audio_dir.mkdir(exist_ok=True)
    
    tts = gTTS(text=text, lang=lang, slow=slow)
    output_path = audio_dir / f"{filename}.mp3"
    tts.save(str(output_path))
    
    return ipd.Audio(str(output_path), autoplay=False)

print("âœ… Setup complete! Ready to build your professional online presence.")
````

#### Examples: Weak vs Strong Headlines

````python
# LinkedIn Headline Examples
headline_examples = pd.DataFrame({
    'Career Stage': [
        'Fresh Graduate',
        'Fresh Graduate',
        'Experienced Professional',
        'Experienced Professional',
        'Career Changer',
        'Career Changer'
    ],
    'Type': ['Weak', 'Strong', 'Weak', 'Strong', 'Weak', 'Strong'],
    'Headline': [
        'Recent Graduate',
        'Computer Science Graduate | Python & Web Development | Seeking Software Developer Role',
        'Marketing Manager',
        'Digital Marketing Manager | SEO & Social Media Specialist | Growing Brands Through Data-Driven Strategies',
        'Looking for opportunities in UX Design',
        'Transitioning to UX Design | Former Educator | Creating User-Centered Digital Experiences'
    ]
})

print("ğŸ“Š LINKEDIN HEADLINE EXAMPLES: Weak vs Strong")
print("=" * 140)
for stage in headline_examples['Career Stage'].unique():
    subset = headline_examples[headline_examples['Career Stage'] == stage]
    print(f"\n{stage.upper()}:")
    print("-" * 140)
    for _, row in subset.iterrows():
        symbol = "âŒ" if row['Type'] == 'Weak' else "âœ…"
        print(f"{symbol} {row['Type']:6}: {row['Headline']}")

print("\n" + "=" * 140)
print("ğŸ’¡ Notice: Strong headlines include KEYWORDS + VALUE PROPOSITION + SPECIFICITY")
print("=" * 140)
````

### Section 3: About / Summary (Your Professional Story)

**Character Limit**: 2,600 characters (about 400 words)

**Purpose**: Tell your professional story in a way that connects with readers

#### Structure: The 3-Paragraph Formula

**Paragraph 1: WHO YOU ARE (Present)**
- Your current role and expertise
- What you're passionate about
- Your unique value proposition

**Paragraph 2: HOW YOU GOT HERE (Past)**
- Your background and journey
- Key experiences that shaped you
- Notable achievements (with numbers!)

**Paragraph 3: WHERE YOU'RE GOING (Future)**
- What you're looking for
- How others can work with you
- Call to action ("Let's connect!")

#### Tone Guidelines

âœ… **DO:**
- Write in first person ("I", not "he/she")
- Be conversational but professional
- Show personality (it's okay to be human!)
- Include specific achievements with metrics
- Use short paragraphs (2-4 sentences each)
- Add strategic line breaks for readability

âŒ **DON'T:**
- Copy your resume word-for-word
- Use jargon that outsiders won't understand
- Write one giant paragraph
- Be overly formal or stiff
- Focus only on duties (focus on impact!)

## 5. What NOT to Post on LinkedIn

LinkedIn is a professional platform. Avoid these mistakes:

### âŒ Don't Post:

**1. Political Rants**
- Keep political opinions off LinkedIn
- Focus on professional topics
- Exception: If politics IS your profession

**2. Complaints About Your Job/Boss/Company**
- Never badmouth current or former employers
- Recruiters WILL see this
- Makes you look unprofessional

**3. Controversial Religious Content**
- Keep religious views personal
- Could alienate potential employers/clients

**4. Party/Vacation Photos**
- This isn't Instagram or Facebook
- Professional content only
- Exception: Work conference photos are fine

**5. Sales Pitches in Connection Messages**
- Don't immediately pitch your product/service
- Build relationship first
- Nobody likes spam

**6. Inappropriate Humor**
- Keep it professional
- Avoid off-color jokes
- What's funny to friends might offend professionals

**7. Personal Drama**
- Relationship issues
- Family problems
- Health details (unless relevant to your professional story)

**8. Job Hunting Desperation**
- Don't: "I NEED A JOB PLEASE HIRE ME!!!"
- Do: "I'm exploring new opportunities in [field]. Open to discussing roles in [area]."

### âœ… DO Post:

- Professional achievements
- Industry insights and trends
- Lessons learned from projects
- Helpful resources for your network
- Thoughtful commentary on industry news
- Congratulations to others
- Your published work or content
- Conference/event attendance
- Certifications and learning milestones
