````python
# Setup: Import required libraries
import warnings
warnings.filterwarnings('ignore')

from gtts import gTTS
import IPython.display as ipd
from pathlib import Path
import pandas as pd

def generate_audio(text, filename='audio', slow=False, lang='en'):
    """Generate audio from text for pronunciation practice."""
    audio_dir = Path('../audio')
    audio_dir.mkdir(exist_ok=True)
    
    tts = gTTS(text=text, lang=lang, slow=slow)
    output_path = audio_dir / f"{filename}.mp3"
    tts.save(str(output_path))
    
    return ipd.Audio(str(output_path), autoplay=False)

print("‚úÖ Setup complete!")
````

## 1. Formal vs. Informal English

### Understanding Register

**Register** refers to the level of formality in language. In professional settings, choosing the right register is crucial.

### Key Differences

| Aspect | Informal (Casual) | Formal (Professional) |
|--------|------------------|----------------------|
| **Contractions** | I'm, you're, we'll | I am, you are, we will |
| **Phrasal Verbs** | figure out, put off | determine, postpone |
| **Vocabulary** | kids, boss, job | children, supervisor, position |
| **Tone** | Friendly, casual | Respectful, polite |
| **Abbreviations** | ASAP, FYI, BTW | As soon as possible, For your information |
| **Idioms** | piece of cake | straightforward, easy to accomplish |
| **Sentence Structure** | Short, simple | Complete, well-structured |

### Examples: Formal vs. Informal

Let's compare how the same message changes based on register:

#### Scenario 1: Requesting Information

**‚ùå Too Informal** (Don't use with supervisors/clients):
```
Hey! Can you send me that report? Need it ASAP. Thanks!
```

**‚úÖ Professional** (Use in workplace):
```
Dear Ms. Johnson,

I hope this message finds you well. Could you please send me the quarterly report when you have a moment? I would appreciate receiving it by Friday if possible.

Thank you for your assistance.

Best regards,
[Your name]
```

#### Scenario 2: Declining a Request

**‚ùå Too Informal**:
```
Sorry, can't do it. I'm swamped right now.
```

**‚úÖ Professional**:
```
Thank you for thinking of me for this project. Unfortunately, I am currently committed to other priorities and would not be able to give this the attention it deserves. I recommend reaching out to [colleague name] who may be able to assist.
```

````python
# Listen to the professional versions
examples = {
    "Professional Request": """
    Dear Ms. Johnson, I hope this message finds you well. Could you please send me 
    the quarterly report when you have a moment? I would appreciate receiving it 
    by Friday if possible. Thank you for your assistance.
    """,
    
    "Professional Decline": """
    Thank you for thinking of me for this project. Unfortunately, I am currently 
    committed to other priorities and would not be able to give this the attention 
    it deserves.
    """
}

print("üîä PRONUNCIATION PRACTICE")
print("=" * 70)

for title, text in examples.items():
    print(f"\n{title}:")
    audio = generate_audio(text.strip(), filename=f"example_{title.replace(' ', '_')}")
    display(audio)

print("\nüí° Notice the polite, professional tone in both examples!")
````

## 2. The 4 Cs of Professional Communication

Professional communication should always be:

### 1. CLEAR üîç

**Definition**: Your message is easy to understand. No confusion about what you mean.

**How to be clear:**
- Use simple, direct language
- Define technical terms if needed
- Organize information logically
- Provide specific details and examples

**Examples:**

‚ùå **Unclear**: "We should do something about the situation soon."
- What situation?
- What action?
- When is "soon"?

‚úÖ **Clear**: "We need to implement the new security protocol by March 15th to comply with data protection regulations."
- Specific action: implement security protocol
- Specific deadline: March 15th
- Reason given: regulatory compliance

### 2. CONCISE ‚ö°

**Definition**: You use only the words necessary to convey your message. Respect others' time.

**How to be concise:**
- Remove unnecessary words and phrases
- Avoid redundancy
- Get to the point quickly
- Use bullet points for lists

**Examples:**

‚ùå **Wordy** (38 words):
```
I am writing to you today in order to inform you that we have made the decision to postpone the scheduled meeting that was originally planned for next Tuesday until a later date to be determined.
```

‚úÖ **Concise** (12 words):
```
We are postponing next Tuesday's meeting. A new date will be announced soon.
```

**Common Wordy Phrases to Avoid:**

| ‚ùå Wordy | ‚úÖ Concise |
|---------|----------|
| in order to | to |
| due to the fact that | because |
| at this point in time | now |
| in the near future | soon |
| make a decision | decide |
| give consideration to | consider |
| for the purpose of | for |

### 3. COURTEOUS ü§ù

**Definition**: You are polite and respectful, regardless of the situation.

**How to be courteous:**
- Use "please" and "thank you"
- Acknowledge others' efforts
- Show appreciation
- Be tactful when delivering bad news
- Avoid accusatory language

**Examples:**

‚ùå **Rude**:
```
Your proposal is wrong. You need to fix it and resubmit.
```

‚úÖ **Courteous**:
```
Thank you for your proposal. I have a few suggestions that might strengthen it. Would you be available to discuss revisions?
```

**Courteous Phrases:**
- "I appreciate your time..."
- "Thank you for bringing this to my attention..."
- "I would be grateful if..."
- "Would you mind..."
- "Please let me know if..."

### 4. CORRECT ‚úîÔ∏è

**Definition**: Your grammar, spelling, and facts are accurate.

**How to be correct:**
- Proofread before sending
- Check facts and figures
- Use grammar checking tools
- Verify recipient names and titles
- Double-check attachments

**Common Errors to Avoid:**

| ‚ùå Incorrect | ‚úÖ Correct |
|------------|----------|
| Your right | You're right |
| Its a good idea | It's a good idea |
| Dear Mr. Sarah | Dear Ms. Sarah |
| Myself and John | John and I |
| Could of | Could have |
| Irregardless | Regardless |

## 3. Adjusting Tone for Different Situations

### The Tone Spectrum

Even within professional communication, tone varies based on:
- **Who** you're communicating with
- **What** the situation is
- **Why** you're communicating

```
VERY FORMAL ‚Üê‚Üí MODERATELY FORMAL ‚Üê‚Üí PROFESSIONAL BUT FRIENDLY
```

### When to Use Different Tones

#### Very Formal (Cold, distant tone)
**Use when:**
- Writing to senior executives you don't know
- Formal complaints or legal matters
- Official announcements
- First contact with important clients

**Example:**
```
Dear Dr. Martinez,

I am writing to formally request approval for the proposed budget allocation. 
Please find the detailed documentation attached for your review.

Respectfully,
[Your name]
```

#### Moderately Formal (Standard professional)
**Use when:**
- Communicating with direct supervisor
- Writing to colleagues in other departments
- Client communications (ongoing relationship)
- Most business emails

**Example:**
```
Hi Sarah,

Thank you for sending the report. I've reviewed it and have a few questions about 
the data in Section 3. Could we schedule a brief call to discuss?

Best regards,
[Your name]
```

#### Professional but Friendly (Warm, approachable)
**Use when:**
- Communicating with team members
- Colleagues you work with regularly
- Less formal internal communications
- Building rapport (while still professional)

**Example:**
```
Hi Mark,

Great job on the presentation yesterday! The client was really impressed. 
Let's grab coffee this week to debrief and plan next steps.

Cheers,
[Your name]
```

## 4. Common Professional Communication Mistakes

### Mistake #1: Using Overly Casual Language

‚ùå **Don't write:**
```
Hey! Just wondering if u got my email? Lol let me know!
```

‚úÖ **Do write:**
```
Hello,

I wanted to follow up on the email I sent yesterday regarding the project timeline. 
Please let me know if you need any clarification.

Best regards,
```

**Why it matters:** Casual language can seem unprofessional and may not be taken seriously.

### Mistake #2: Being Too Wordy

‚ùå **Don't write:**
```
I am writing this email to inform you that we have reached the conclusion that it would 
be beneficial for us to have a discussion regarding the matter at hand in order to 
ensure that we are all on the same page moving forward.
```

‚úÖ **Do write:**
```
Let's discuss this matter to ensure we're aligned on next steps.
```

**Why it matters:** Busy professionals appreciate concise communication.

### Mistake #3: Not Proofreading

‚ùå **Don't write:**
```
Thank you for you're email. I will send the report to Sarah and myself tommorow.
```
(Errors: "you're" should be "your", "myself" should be "me", "tommorow" should be "tomorrow")

‚úÖ **Do write:**
```
Thank you for your email. I will send the report to Sarah and me tomorrow.
```

**Why it matters:** Errors damage your credibility and professional image.

### Mistake #4: Unclear Subject Lines

‚ùå **Don't write:**
```
Subject: Question
Subject: Hi
Subject: Important
```

‚úÖ **Do write:**
```
Subject: Budget Approval Request - Marketing Campaign Q2
Subject: Meeting Reschedule Request - Project Review (April 15)
Subject: Technical Support Needed - Database Connection Error
```

**Why it matters:** Specific subject lines help recipients prioritize and find emails later.

### Mistake #5: Emotional or Aggressive Tone

‚ùå **Don't write:**
```
I can't believe you missed ANOTHER deadline! This is completely unacceptable and I'm 
extremely frustrated. You need to get your act together immediately.
```

‚úÖ **Do write:**
```
I notice the deadline was not met. Could we discuss what challenges you're facing so 
I can provide support? Let's work together to get this back on track.
```

**Why it matters:** Emotional emails damage relationships and reflect poorly on you.

## 5. Essential Professional Phrases

### Opening Phrases

| Situation | Professional Phrase |
|-----------|--------------------|
| **First contact** | "I am writing to inquire about..." |
| **Following up** | "I wanted to follow up on..." |
| **Introducing yourself** | "My name is [X] and I am the [title] at [company]." |
| **Responding to email** | "Thank you for your email regarding..." |
| **Referencing previous communication** | "As discussed in our meeting on [date]..." |

### Making Requests

| Formality | Phrase |
|-----------|--------|
| **Very Formal** | "I would be grateful if you could..." |
| **Formal** | "Could you please..." / "Would you mind..." |
| **Standard** | "Please..." |
| **Urgent** | "I would appreciate it if you could provide this by [date/time]" |

### Giving Information

- "I am pleased to inform you that..."
- "I would like to bring to your attention..."
- "Please note that..."
- "For your information..."
- "I wanted to update you on..."

### Apologizing

| Situation | Professional Phrase |
|-----------|--------------------|
| **Delay** | "I apologize for the delay in responding." |
| **Error** | "Thank you for bringing this to my attention. I will correct it immediately." |
| **Inconvenience** | "I apologize for any inconvenience this may have caused." |
| **General** | "Please accept my apologies for..." |

### Closing Phrases

- "Please let me know if you need any further information."
- "I look forward to hearing from you."
- "Thank you for your time and consideration."
- "I appreciate your assistance with this matter."
- "Please feel free to contact me if you have any questions."

````python
# Practice pronunciation of common professional phrases
professional_phrases = [
    "I would be grateful if you could provide an update.",
    "Thank you for bringing this to my attention.",
    "I apologize for any inconvenience this may have caused.",
    "Please let me know if you need any further information.",
    "I look forward to hearing from you soon."
]

print("üîä COMMON PROFESSIONAL PHRASES")
print("=" * 70)
print("Listen and repeat these frequently-used phrases:\n")

for i, phrase in enumerate(professional_phrases, 1):
    print(f"{i}. {phrase}")
    audio = generate_audio(phrase, filename=f"professional_phrase_{i}")
    display(audio)
    print()

print("üí° Practice saying these aloud until they feel natural!")
````
