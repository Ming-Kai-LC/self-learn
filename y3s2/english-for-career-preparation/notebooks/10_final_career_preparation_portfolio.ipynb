## 1. Setup and Imports

````python
# Import required libraries
import warnings
warnings.filterwarnings('ignore')

from gtts import gTTS
import IPython.display as ipd
from pathlib import Path
import pandas as pd
from datetime import datetime

def generate_audio(text, filename='audio', slow=False, lang='en'):
    """
    Generate audio from text for pronunciation practice.
    """
    audio_dir = Path('../audio')
    audio_dir.mkdir(exist_ok=True)
    
    tts = gTTS(text=text, lang=lang, slow=slow)
    output_path = audio_dir / f"{filename}.mp3"
    tts.save(str(output_path))
    
    return ipd.Audio(str(output_path), autoplay=False)

# Get current date for portfolio
today = datetime.now().strftime("%B %d, %Y")

print("âœ… Setup complete!")
print(f"ğŸ“… Today's Date: {today}")
print("ğŸ‰ Welcome to your final project - let's showcase everything you've learned!")
````
