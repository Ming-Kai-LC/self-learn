## Part 2: Time Tracking System

### Log Every Hour

```python
def log_time(client, project, task, hours, billable=True, category='Development'):
    """
    Log a time entry to your tracking system.
    
    Args:
        client: Client name (or 'Internal' for non-billable)
        project: Project name
        task: Task description
        hours: Hours spent (can be decimal like 1.5)
        billable: True if client-billable, False for proposals/admin
        category: Type of work (Development, Communication, Proposals, Admin)
    """
    tracker_file = Path('../data/tracking/time_log.csv')
    df = pd.read_csv(tracker_file)
    
    new_entry = {
        'date': datetime.now().strftime('%Y-%m-%d'),
        'client': client,
        'project': project,
        'task_description': task,
        'hours': hours,
        'billable': 'yes' if billable else 'no',
        'category': category,
        'notes': '',
    }
    
    df = pd.concat([df, pd.DataFrame([new_entry])], ignore_index=True)
    df.to_csv(tracker_file, index=False)
    
    return {
        'entry_number': len(df),
        'client': client,
        'hours': hours,
        'billable': billable,
    }

# Example time entries
print("EXAMPLE TIME LOG ENTRIES")
print("="*70)

examples = [
    log_time('Sarah M.', 'Data Cleaning', 'Removed duplicates and handled missing values', 2.0, True, 'Development'),
    log_time('Internal', 'Job Search', 'Wrote 5 proposals for data analysis jobs', 1.0, False, 'Proposals'),
    log_time('John K.', 'Web Scraper', 'Fixed pagination bug and added error handling', 1.5, True, 'Development'),
    log_time('Sarah M.', 'Data Cleaning', 'Email exchange about final deliverable format', 0.2, False, 'Communication'),
]

for entry in examples:
    billable_status = "Billable" if entry['billable'] else "Non-billable"
    print(f"  Entry #{entry['entry_number']}: {entry['client']} - {entry['hours']} hrs ({billable_status})")

print("\nâœ“ All entries saved to: ../data/tracking/time_log.csv")
```

### Weekly Time Reports

```python
def generate_weekly_time_report():
    """
    Generate report of time usage for current week.
    """
    tracker_file = Path('../data/tracking/time_log.csv')
    
    if not tracker_file.exists():
        return "No time entries logged yet."
    
    df = pd.read_csv(tracker_file)
    
    if len(df) == 0:
        return "No time entries logged yet."
    
    # Convert date
    df['date'] = pd.to_datetime(df['date'])
    
    # Get current week (last 7 days)
    today = datetime.now()
    week_ago = today - timedelta(days=7)
    df_week = df[df['date'] >= week_ago]
    
    if len(df_week) == 0:
        return "No time entries in the last 7 days."
    
    # Calculate totals
    total_hours = df_week['hours'].sum()
    billable_hours = df_week[df_week['billable'] == 'yes']['hours'].sum()
    non_billable_hours = total_hours - billable_hours
    billable_percentage = (billable_hours / total_hours * 100) if total_hours > 0 else 0
    
    # By category
    by_category = df_week.groupby('category')['hours'].sum().to_dict()
    
    # By client (billable only)
    billable_df = df_week[df_week['billable'] == 'yes']
    by_client = billable_df.groupby('client')['hours'].sum().to_dict()
    
    report = f"""
WEEKLY TIME REPORT ({week_ago.strftime('%Y-%m-%d')} to {today.strftime('%Y-%m-%d')})
{'='*70}

Summary:
  Total hours logged: {total_hours:.1f}
  Billable hours: {billable_hours:.1f} ({billable_percentage:.0f}%)
  Non-billable hours: {non_billable_hours:.1f}

By Category:
"""
    
    for category, hours in sorted(by_category.items(), key=lambda x: x[1], reverse=True):
        pct = (hours / total_hours * 100)
        report += f"  {category}: {hours:.1f} hrs ({pct:.0f}%)\n"
    
    if by_client:
        report += "\nBy Client (Billable):\n"
        for client, hours in sorted(by_client.items(), key=lambda x: x[1], reverse=True):
            report += f"  {client}: {hours:.1f} hrs\n"
    
    # Weekly target check
    report += f"\nWeekly Target: 5.0 hours\n"
    if total_hours >= 5:
        report += f"âœ“ Target met! ({total_hours:.1f} hours logged)\n"
    else:
        report += f"âš  {5 - total_hours:.1f} hours remaining this week\n"
    
    # Efficiency insight
    if billable_percentage >= 80:
        report += "\nâœ“ Excellent billable percentage (80%+)\n"
    elif billable_percentage >= 60:
        report += "\nâ†’ Good billable percentage, aim for 80%+\n"
    else:
        report += "\nâš  Low billable percentage - reduce proposal/admin time\n"
    
    return report

# Generate report
print(generate_weekly_time_report())
```
